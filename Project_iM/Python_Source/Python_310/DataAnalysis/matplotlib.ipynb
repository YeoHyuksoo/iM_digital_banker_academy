## Matplotlib 실습

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
from matplotlib import font_manager, rc
import platform
#import matplotlib.font_manager

font_list = [f.name for f in matplotlib.font_manager.fontManager.ttflist if 'Nanum' in f.name]
print(font_list)

plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Windows':
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    print(font_name)
    rc('font', family=font_name)
elif platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Linux':
    rc('font', family='NanumBarunGothic')
else:
    print('Unknown system... sorry~~~~~~')
```

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("LaTeX, 문자열로 tick label 정의")
plt.plot(X, C)

plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])
plt.grid(True)
plt.show()
```

```python
t = np.arange(0., 5., 0.2)
plt.title("Line Plot")
plt.plot(t, t, 'r-.', t, 0.5 * t**2, 'b2:', t, 0.2 * t**3, 'g^-', linewidth = 2)
plt.legend(['Red', 'Blue', 'Green'], loc = 'best')
plt.show()
```

```python
x1 = np.linspace(start = 0.0, stop = 5.0, num = 50)
x2 = np.linspace(0.0, 2.0, 50)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.suptitle('Main title', fontsize = 15)

plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'yo-')
plt.xlabel('time (s)')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.title('A tale of 2 subplots')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

```python
x = np.linspace(-3, 3, 100)

y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)

fig, ax = plt.subplots(2, 2, figsize = (10, 10))
ax[0, 0].plot(x, y1, 'r--')
ax[0, 1].plot(x, y2)
ax[1, 0].plot(x, y3)
ax[1, 1].plot(x, y4)

ax[0, 0].set_title("Sine function")
ax[0, 1].set_title("Cosine function")
ax[1, 0].set_title("Tan function")
ax[1, 1].set_title("Exponential function")

fig.suptitle("Trigonometric functions", fontsize = 16)
fig.tight_layout()
plt.show()
```

```python
fig, ax0 = plt.subplots()
ax1 = ax0.twinx()

ax0.set_title("2개의 x축 공유 figures")
ax0.plot([10, 5, 2, 9, 7], 'r-', label = "y0")
ax0.set_ylabel("y0")
ax0.set_xlabel("common x-axis")
ax0.grid(True)

ax1.plot([100, 200, 220, 180, 120], 'g:', label="y1")
ax1.set_ylabel("y1")
#ax1.grid(True)

plt.show()
```

```python
font = {'family' : font_name, 'color' : 'red', 'size' : 20}
font2 = {'family' : font_name, 'color' : 'black', 'size' : 11}

np.random.seed(0)
X = np.random.normal(0, 1, 100)
Y = np.random.normal(0, 1, 100)

plt.title("Scatter Plot(산점도)", fontdict = font)
plt.scatter(X, Y)
plt.xlabel('x축', fontdict = font2)
plt.ylabel('y축', fontdict = font2)
plt.show()
```

```python
def f(x, y) : 
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
XX, YY = np.meshgrid(x, y)
ZZ = f(XX, YY)

plt.title("Contour")
plt.contourf(XX, YY, ZZ, alpha = .77, cmap = 'gnuplot2')

plt.contour(XX, YY, ZZ, colors = 'green', size = 100)

plt.show()
```

```python
y1 = [70, 90, 80]
y2= [60, 70, 90]

x = np.arange(len(y1))
err1 = np.abs(np.random.randn(3)) + 5
err2 = np.abs(np.random.randn(3)) + 3
ylabel = ['국어', '영어', '수학']

plt.title("Bar Chart")
plt.barh(x, y1, xerr = err1, ecolor = 'yellow', height = 0.4, color = 'red', align = 'center', label = '점수')
plt.barh(x + 0.4, y2, xerr = err2, height = 0.4, color = 'blue', align = 'center', label = '반평균')
plt.yticks(x + 0.4 / 2, ylabel)
plt.xlim(0, 100)
plt.xlabel("점수")
plt.ylabel("과목")
plt.legend()
plt.show()
```

```python
labels = ['frog', 'pig', 'dog', 'wood']

sizes = [15.55, 30, 45, 9.45]

colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']

explode = (0.1, 0.4, -0.2, 0.2)
plt.title("Pie Chart")
plt.pie(sizes, explode = explode, labels = labels, colors = colors, autopct = '%1.2f%%',
        shadow = True, startangle = 180, counterclock = False)
plt.axis('equal')
plt.show()
```
