```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.alert import Alert
from selenium.common.exceptions import UnexpectedAlertPresentException, NoAlertPresentException
import time
import re
import urllib.parse

# 웹드라이버 설정
# driver_path = "chromedriver.exe"  # ChromeDriver 경로
driver = webdriver.Chrome()

try:
    # 카테고리 페이지 열기
    url = "https://category.gmarket.co.kr/listview/L100000032.aspx"
    driver.get(url)
    time.sleep(3)

    # 인기상품 5위의 li 태그 찾기
    item_xpath = "//li[.//div[contains(@class, 'best-num') and text()='5']]"
    item_element = driver.find_element(By.XPATH, item_xpath)

    # 상품의 href 속성에서 URL 추출
    link_element = item_element.find_element(By.CSS_SELECTOR, "a")
    href_attribute = link_element.get_attribute("href")
    print("href 속성 내용:", href_attribute)

    # http로 시작하는 URL 추출 및 디코딩
    match = re.search(r"http[s]?://[^\s'\"]+", href_attribute)
    if match:
        full_url = match.group(0)
        trimmed_url = full_url.split(",")[0]
        decoded_url = urllib.parse.unquote(trimmed_url)
        actual_url = decoded_url.rstrip("'")
        print("실제 상품 페이지 URL:", actual_url)

        # 상품 페이지로 이동
        driver.get(actual_url)
        time.sleep(3)  # 페이지 로드 대기

        # 경고창 처리
        try:
            alert = Alert(driver)
            print(f"경고창이 나타났습니다: {alert.text}")
            alert.accept()  # 경고창 닫기
        except NoAlertPresentException:
            print("경고창이 없습니다.")

        # 현재 페이지 URL 출력
        print("현재 페이지 URL:", driver.current_url)
    else:
        print("http로 시작하는 URL을 찾을 수 없습니다.")

except UnexpectedAlertPresentException:
    alert = Alert(driver)
    print(f"예기치 않은 경고창이 나타났습니다: {alert.text}")
    alert.accept()  # 경고창 닫기

finally:
    driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.alert import Alert
from selenium.common.exceptions import UnexpectedAlertPresentException, NoAlertPresentException
import time
import re
import urllib.parse

# 웹드라이버 설정
driver = webdriver.Chrome()

# 카테고리 딕셔너리
categories = {
    "가전디지털": "https://category.gmarket.co.kr/listview/L100000055.aspx",
    "완구취미": "https://category.gmarket.co.kr/listview/L100000005.aspx",
    "도서음반DVD": "https://category.gmarket.co.kr/listview/L100000009.aspx",
    "주방용품": "https://category.gmarket.co.kr/listview/L100000063.aspx",
    "스포츠레저": "https://category.gmarket.co.kr/listview/L100000021.aspx",
    "홈인테리어": "https://category.gmarket.co.kr/listview/L100000022.aspx",
    "반려동물용품": "https://category.gmarket.co.kr/listview/L100000045.aspx",
}

# 카테고리별로 반복
for category_name, category_url in categories.items():
    try:
        print(f"크롤링 중: {category_name} 카테고리")
        driver.get(category_url)
        time.sleep(3)  # 페이지 로드 대기

        # 1위부터 10위까지 반복
        for rank in range(1, 11):
            try:
                # 순위에 해당하는 상품의 li 태그 찾기
                item_xpath = f"//li[.//div[contains(@class, 'best-num') and text()='{rank}']]"
                item_element = driver.find_element(By.XPATH, item_xpath)

                # 상품의 href 속성에서 URL 추출
                link_element = item_element.find_element(By.CSS_SELECTOR, "a")
                href_attribute = link_element.get_attribute("href")
                print(f"{rank}위 상품의 href 속성 내용:", href_attribute)

                # http로 시작하는 URL 추출 및 디코딩
                match = re.search(r"http[s]?://[^\s'\"]+", href_attribute)
                if match:
                    full_url = match.group(0)
                    trimmed_url = full_url.split(",")[0]
                    decoded_url = urllib.parse.unquote(trimmed_url)
                    actual_url = decoded_url.rstrip("'")
                    print(f"{rank}위 상품 페이지 URL:", actual_url)

                    # 상품 페이지로 이동
                    driver.get(actual_url)
                    time.sleep(3)  # 페이지 로드 대기

                    # 경고창 처리
                    try:
                        alert = Alert(driver)
                        print(f"경고창이 나타났습니다: {alert.text}")
                        alert.accept()  # 경고창 닫기
                    except NoAlertPresentException:
                        print("경고창이 없습니다.")

                    # 현재 페이지 URL 출력
                    print("현재 페이지 URL:", driver.current_url)

                    # 뒤로 가기
                    driver.back()
                    time.sleep(2)  # 페이지 로드 대기
                else:
                    print(f"{rank}위 상품 URL을 찾을 수 없습니다.")
            except Exception as e:
                print(f"{rank}위 상품 크롤링 중 오류 발생: {e}")
                continue
    except Exception as e:
        print(f"{category_name} 카테고리 크롤링 중 오류 발생: {e}")
        continue

# finally:
driver.quit()
```
