```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#  
df = pd.read_csv('fake_reviews_dataset.csv')

# AI   
ai_reviews = df[df['label'] == 'CG']['text_']
human_reviews = df[df['label'] == 'OR']['text_']

#      (: 2000)
sample_size = 2000
ai_reviews_sample = ai_reviews.sample(n=sample_size, random_state=42)
human_reviews_sample = human_reviews.sample(n=sample_size, random_state=42)

# TF-IDF 
vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)
ai_tfidf = vectorizer.fit_transform(ai_reviews_sample)
human_tfidf = vectorizer.fit_transform(human_reviews_sample)

#   ( )
ai_similarity = cosine_similarity(ai_tfidf)
human_similarity = cosine_similarity(human_tfidf)

# AI ,      ( )
#       
ai_similarity_values = ai_similarity[np.triu_indices_from(ai_similarity, k=1)]
human_similarity_values = human_similarity[np.triu_indices_from(human_similarity, k=1)]

# : AI      
plt.figure(figsize=(8, 6))
sns.boxplot(data=[ai_similarity_values, human_similarity_values])
plt.title('Distribution of Cosine Similarity between AI and Human Reviews')
plt.ylabel('Cosine Similarity')

# x  
plt.xticks([0, 1], ['AI Reviews', 'Human Reviews'])

plt.show()

# : AI      
plt.figure(figsize=(8, 6))
plt.hist(ai_similarity_values, bins=50, alpha=0.6, label='AI Reviews', color='blue')
plt.hist(human_similarity_values, bins=50, alpha=0.6, label='Human Reviews', color='orange')
plt.title('Cosine Similarity Distribution between AI and Human Reviews')
plt.xlabel('Cosine Similarity')
plt.ylabel('Frequency')
plt.legend(loc='upper right')
plt.show()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

#  
df = pd.read_csv('fake_reviews_dataset.csv')

# AI   
ai_reviews = df[df['label'] == 'CG']['text_']
human_reviews = df[df['label'] == 'OR']['text_']

#      (: 2000)
sample_size = 2000
ai_reviews_sample = ai_reviews.sample(n=sample_size, random_state=42)
human_reviews_sample = human_reviews.sample(n=sample_size, random_state=42)

# TF-IDF 
vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)
ai_tfidf = vectorizer.fit_transform(ai_reviews_sample)
human_tfidf = vectorizer.fit_transform(human_reviews_sample)

#   ( )
ai_similarity = cosine_similarity(ai_tfidf)
human_similarity = cosine_similarity(human_tfidf)

# AI    ( )
ai_similarity_values = ai_similarity[np.triu_indices_from(ai_similarity, k=1)]

#     ( )
human_similarity_values = human_similarity[np.triu_indices_from(human_similarity, k=1)]

# AI    10
top_10_ai_similarity = np.sort(ai_similarity_values)[-10:]

#     10
top_10_human_similarity = np.sort(human_similarity_values)[-10:]

#   10 
print("Top 10 AI Review Similarities:", top_10_ai_similarity)
print("Top 10 Human Review Similarities:", top_10_human_similarity)

#      (AI  )
ai_similarity_avg = np.mean(ai_similarity_values)
human_similarity_avg = np.mean(human_similarity_values)

print("\nAverage Cosine Similarity for AI Reviews:", ai_similarity_avg)
print("Average Cosine Similarity for Human Reviews:", human_similarity_avg)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#    (      )
plt.rc('font', family='Malgun Gothic')  #  
plt.rc('axes', unicode_minus=False)  #    

plt.figure(figsize=(10, 5))
ax = sns.barplot(x=np.arange(1, 11), y=top_10_ai_similarity, palette="Blues_d")
plt.title('AI    10', fontsize=14, weight='bold')
plt.xlabel(' ', fontsize=12, weight='bold')
plt.ylabel(' ', fontsize=12, weight='bold')
plt.xticks(ticks=np.arange(10), labels=[f'top{10-i}' for i in range(10)])  # top10 top1 
plt.ylim(0, 1)

#    
for i, v in enumerate(top_10_ai_similarity):
    ax.text(i, v + 0.01, f'{v:.2f}', ha='center', fontsize=10, weight='bold')

plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 5))
ax = sns.barplot(x=np.arange(1, 11), y=top_10_human_similarity, palette="Greens_d")
plt.title('    10', fontsize=14, weight='bold')
plt.xlabel(' ', fontsize=12, weight='bold')
plt.ylabel(' ', fontsize=12, weight='bold')
plt.xticks(ticks=np.arange(10), labels=[f'top{10-i}' for i in range(10)])  # top10 top1 
plt.ylim(0, 1)

#    
for i, v in enumerate(top_10_human_similarity):
    ax.text(i, v + 0.01, f'{v:.2f}', ha='center', fontsize=10, weight='bold')

plt.tight_layout()
plt.show()

```

```python
#   
avg_ai_similarity = ai_similarity[np.triu_indices_from(ai_similarity, k=1)].mean()
avg_human_similarity = human_similarity[np.triu_indices_from(human_similarity, k=1)].mean()

#   
# plt.figure(figsize=(6, 4))
colors = ['skyblue', 'green']
sns.barplot(x=['AI   ', '   '], y=[avg_ai_similarity, avg_human_similarity], palette=colors)
plt.title('AI      ', fontsize=14, weight='bold')
plt.ylabel(' ', fontsize=12, weight='bold')

#    
for i, v in enumerate([avg_ai_similarity, avg_human_similarity]):
    plt.text(i, v, f'{v:.4f}', ha='center', fontsize=10, weight='bold')

plt.ylim(0, 0.03)
plt.tight_layout()
plt.show()
```
