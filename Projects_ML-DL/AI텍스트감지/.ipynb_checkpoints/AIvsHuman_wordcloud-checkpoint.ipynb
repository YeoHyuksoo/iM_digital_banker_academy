### Konlpy Kkma

```python
from konlpy.tag import Kkma

kkma = Kkma(max_heap_size=1024*12)
```

```python
human = train[train['label'] == 0]
ai = train[train['label'] == 1]
```

```python
human.drop(index=2665, axis=0, inplace=True)
```

```python
human = human['text'].to_list()
ai = ai['text'].to_list()
```

```python
# human_sentence = []

# for i in range(len(human)):
#     human_sentence.append(Kkma().sentences(human[i]))
#     print(i)
```

```python
# ai_sentence = []

# for i in range(len(ai)):
#     ai_sentence.append(Kkma().sentences(ai[i]))
#     print(i)
```

```python
human = ' '.join(human)
ai = ' '.join(ai)
```

```python
import re

pattern = r'[^a-zA-Z가-힣0-9]'
human_cl = re.sub(pattern=pattern, repl=' ', string=human)
ai_cl = re.sub(pattern=pattern, repl=' ', string=ai)
```

```python
human_noun = Kkma().nouns(human_cl)
```

```python
ai_noun = Kkma().nouns(ai_cl)
```

```python
from collections import Counter

human_count = Counter(human_noun)
ai_count = Counter(ai_noun)
```

```python
human_count.most_common()
```

```python
for x,y in list(human_count.items()):
    if (len(x) == 1 or len(x) >= 6):
        del human_count[x]

# 한글자 or 6글자 이상 (에어프라이어 등) 제외
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

font_path = '/Users/haegyul/Library/Fonts/SCDream4.otf'
wordcloud = WordCloud(font_path=font_path,
                     colormap = 'PuBu',
                      background_color='white').generate_from_frequencies(human_count)

plt.figure(figsize=(8,4))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
