```python
#### 주요한 파라미터
- C
    - 잘못 분류된 데이터로부터 해당 클래스의 결정 경계까지의 거리로 데이터점이 잘못 분류된 정도를 나타낸다.
    - SVM 모델이 오류를 어느정도 허용할 것인지 파라미터 C를 통해 지정
    - C값이 클수록 하드마진(오류 허용 안 함), 작을수록 소프트마진(오류를 허용함)이다.
- gamma
    - 결정 경계를 얼마나 유연하게 그을 것인지 정해주는 역할
    - gamma값을 높이면 학습 데이터에 많이 의존해서 결정 경계를 구불구불 긋게 된다. 오버피팅을 초래
    - gamma값을 낮추면 학습 데이터에 별로 의존하지 않고 결정 경계를 직선에 가깝게 긋게 된다. 언더피팅 초래
C와 gamma을 지나치게 높게 잡는다면 오버피팅이 될 수 있다.
- kernel
    - 판별 경계를 찾기위한 내적 함수
    - 'linear', 'poly', 'rbf', 'sigmoid', 'precomputed'.default는 rbf
```

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

plt.rcParams['axes.unicode_minus'] = False
plt.rcParams["font.family"] = 'Malgun Gothic'
plt.title("세개의 클러스터를 가진 가상 데이터")
X, y = make_blobs(n_samples = 300, n_features = 2, centers = 3, random_state = 1)
plt.scatter(X[:, 0], X[:, 1], marker = 'o', c = y, s = 100, edgecolor = 'k', linewidth = 2)
plt.xlabel("$X_1$")
plt.ylabel("$X_2$")
plt.show()
```

```python
from sklearn.svm import SVC

model = SVC(kernel = 'linear', C = 1e10).fit(X, y)
```

```python
model
```

SVC를 사용하여 모형을 구하면 다음과 같은 속성값을 가진다.
• n_support_ : 각 클래스의 서포트 벡터의 개수
• support_ : 각 클래스의 서포트 벡터의 인덱스
• support_vectors_: 각 클래스의 서포트의 x값.
• coef_ : 기울기
• intercept_: 절편
• dual_coef_: 각 원소가 aiyi로 이루어진 벡터. 음의 계수와 양의 계수를 나타낸다.

```python
import numpy as np

xmin = X[:, 0].min()
xmax = X[:, 0].max()
ymin = X[:, 1].min()
ymax = X[:, 1].max()

xx = np.linspace(xmin, xmax, 10)
yy = np.linspace(ymin, ymax, 10)
X1, X2 = np.meshgrid(xx, yy)

Z = np.empty(X1.shape)
for (i,j), val in np.ndenumerate(X1):
    x1 = val
    x2 = X2[i,j]
    p = model.decision_function([[x1, x2]])
    Z[i,j] = p[0]

levels = [-1, 0, 1]
linestyles = ['dashed', 'solid', 'dashed']

plt.scatter(X[y == -1, 0], X[y == -1, 1], marker = 'o', label = '-1클래스')
plt.scatter(X[y == 1, 0], X[y == 1, 1], marker = 'x', label = '+1클래스')
plt.contour(X1, X2, Z, levels, colors = 'k', linestyles = linestyles)
plt.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1], s = 300, alpha = 0.3)

# 테스트 데이터 포인트 추가
x_new = [10, 2]  # 새로운 데이터 포인트
plt.scatter(x_new[0], x_new[1], marker='^', s=100)  # 테스트 데이터 포인트 표시
plt.text(x_new[0] + 0.03, x_new[1] + 0.08, '테스트 데이터')  # 라벨 추가

# 6. 그래프 설정
plt.xlabel('x1')  # x축 라벨
plt.ylabel('x2')  # y축 라벨
plt.legend() # 범례 추가
plt.title('SVM 예측 결과')  # 그래프 제목
plt.show()  # 그래프 출력
```

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.datasets import load_iris
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

iris = load_iris()
iris.target
```

```python
X_data = iris.data[(iris.target == 0) | (iris.target == 1), :2]
X_data[:5]
```

```python
X_data.shape
```

```python
y = iris.target[(iris.target == 0) | (iris.target == 1)]
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_data, y, test_size = 0.3, random_state = 10)
```

```python
svm = SVC(kernel = "linear", C = 1e10)
svm.fit(X_train, y_train)
```

```python
pred_y = svm.predict(X_test)
confusion_matrix(y_test, pred_y)
```

```python
print(classification_report(y_test, pred_y))
```

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import pandas as pd
```

```python
svm = SVC(kernel='linear', C=1e10)
svm.fit(X_train, y_train)


pred_y = svm.predict(X_test)
confusion_matrix(y_test, pred_y)

print(classification_report(y_test, pred_y))

df = pd.read_csv('boston_house.csv')
df.head()

x = df.drop('Target', axis=1)
y = df['Target']

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3)

svr = SVR(kernel='linear', gamma='auto')
svr.fit(xtrain, ytrain)

print('가중치 : ', svr.coef_)
print('절펀향 : ', svr.intercept_)

score = svr.score(xtest, ytest)
print('R-squared : ', score)

ypred = svr.predict(xtest)
mse = mean_squared_error(ytest, ypred)
print('MSE : ', mse)
```

```python
from sklearn.metrics import r2_score
r2_score(ytest, ypred)
```

```python
x_ax = range(len(ytest))
plt.plot(x_ax, ytest, label = 'original')
plt.plot(x_ax, ypred, label = 'predicted')
plt.title("Boston test and predicted data")
plt.legend()
plt.show()
```
