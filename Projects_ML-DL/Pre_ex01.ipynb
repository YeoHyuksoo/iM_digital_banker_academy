```python
import seaborn as sns

iris = sns.load_dataset("iris")
x = iris.iloc[:,:-1]
x.head()
```

```python
iris.info()
```

```python
import random

for col in range(4):
    x.iloc[[random.sample(range(len(iris)), 10)],col] = float('nan')
x.head()
```

```python
x.mean(axis=0) 
```

```python
from sklearn.impute import SimpleImputer
import numpy as np

imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
imp_mean.fit_transform(x)[0:5,:]
```

```python
x.median(axis=0)
```

```python
imp_median = SimpleImputer(missing_values=np.nan, strategy='median')
imp_median.fit_transform(x)[0:5,:]
```

### **전처리 실습 예제(타이타닉 데이터 이용)**
- 결측값 처리
    - Age : 평균값으로 결측값 처리
    - Embarked : 최빈값으로 결측값 처리
- 레이블 인코딩
    - Sex
    - Embarked
- 원핫인코딩
    - Sex
    - Embarked

```python
import pandas as pd
from sklearn.impute import SimpleImputer
import numpy as np

df = pd.read_csv('titanic_train.csv')
df.head(2)
```

```python
df.info()
```

```python
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
age_mean = imp_mean.fit_transform(df[["Age"]])
age_mean
```

```python
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
df["Age"]= imp_mean.fit_transform(df[["Age"]])
df
```

```python
imp_most = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
embarked_mean = imp_most.fit_transform(df[["Embarked"]])
embarked_mean
```

```python
df["Embarked"] = embarked_mean
df
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
res1 = le.fit_transform(df[["Sex"]])
print(le.classes_)
```

```python
res1
```

```python
df["Sex"] = res1
df
```

```python
import pandas as pd
from sklearn.impute import SimpleImputer
import numpy as np

df = pd.read_csv('./titanic_train.csv')
df.head(2)
```

```python
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder()
res3 = enc.fit_transform(df[["Sex"]].copy())
print(enc.categories_)
res3.toarray()
```

```python
from sklearn.impute import SimpleImputer
import  numpy as np

imp_most = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
embarked_mean = imp_most.fit_transform(df[["Embarked"]])
embarked_mean
```

```python
df["Embarked"] = embarked_mean
```

```python
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder()
res4 = enc.fit_transform(df[["Embarked"]].copy())
print(enc.categories_)
res4.toarray()
```

```python
df2 = pd.DataFrame(res4.toarray(), columns=["Embarked_C", "Embarked_Q", "Embarked_S"])
df2
```

```python
df3 = df.drop(["Embarked"], axis=1)
df3
```

```python
df4 = pd.concat([df3, df2], axis=1)
df4
```
