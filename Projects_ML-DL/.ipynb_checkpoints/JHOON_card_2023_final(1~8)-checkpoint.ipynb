```python
import pandas as pd
from IPython.display import display

# ì—…ì¢… ë¶„ë¥˜ í•¨ìˆ˜
def classify_category(business_type):
    categories = {
        "ëŒ€ì¤‘êµí†µ": {
            "ì² ë„": ["ì² ë„", "KTX"],
            "ê³ ì†ë²„ìŠ¤": ["ê³ ì†ë²„ìŠ¤"],
            "íƒì‹œ": ["íƒì‹œ", "íƒì‹œíšŒì‚¬"],
            "ëŒ€ì¤‘êµí†µ": ["ë²„ìŠ¤", "ì§€í•˜ì² "]
        },
        "ë°°ë‹¬ì•±": ["ë°°ë‹¬ì˜ë¯¼ì¡±", "ìš”ê¸°ìš”", "ì¿ íŒ¡ì´ì¸ "],
        "OTT (ìŠ¤íŠ¸ë¦¬ë°)": ["ë„·í”Œë¦­ìŠ¤", "ë””ì¦ˆë‹ˆ+", "ì›¨ì´ë¸Œ", "í‹°ë¹™", "ì™“ì± "],
        "ì´ë™í†µì‹ , í†µì‹ ìš”ê¸ˆ": {
            "ì´ë™í†µì‹ ìš”ê¸ˆ": ["ì´ë™í†µì‹ ìš”ê¸ˆ"],
            "ktí†µì‹ ": ["ì´ë™í†µì‹ ìš”ê¸ˆ", "í†µì‹ ê¸°ê¸°", "í†µì‹ ì„œë¹„ìŠ¤/ì†Œë“ê³µì œë¹„ëŒ€ìƒ", "í†µì‹ íŒë§¤ì—…1", "ì¼ë°˜(í†µì‹ íŒë§¤)", "ìœ„ì„±ë°©ì†¡"]
        },
        "ì˜í™”": ["ì˜í™”ê´€", "ë¡¯ë°ì‹œë„¤ë§ˆ", "ë©”ê°€ë°•ìŠ¤"],
        "ì»¤í”¼": {
            "ê¸°íƒ€ìŒë£Œì‹í’ˆ": ["ê¸°íƒ€ìŒë£Œì‹í’ˆ"],
            "ì»¤í”¼ë¸Œëœë“œ": ["ìŠ¤íƒ€ë²…ìŠ¤", "ì»¤í”¼ë¹ˆ", "ì—”ì ¤ë¦¬ë„ˆìŠ¤", "ì‹œì• í‹€", "Sleepless In Seattle", "ì¹´í˜ë² ë„¤"]
        },
        "í¸ì˜ì ": {
            "ì¼ë°˜í¸ì˜ì ": ["GS25", "CU", "ì„¸ë¸ì¼ë ˆë¸", "ë°”ì´ë”ì›¨ì´", "ë¯¸ë‹ˆìŠ¤í†±", "ë‹¤ì´ì†Œ", "ë¯¸ë‹ˆìƒµ"]
        },
        "ì˜¨ë¼ì¸ì‡¼í•‘": {
            "ì¢…í•©ì‡¼í•‘ëª°": ["Gë§ˆì¼“", "ì˜¥ì…˜", "11ë²ˆê°€", "ì¿ íŒ¡", "ìœ„ë©”í”„", "í‹°ëª¬", "í˜ì´ë¶ì‡¼í•‘", "ì¸í„°ë„·Mall", "ì¸í„°ë„·ì¢…í•©Mall", "PGìƒí’ˆê¶Œ"],
            "ë°±í™”ì ëª°": ["ì‹ ì„¸ê³„ëª°", "ë¡¯ë°ë‹·ì»´", "ê°¤ëŸ¬ë¦¬ì•„ëª°", "AKëª°"],
            "í™ˆì‡¼í•‘": ["GSìƒµ", "CJì˜¤ì‡¼í•‘", "ë¡¯ë°í™ˆì‡¼í•‘", "í˜„ëŒ€Hëª°", "NSí™ˆì‡¼í•‘", "í™ˆì•¤ì‡¼í•‘", "ë¡¯ë°iëª°"],
            "ëŒ€í˜•ë§ˆíŠ¸ëª°": ["ì´ë§ˆíŠ¸ëª°", "í™ˆí”ŒëŸ¬ìŠ¤ëª°", "ë¡¯ë°ë§ˆíŠ¸ëª°", "ë¡¯ë°ìŠˆí¼ëª°"],
            "ë©´ì„¸ì ëª°": ["ì‹ ë¼ë©´ì„¸ì ", "ë¡¯ë°ë©´ì„¸ì ", "ë™í™”ë©´ì„¸ì ", "ì›Œì»¤íë©´ì„¸ì "]
        },
        "ë°±í™”ì ": ["ìì‚¬ì¹´ë“œë°œí–‰ë°±í™”ì ", "ë°±í™”ì "],
        "ëŒ€í˜•ë§ˆíŠ¸": ["ëŒ€í˜•í• ì¸ì ", "ë†í˜‘í•˜ë‚˜ë¡œí´ëŸ½", "ë†,ì¶•í˜‘ì§ì˜ë§¤ì¥", "ì—°ì‡„ì ", "ëŒ€í˜•ë§ˆíŠ¸"],
        "ìŠˆí¼ë§ˆì¼“": ["ìŠˆí¼ë§ˆì¼“"],
        "ë©´ì„¸ì ": ["ì‹ ë¼", "ë¡¯ë°", "ë™í™”", "ì›Œì»¤í", "ë©´ì„¸ì "],
        "í™”ì¥í’ˆ": ["í™”ì¥í’ˆ"],
        "ì˜ë¥˜": ["ì¼ë°˜ì˜ë¥˜", "ìŠ¤í¬ì¸ ", "ê³¨í”„", "ì•„ì›ƒë„ì–´"],
        "ì•ˆê²½": ["ì•ˆê²½"],
        "í•´ì™¸ê²°ì œ": ["í•´ì™¸ ì§êµ¬", "í•´ì™¸ ê°€ë§¹ì "],
        "ì™¸ì‹, ìŒì‹ì ": ["í•œì‹", "ì–‘ì‹", "ì¤‘ì‹", "ì¼ì‹", "íŒ¨ë°€ë¦¬ ë ˆìŠ¤í† ë‘", "ì•„ì›ƒë°±", "TGIF", "VIPS"],
        "ì£¼ìœ ": ["ì£¼ìœ ì†Œ", "GSì¹¼í…ìŠ¤", "GSì£¼ìœ ì†Œ", "SKì£¼ìœ ì†Œ", "í˜„ëŒ€ì˜¤ì¼ë±…í¬", "S-OIL", "LPGì¶©ì „ì†Œ"],
        "ìë™ì°¨ ê´€ë ¨": ["ìë™ì°¨ì •ë¹„", "ì„¸ì°¨ì¥", "ë Œí„°ì¹´"],
        "ë ˆì € ë° ì·¨ë¯¸": ["ê³¨í”„ì¥", "ê³¨í”„ì—°ìŠµì¥", "ìŠ¤í¬ë¦°ê³¨í”„", "ë‹¹êµ¬ì¥", "ë³¼ë§ì¥", "ìˆ˜ì˜ì¥", "ë…¸ë˜ë°©", "í—¬ìŠ¤í´ëŸ½", "ìŠ¤í¬ì¸ ì„¼í„°"],
        "ìˆ™ë°•": ["í˜¸í…”", "ëª¨í…”", "ë¦¬ì¡°íŠ¸", "ì½˜ë„", "ì¹´ë¼ë°˜", "ê¸€ë¨í•‘", "íŒ¬ì…˜"],
        "ë³´í—˜": ["ìƒëª…ë³´í—˜", "ì†í•´ë³´í—˜"],
        "ê±´ì¶• ë° ì¸í…Œë¦¬ì–´": ["ê°€êµ¬", "ì¸í…Œë¦¬ì–´", "ì¡°ëª…", "ëª©ì¬", "ì² ë¬¼", "ê±´ì¶•ìì¬"],
        "ìœ¡ì•„/êµìœ¡": ["ì–´ë¦°ì´ì§‘", "ìœ ì¹˜ì›", "í•™ì›"],
        "ì„œì ": ["êµë³´ë¬¸ê³ ", "ì˜í’ë¬¸ê³ ", "YES24", "ì˜ê´‘ì„œì "],
        "ë†€ì´ê³µì›": ["ì—ë²„ëœë“œ", "ë¡¯ë°ì›”ë“œ", "ì„œìš¸ëœë“œ", "ê²½ì£¼ì›”ë“œ", "í†µë„í™˜íƒ€ì§€ì•„"],
        "ìƒí™œí¸ì˜": ["ì„œì ", "í•™ì›", "ì™¸ì‹"],
        "ì˜ë£Œ": ["ë³‘ì›", "ì˜ì›", "ì¡°ì‚°ì›", "ì‚°í›„ì¡°ë¦¬ì›"],
        "ì•„íŒŒíŠ¸ê´€ë¦¬ë¹„": ["ì•„íŒŒíŠ¸ê´€ë¦¬ë¹„"],
        "ì•½êµ­": ["ì•½êµ­"],
        "ë² ì´ì»¤ë¦¬/ë””ì €íŠ¸": ["ë˜í‚¨ë„ë„ˆì¸ ", "ë² ìŠ¤í‚¨ë¼ë¹ˆìŠ¤"],
        "íŒ¨ì…˜": ["ì—¬ì„±í¬ë¡œì»¤ë‹¤ì¼", "ìƒ¤íŠ¸ë Œ", "ì˜¬ë¦¬ë¹„ì•„í•˜ìŠ¬ëŸ¬", "ì™€ì¼ë“œë¡œì¦ˆ"],
        "ì—¬í–‰": ["êµ­ë‚´ì—¬í–‰", "í•´ì™¸ì—¬í–‰"]
    }

    for category, subcategories in categories.items():
        if isinstance(subcategories, dict):  # ì„œë¸Œ ì¹´í…Œê³ ë¦¬ ì¡´ì¬
            for subcategory, keywords in subcategories.items():
                if business_type in keywords:
                    return category
        elif isinstance(subcategories, list):  # ë¦¬ìŠ¤íŠ¸ í˜•ì‹ì˜ ë‹¨ì¼ ì¹´í…Œê³ ë¦¬
            if business_type in subcategories:
                return category

    return "ê¸°íƒ€ (ë¶„ë¥˜ ë¶ˆê°€)"

# iM_ì˜¬ë ˆ_Super_DC_ì¹´ë“œ
# 1. í• ì¸ í•œë„ ê³„ì‚° í•¨ìˆ˜
def get_discount_limits_Super_DC_card(total_usage):
    if total_usage >= 700000:
        return {'KTí†µì‹ ': 15000, 'ì»¤í”¼í• ì¸ìœ¨': 0.1}
    elif total_usage >= 300000:
        return {'KTí†µì‹ ': 7000, 'ì»¤í”¼í• ì¸ìœ¨': 0.1}
    else:
        return {'KTí†µì‹ ': 0, 'ì»¤í”¼í• ì¸ìœ¨': 0.0}

# 2. í˜œíƒ ê³„ì‚° í•¨ìˆ˜
def calculate_benefits_Super_DC_card(df, months, card_type):
    df['ì¹´í…Œê³ ë¦¬'] = df['ê°€ë§¹ì ì—…ì¢…ëª…'].apply(classify_category)
    df['ìŠ¹ì¸ê¸ˆì•¡'] = df['ìŠ¹ì¸ê¸ˆì•¡'].fillna(0)

    target_month = 202312
    benefits_list = []

    for customer_id in df['ê³ ê°ID'].unique():
        customer_data = df[df['ê³ ê°ID'] == customer_id]
        total_benefit = 0

        for i in range(months):
            year = target_month // 100
            month = target_month % 100 - i
            if month <= 0:
                month += 12
                year -= 1
            current_month = year * 100 + month

            month_data = customer_data[customer_data['ê±°ë˜ë…„ì›”'] == current_month]
            total_usage = month_data['ìŠ¹ì¸ê¸ˆì•¡'].sum()
            limits = get_discount_limits_Super_DC_card(total_usage)

            month_discount = 0
            telecom_usage = month_data[month_data['ì¹´í…Œê³ ë¦¬'] == 'ì´ë™í†µì‹ ']['ìŠ¹ì¸ê¸ˆì•¡'].sum()
            kt_discount = min(limits['KTí†µì‹ '], telecom_usage)
            month_discount += kt_discount

            coffee_usage = month_data[month_data['ì¹´í…Œê³ ë¦¬'] == 'ì»¤í”¼']['ìŠ¹ì¸ê¸ˆì•¡'].sum()
            coffee_discount = coffee_usage * limits['ì»¤í”¼í• ì¸ìœ¨']
            month_discount += coffee_discount

            total_benefit += month_discount

        annual_fee = 8000 * (months / 12) if card_type == 'CUP' else 13000 * (months / 12)
        net_benefit = total_benefit - annual_fee

        benefits_list.append({'ê³ ê°ID': customer_id, f'{months}ê°œì›”_í˜œíƒ': net_benefit})

    return pd.DataFrame(benefits_list)



# iM_ì‡¼í•‘ì¹´ë“œ
# 1. í• ì¸ í•œë„ ê³„ì‚° í•¨ìˆ˜
def get_discount_limits_Shopping_card(total_usage):
    if total_usage >= 1200000:
        return {'í• ì¸ìœ¨': 0.1, 'ì›”ê°„ìµœëŒ€í• ì¸': 20000}
    elif total_usage >= 600000:
        return {'í• ì¸ìœ¨': 0.07, 'ì›”ê°„ìµœëŒ€í• ì¸': 10000}
    elif total_usage >= 300000:
        return {'í• ì¸ìœ¨': 0.05, 'ì›”ê°„ìµœëŒ€í• ì¸': 5000}
    else:
        return {'í• ì¸ìœ¨': 0.0, 'ì›”ê°„ìµœëŒ€í• ì¸': 0}

# 2. í˜œíƒ ê³„ì‚° í•¨ìˆ˜
def calculate_benefits_Shopping_card(df, months, card_type):
    df['ì¹´í…Œê³ ë¦¬'] = df['ê°€ë§¹ì ì—…ì¢…ëª…'].apply(classify_category)
    df['ìŠ¹ì¸ê¸ˆì•¡'] = df['ìŠ¹ì¸ê¸ˆì•¡'].fillna(0)

    target_month = 202312
    benefits_list = []

    for customer_id in df['ê³ ê°ID'].unique():
        customer_data = df[df['ê³ ê°ID'] == customer_id]
        total_benefit = 0

        for i in range(months):
            year = target_month // 100
            month = target_month % 100 - i
            if month <= 0:
                month += 12
                year -= 1
            current_month = year * 100 + month

            month_data = customer_data[customer_data['ê±°ë˜ë…„ì›”'] == current_month]
            total_usage = month_data['ìŠ¹ì¸ê¸ˆì•¡'].sum()
            limits = get_discount_limits_Shopping_card(total_usage)

            month_discount = 0
            category_discount = 0

            eligible_categories_1 = ["ì˜¨ë¼ì¸ì‡¼í•‘", "í¸ì˜ì ", "ì»¤í”¼", "ë©´ì„¸ì "]
            eligible_categories_5 = ["ë°±í™”ì ", "ëŒ€í˜•ë§ˆíŠ¸", "ìŠˆí¼ë§ˆì¼“", "í™”ì¥í’ˆ", "ì˜ë¥˜", "ì•ˆê²½"]

            for category in eligible_categories_1:
                category_data = month_data[(month_data['ì¹´í…Œê³ ë¦¬'] == category) & (month_data['ìŠ¹ì¸ê¸ˆì•¡'] >= 10000)]
                category_discount += category_data['ìŠ¹ì¸ê¸ˆì•¡'].sum() * limits['í• ì¸ìœ¨']

            for category in eligible_categories_5:
                category_data = month_data[(month_data['ì¹´í…Œê³ ë¦¬'] == category) & (month_data['ìŠ¹ì¸ê¸ˆì•¡'] >= 50000)]
                category_discount += category_data['ìŠ¹ì¸ê¸ˆì•¡'].sum() * limits['í• ì¸ìœ¨']

            month_discount = min(category_discount, limits['ì›”ê°„ìµœëŒ€í• ì¸'])

            movie_data = month_data[(month_data['ì¹´í…Œê³ ë¦¬'] == 'ì˜í™”') & (month_data['ìŠ¹ì¸ê¸ˆì•¡'] >= 16000)]
            if not movie_data.empty:
                month_discount += 4000

            total_benefit += month_discount

        annual_fee = 5000 if card_type == 'BC Global' else 8000
        if customer_data['ìŠ¹ì¸ê¸ˆì•¡'].sum() >= 5000000:
            annual_fee = 0

        benefits_list.append({'ê³ ê°ID': customer_id, f'{months}ê°œì›”_í˜œíƒ': total_benefit - annual_fee})

    return pd.DataFrame(benefits_list)

# ë‹¨ë””(DANDI)ì¹´ë“œ
# 1. í• ì¸ í•œë„ ê³„ì‚° í•¨ìˆ˜
def get_discount_limits_Dandi_card(total_usage):
    if total_usage >= 1000000:
        return {"ìƒí™œí¸ì˜í•œë„": 10000, "í¬ì¸íŠ¸ì ë¦½ë¥ ": 0.006}
    elif total_usage >= 500000:
        return {"ìƒí™œí¸ì˜í•œë„": 5000, "í¬ì¸íŠ¸ì ë¦½ë¥ ": 0.005}
    elif total_usage >= 300000:
        return {"ìƒí™œí¸ì˜í•œë„": 5000, "í¬ì¸íŠ¸ì ë¦½ë¥ ": 0.004}
    else:
        return {"ìƒí™œí¸ì˜í•œë„": 0, "í¬ì¸íŠ¸ì ë¦½ë¥ ": 0.003}


# 2. í˜œíƒ ê³„ì‚° í•¨ìˆ˜
def calculate_benefits_Dandi_card(df, months):
    df["ì¹´í…Œê³ ë¦¬"] = df["ê°€ë§¹ì ì—…ì¢…ëª…"].apply(classify_category)
    df["ìŠ¹ì¸ê¸ˆì•¡"] = df["ìŠ¹ì¸ê¸ˆì•¡"].fillna(0)

    target_month = 202312
    benefits_list = []

    for customer_id in df["ê³ ê°ID"].unique():
        customer_data = df[df["ê³ ê°ID"] == customer_id]
        total_benefit = 0

        for i in range(months):
            year = target_month // 100
            month = target_month % 100 - i
            if month <= 0:
                month += 12
                year -= 1
            current_month = year * 100 + month

            month_data = customer_data[customer_data["ê±°ë˜ë…„ì›”"] == current_month]
            total_usage = month_data["ìŠ¹ì¸ê¸ˆì•¡"].sum()
            limits = get_discount_limits_Dandi_card(total_usage)

            # ìƒí™œí¸ì˜ í• ì¸ ì ìš©
            ìƒí™œí¸ì˜_data = month_data[month_data["ì¹´í…Œê³ ë¦¬"] == "ìƒí™œí¸ì˜"]
            ìƒí™œí¸ì˜_discount = min(ìƒí™œí¸ì˜_data["ìŠ¹ì¸ê¸ˆì•¡"].sum() * 0.05, limits["ìƒí™œí¸ì˜í•œë„"])

            # ì£¼ìœ  í• ì¸ ì ìš©
            ì£¼ìœ _data = month_data[month_data["ì¹´í…Œê³ ë¦¬"] == "ì£¼ìœ "]
            ì£¼ìœ _discount = min(ì£¼ìœ _data["ìŠ¹ì¸ê¸ˆì•¡"].sum() * (60 / 2000), 40000)

            # í†µì‹ ìš”ê¸ˆ í• ì¸ ì ìš©
            í†µì‹ _data = month_data[month_data["ì¹´í…Œê³ ë¦¬"] == "ì´ë™í†µì‹ "]
            í†µì‹ _discount = min(í†µì‹ _data["ìŠ¹ì¸ê¸ˆì•¡"].sum() * 0.03, 1500)

            # ë†€ì´ê³µì› í• ì¸ ì ìš©
            ë†€ì´ê³µì›_data = month_data[month_data["ì¹´í…Œê³ ë¦¬"] == "ë†€ì´ê³µì›"]
            ë†€ì´ê³µì›_discount = min(ë†€ì´ê³µì›_data["ìŠ¹ì¸ê¸ˆì•¡"].sum() * 0.5, 30000)

            # í¬ì¸íŠ¸ ì ë¦½
            ì ë¦½_points = total_usage * limits["í¬ì¸íŠ¸ì ë¦½ë¥ "]
            ì ë¦½_ê¸ˆì•¡ = min(ì ë¦½_points, 100000)  # ì›” ìµœëŒ€ 10ë§Œ í¬ì¸íŠ¸ ì ë¦½

            # í˜œíƒê³¼ í¬ì¸íŠ¸ë¥¼ í•©ì‚°
            month_discount = sum([ìƒí™œí¸ì˜_discount, ì£¼ìœ _discount, í†µì‹ _discount, ë†€ì´ê³µì›_discount, ì ë¦½_ê¸ˆì•¡])
            total_benefit += month_discount

        # ì—°íšŒë¹„ ë©´ì œ ì ìš©
        annual_fee = 7000
        if customer_data["ìŠ¹ì¸ê¸ˆì•¡"].sum() >= 3000000:
            annual_fee = 0

        benefits_list.append({"ê³ ê°ID": customer_id, f"{months}ê°œì›”_í˜œíƒ": total_benefit - annual_fee})

    return pd.DataFrame(benefits_list)


# êµ­ë¯¼í–‰ë³µì¹´ë“œ ì‹ ìš©ì¹´ë“œ
# 1. í• ì¸ í•œë„ ê³„ì‚° í•¨ìˆ˜
def get_discount_limits_Happy_card(total_usage):
    if total_usage >= 2000000:
        return {"ì›”ê°„í†µí•©í•œë„": 30000, "ê¸°ë³¸ì ë¦½ë¥ ": 0.008}
    elif total_usage >= 1000000:
        return {"ì›”ê°„í†µí•©í•œë„": 20000, "ê¸°ë³¸ì ë¦½ë¥ ": 0.006}
    elif total_usage >= 500000:
        return {"ì›”ê°„í†µí•©í•œë„": 10000, "ê¸°ë³¸ì ë¦½ë¥ ": 0.004}
    elif total_usage >= 300000:
        return {"ì›”ê°„í†µí•©í•œë„": 5000, "ê¸°ë³¸ì ë¦½ë¥ ": 0.002}
    else:
        return {"ì›”ê°„í†µí•©í•œë„": 0, "ê¸°ë³¸ì ë¦½ë¥ ": 0.002}


# 2. í˜œíƒ ê³„ì‚° í•¨ìˆ˜
def calculate_benefits_Happy_card(df, months):
    df["ì¹´í…Œê³ ë¦¬"] = df["ê°€ë§¹ì ì—…ì¢…ëª…"].apply(classify_category)
    df["ìŠ¹ì¸ê¸ˆì•¡"] = df["ìŠ¹ì¸ê¸ˆì•¡"].fillna(0)

    target_month = 202312
    benefits_list = []

    for customer_id in df["ê³ ê°ID"].unique():
        customer_data = df[df["ê³ ê°ID"] == customer_id]
        total_benefit = 0

        for i in range(months):
            year = target_month // 100
            month = target_month % 100 - i
            if month <= 0:
                month += 12
                year -= 1
            current_month = year * 100 + month

            month_data = customer_data[customer_data["ê±°ë˜ë…„ì›”"] == current_month]
            total_usage = month_data["ìŠ¹ì¸ê¸ˆì•¡"].sum()
            limits = get_discount_limits_Happy_card(total_usage)

            # ê° í˜œíƒ ê³„ì‚°
            discount_amount = 0
            for category in ["ì˜ë£Œ", "ì˜¨ë¼ì¸ì‡¼í•‘", "ì»¤í”¼", "íŒ¨ë°€ë¦¬ë ˆìŠ¤í† ë‘"]:
                category_data = month_data[month_data["ì¹´í…Œê³ ë¦¬"] == category]
                discount_amount += category_data["ìŠ¹ì¸ê¸ˆì•¡"].sum() * 0.05

            ëŒ€ì¤‘êµí†µ_data = month_data[month_data["ì¹´í…Œê³ ë¦¬"] == "ëŒ€ì¤‘êµí†µ"]
            ëŒ€ì¤‘êµí†µ_discount = min(ëŒ€ì¤‘êµí†µ_data["ìŠ¹ì¸ê¸ˆì•¡"].sum() * 0.05, 1500)

            ì´ë™í†µì‹ _data = month_data[month_data["ì¹´í…Œê³ ë¦¬"] == "ì´ë™í†µì‹ "]
            ì´ë™í†µì‹ _discount = min(1000, ì´ë™í†µì‹ _data["ìŠ¹ì¸ê¸ˆì•¡"].sum())

            ë†€ì´ê³µì›_data = month_data[month_data["ì¹´í…Œê³ ë¦¬"] == "ë†€ì´ê³µì›"]
            ë†€ì´ê³µì›_discount = min(ë†€ì´ê³µì›_data["ìŠ¹ì¸ê¸ˆì•¡"].sum() * 0.5, 30000)

            # ì´ í• ì¸ ì ìš©
            month_discount = sum([discount_amount, ëŒ€ì¤‘êµí†µ_discount, ì´ë™í†µì‹ _discount, ë†€ì´ê³µì›_discount])
            total_benefit += min(month_discount, limits["ì›”ê°„í†µí•©í•œë„"])

        benefits_list.append({"ê³ ê°ID": customer_id, f"{months}ê°œì›”_í˜œíƒ": total_benefit})

    return pd.DataFrame(benefits_list)


# ë¶€ìë˜ì„¸ìš” ì•„íŒŒíŠ¸ì¹´ë“œ
# 1. í• ì¸ í•œë„ ê³„ì‚° í•¨ìˆ˜
def get_discount_limits_Apartment_card(total_usage):
    if total_usage >= 600000:
        return {"í†µí•©í• ì¸í•œë„": 20000}
    elif total_usage >= 300000:
        return {"í†µí•©í• ì¸í•œë„": 10000}
    else:
        return {"í†µí•©í• ì¸í•œë„": 0}

# 2. í˜œíƒ ê³„ì‚° í•¨ìˆ˜
def calculate_benefits_Apartment_card(df, months):
    df["ì¹´í…Œê³ ë¦¬"] = df["ê°€ë§¹ì ì—…ì¢…ëª…"].apply(classify_category)
    df["ìŠ¹ì¸ê¸ˆì•¡"] = df["ìŠ¹ì¸ê¸ˆì•¡"].fillna(0)

    target_month = 202312
    benefits_list = []

    for customer_id in df["ê³ ê°ID"].unique():
        customer_data = df[df["ê³ ê°ID"] == customer_id]
        total_benefit = 0

        for i in range(months):
            year = target_month // 100
            month = target_month % 100 - i
            if month <= 0:
                month += 12
                year -= 1
            current_month = year * 100 + month

            month_data = customer_data[customer_data["ê±°ë˜ë…„ì›”"] == current_month]
            total_usage = month_data["ìŠ¹ì¸ê¸ˆì•¡"].sum()
            limits = get_discount_limits_Apartment_card(total_usage)

            apt_data = month_data[month_data["ì¹´í…Œê³ ë¦¬"] == "ì•„íŒŒíŠ¸ê´€ë¦¬ë¹„"]
            apt_discount = 5000 if total_usage >= 300000 else 0
            apt_discount = 10000 if total_usage >= 600000 else apt_discount

            ìƒí™œí• ì¸ = sum([
                min(category_data["ìŠ¹ì¸ê¸ˆì•¡"].sum() * 0.05, 5000)
                for category in ["ë³‘ì›", "ì•½êµ­", "ë² ì´ì»¤ë¦¬/ë””ì €íŠ¸"]
                for category_data in [month_data[(month_data["ì¹´í…Œê³ ë¦¬"] == category) & (month_data["ìŠ¹ì¸ê¸ˆì•¡"] >= 10000)]]
            ])

            total_benefit += min(apt_discount + ìƒí™œí• ì¸, limits["í†µí•©í• ì¸í•œë„"])

        benefits_list.append({"ê³ ê°ID": customer_id, f"{months}ê°œì›”_í˜œíƒ": total_benefit})

    return pd.DataFrame(benefits_list)
```

```python
import pandas as pd
import gc  # ê°€ë¹„ì§€ ì»¬ë ‰ì…˜ ëª¨ë“ˆ
from IPython.display import display
from tqdm import tqdm  # ì§„í–‰ìƒí™© ì‹œê°í™” ëª¨ë“ˆ

# ì‚¬ìš©í•  ì¹´ë“œ íƒ€ì…ì„ ëª…í™•í•˜ê²Œ ì •ì˜
card_type = "VISA"  # ë˜ëŠ” "CUP", í•„ìš”ì— ë”°ë¼ ë³€ê²½ ê°€ëŠ¥

# ì¹´ë“œ ë¦¬ìŠ¤íŠ¸ ì •ì˜ (card_typeì„ í¬í•¨í•˜ë„ë¡ ìˆ˜ì •)
cards = [
    ("iM_ì˜¬ë ˆ_Super_DC_ì¹´ë“œ", lambda df, months: calculate_benefits_Super_DC_card(df, months, card_type)),
    ("iM_ì‡¼í•‘ì¹´ë“œ", lambda df, months: calculate_benefits_Shopping_card(df, months, card_type)),
    ("ë‹¨ë””ì¹´ë“œ", calculate_benefits_Dandi_card),  # card_type í•„ìš” ì—†ìŒ
    ("êµ­ë¯¼í–‰ë³µì¹´ë“œ", calculate_benefits_Happy_card),  # card_type í•„ìš” ì—†ìŒ
    ("ë¶€ìë˜ì„¸ìš”_ì•„íŒŒíŠ¸ì¹´ë“œ", calculate_benefits_Apartment_card),  # card_type í•„ìš” ì—†ìŒ
]

# sample1ë¶€í„° sample8ê¹Œì§€ tqdmì„ ì‚¬ìš©í•˜ì—¬ ì§„í–‰ë¥  í‘œì‹œ
for i in tqdm(range(1, 9), desc="ğŸ“‚ Processing CSV Files"):
    file_name = f'card_data_2023_sample{i}_utf8.csv'

    try:
        df = pd.read_csv(file_name, encoding='utf-8')
        print(f"\n=== Processing {file_name} ===")

        final_result = None  # ì¹´ë“œë³„ ê³„ì‚° ê²°ê³¼ë¥¼ ì €ì¥í•  ë°ì´í„°í”„ë ˆì„ ì´ˆê¸°í™”

        # ì¹´ë“œë³„ í˜œíƒ ê³„ì‚°
        for card_name, calc_function in tqdm(cards, desc=f"ğŸƒ Processing Cards for {file_name}", leave=False):
            result_1 = calc_function(df, 1).rename(columns={'1ê°œì›”_í˜œíƒ': f'{card_name}_1ê°œì›”í˜œíƒ'})
            result_3 = calc_function(df, 3).rename(columns={'3ê°œì›”_í˜œíƒ': f'{card_name}_3ê°œì›”í˜œíƒ'})
            result_6 = calc_function(df, 6).rename(columns={'6ê°œì›”_í˜œíƒ': f'{card_name}_6ê°œì›”í˜œíƒ'})
            result_12 = calc_function(df, 12).rename(columns={'12ê°œì›”_í˜œíƒ': f'{card_name}_12ê°œì›”í˜œíƒ'})

            # ê²°ê³¼ ë³‘í•©
            if final_result is None:
                final_result = result_1.merge(result_3, on="ê³ ê°ID")\
                                       .merge(result_6, on="ê³ ê°ID")\
                                       .merge(result_12, on="ê³ ê°ID")
            else:
                final_result = final_result.merge(result_1, on="ê³ ê°ID")\
                                           .merge(result_3, on="ê³ ê°ID")\
                                           .merge(result_6, on="ê³ ê°ID")\
                                           .merge(result_12, on="ê³ ê°ID")

        # ê²°ê³¼ ì €ì¥
        output_file = f'im_i_card_jhoon_sample{i}.csv'
        final_result.to_csv(output_file, index=False, encoding="utf-8")
        print(f"ğŸ“ ì €ì¥ ì™„ë£Œ: {output_file}")

        # âœ… **ë©”ëª¨ë¦¬ ì ˆì•½ì„ ìœ„í•œ ë°ì´í„°í”„ë ˆì„ ì‚­ì œ**
        del final_result
        del df
        gc.collect()  # ê°€ë¹„ì§€ ì»¬ë ‰ì…˜ ì‹¤í–‰

    except Exception as e:
        print(f"âš ï¸ íŒŒì¼ {file_name} ì²˜ë¦¬ ì¤‘ ì˜¤ë¥˜ ë°œìƒ: {e}")
```
