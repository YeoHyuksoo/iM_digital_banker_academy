```python
import pandas as pd
from IPython.display import display

# 업종 분류 함수
def classify_category(business_type):
    categories = {
        "대중교통": {
            "철도": ["철도", "KTX"],
            "고속버스": ["고속버스"],
            "택시": ["택시", "택시회사"],
            "대중교통": ["버스", "지하철"]
        },
        "배달앱": ["배달의민족", "요기요", "쿠팡이츠"],
        "OTT (스트리밍)": ["넷플릭스", "디즈니+", "웨이브", "티빙", "왓챠"],
        "이동통신, 통신요금": {
            "이동통신요금": ["이동통신요금"],
            "kt통신": ["이동통신요금", "통신기기", "통신서비스/소득공제비대상", "통신판매업1", "일반(통신판매)", "위성방송"]
        },
        "영화": ["영화관", "롯데시네마", "메가박스"],
        "커피": {
            "기타음료식품": ["기타음료식품"],
            "커피브랜드": ["스타벅스", "커피빈", "엔젤리너스", "시애틀", "Sleepless In Seattle", "카페베네"]
        },
        "편의점": {
            "일반편의점": ["GS25", "CU", "세븐일레븐", "바이더웨이", "미니스톱", "다이소", "미니샵"]
        },
        "온라인쇼핑": {
            "종합쇼핑몰": ["G마켓", "옥션", "11번가", "쿠팡", "위메프", "티몬", "페이북쇼핑", "인터넷Mall", "인터넷종합Mall", "PG상품권"],
            "백화점몰": ["신세계몰", "롯데닷컴", "갤러리아몰", "AK몰"],
            "홈쇼핑": ["GS샵", "CJ오쇼핑", "롯데홈쇼핑", "현대H몰", "NS홈쇼핑", "홈앤쇼핑", "롯데i몰"],
            "대형마트몰": ["이마트몰", "홈플러스몰", "롯데마트몰", "롯데슈퍼몰"],
            "면세점몰": ["신라면세점", "롯데면세점", "동화면세점", "워커힐면세점"]
        },
        "백화점": ["자사카드발행백화점", "백화점"],
        "대형마트": ["대형할인점", "농협하나로클럽", "농,축협직영매장", "연쇄점", "대형마트"],
        "슈퍼마켓": ["슈퍼마켓"],
        "면세점": ["신라", "롯데", "동화", "워커힐", "면세점"],
        "화장품": ["화장품"],
        "의류": ["일반의류", "스포츠", "골프", "아웃도어"],
        "안경": ["안경"],
        "해외결제": ["해외 직구", "해외 가맹점"],
        "외식, 음식점": ["한식", "양식", "중식", "일식", "패밀리 레스토랑", "아웃백", "TGIF", "VIPS"],
        "주유": ["주유소", "GS칼텍스", "GS주유소", "SK주유소", "현대오일뱅크", "S-OIL", "LPG충전소"],
        "자동차 관련": ["자동차정비", "세차장", "렌터카"],
        "레저 및 취미": ["골프장", "골프연습장", "스크린골프", "당구장", "볼링장", "수영장", "노래방", "헬스클럽", "스포츠센터"],
        "숙박": ["호텔", "모텔", "리조트", "콘도", "카라반", "글램핑", "팬션"],
        "보험": ["생명보험", "손해보험"],
        "건축 및 인테리어": ["가구", "인테리어", "조명", "목재", "철물", "건축자재"],
        "육아/교육": ["어린이집", "유치원", "학원"],
        "서적": ["교보문고", "영풍문고", "YES24", "영광서적"],
        "놀이공원": ["에버랜드", "롯데월드", "서울랜드", "경주월드", "통도환타지아"],
        "생활편의": ["서적", "학원", "외식"],
        "의료": ["병원", "의원", "조산원", "산후조리원"],
        "아파트관리비": ["아파트관리비"],
        "약국": ["약국"],
        "베이커리/디저트": ["던킨도너츠", "베스킨라빈스"],
        "패션": ["여성크로커다일", "샤트렌", "올리비아하슬러", "와일드로즈"],
        "여행": ["국내여행", "해외여행"]
    }

    for category, subcategories in categories.items():
        if isinstance(subcategories, dict):  # 서브 카테고리 존재
            for subcategory, keywords in subcategories.items():
                if business_type in keywords:
                    return category
        elif isinstance(subcategories, list):  # 리스트 형식의 단일 카테고리
            if business_type in subcategories:
                return category

    return "기타 (분류 불가)"

# iM_올레_Super_DC_카드
# 1. 할인 한도 계산 함수
def get_discount_limits_Super_DC_card(total_usage):
    if total_usage >= 700000:
        return {'KT통신': 15000, '커피할인율': 0.1}
    elif total_usage >= 300000:
        return {'KT통신': 7000, '커피할인율': 0.1}
    else:
        return {'KT통신': 0, '커피할인율': 0.0}

# 2. 혜택 계산 함수
def calculate_benefits_Super_DC_card(df, months, card_type):
    df['카테고리'] = df['가맹점업종명'].apply(classify_category)
    df['승인금액'] = df['승인금액'].fillna(0)

    target_month = 202312
    benefits_list = []

    for customer_id in df['고객ID'].unique():
        customer_data = df[df['고객ID'] == customer_id]
        total_benefit = 0

        for i in range(months):
            year = target_month // 100
            month = target_month % 100 - i
            if month <= 0:
                month += 12
                year -= 1
            current_month = year * 100 + month

            month_data = customer_data[customer_data['거래년월'] == current_month]
            total_usage = month_data['승인금액'].sum()
            limits = get_discount_limits_Super_DC_card(total_usage)

            month_discount = 0
            telecom_usage = month_data[month_data['카테고리'] == '이동통신']['승인금액'].sum()
            kt_discount = min(limits['KT통신'], telecom_usage)
            month_discount += kt_discount

            coffee_usage = month_data[month_data['카테고리'] == '커피']['승인금액'].sum()
            coffee_discount = coffee_usage * limits['커피할인율']
            month_discount += coffee_discount

            total_benefit += month_discount

        annual_fee = 8000 * (months / 12) if card_type == 'CUP' else 13000 * (months / 12)
        net_benefit = total_benefit - annual_fee

        benefits_list.append({'고객ID': customer_id, f'{months}개월_혜택': net_benefit})

    return pd.DataFrame(benefits_list)



# iM_쇼핑카드
# 1. 할인 한도 계산 함수
def get_discount_limits_Shopping_card(total_usage):
    if total_usage >= 1200000:
        return {'할인율': 0.1, '월간최대할인': 20000}
    elif total_usage >= 600000:
        return {'할인율': 0.07, '월간최대할인': 10000}
    elif total_usage >= 300000:
        return {'할인율': 0.05, '월간최대할인': 5000}
    else:
        return {'할인율': 0.0, '월간최대할인': 0}

# 2. 혜택 계산 함수
def calculate_benefits_Shopping_card(df, months, card_type):
    df['카테고리'] = df['가맹점업종명'].apply(classify_category)
    df['승인금액'] = df['승인금액'].fillna(0)

    target_month = 202312
    benefits_list = []

    for customer_id in df['고객ID'].unique():
        customer_data = df[df['고객ID'] == customer_id]
        total_benefit = 0

        for i in range(months):
            year = target_month // 100
            month = target_month % 100 - i
            if month <= 0:
                month += 12
                year -= 1
            current_month = year * 100 + month

            month_data = customer_data[customer_data['거래년월'] == current_month]
            total_usage = month_data['승인금액'].sum()
            limits = get_discount_limits_Shopping_card(total_usage)

            month_discount = 0
            category_discount = 0

            eligible_categories_1 = ["온라인쇼핑", "편의점", "커피", "면세점"]
            eligible_categories_5 = ["백화점", "대형마트", "슈퍼마켓", "화장품", "의류", "안경"]

            for category in eligible_categories_1:
                category_data = month_data[(month_data['카테고리'] == category) & (month_data['승인금액'] >= 10000)]
                category_discount += category_data['승인금액'].sum() * limits['할인율']

            for category in eligible_categories_5:
                category_data = month_data[(month_data['카테고리'] == category) & (month_data['승인금액'] >= 50000)]
                category_discount += category_data['승인금액'].sum() * limits['할인율']

            month_discount = min(category_discount, limits['월간최대할인'])

            movie_data = month_data[(month_data['카테고리'] == '영화') & (month_data['승인금액'] >= 16000)]
            if not movie_data.empty:
                month_discount += 4000

            total_benefit += month_discount

        annual_fee = 5000 if card_type == 'BC Global' else 8000
        if customer_data['승인금액'].sum() >= 5000000:
            annual_fee = 0

        benefits_list.append({'고객ID': customer_id, f'{months}개월_혜택': total_benefit - annual_fee})

    return pd.DataFrame(benefits_list)

# 단디(DANDI)카드
# 1. 할인 한도 계산 함수
def get_discount_limits_Dandi_card(total_usage):
    if total_usage >= 1000000:
        return {"생활편의한도": 10000, "포인트적립률": 0.006}
    elif total_usage >= 500000:
        return {"생활편의한도": 5000, "포인트적립률": 0.005}
    elif total_usage >= 300000:
        return {"생활편의한도": 5000, "포인트적립률": 0.004}
    else:
        return {"생활편의한도": 0, "포인트적립률": 0.003}


# 2. 혜택 계산 함수
def calculate_benefits_Dandi_card(df, months):
    df["카테고리"] = df["가맹점업종명"].apply(classify_category)
    df["승인금액"] = df["승인금액"].fillna(0)

    target_month = 202312
    benefits_list = []

    for customer_id in df["고객ID"].unique():
        customer_data = df[df["고객ID"] == customer_id]
        total_benefit = 0

        for i in range(months):
            year = target_month // 100
            month = target_month % 100 - i
            if month <= 0:
                month += 12
                year -= 1
            current_month = year * 100 + month

            month_data = customer_data[customer_data["거래년월"] == current_month]
            total_usage = month_data["승인금액"].sum()
            limits = get_discount_limits_Dandi_card(total_usage)

            # 생활편의 할인 적용
            생활편의_data = month_data[month_data["카테고리"] == "생활편의"]
            생활편의_discount = min(생활편의_data["승인금액"].sum() * 0.05, limits["생활편의한도"])

            # 주유 할인 적용
            주유_data = month_data[month_data["카테고리"] == "주유"]
            주유_discount = min(주유_data["승인금액"].sum() * (60 / 2000), 40000)

            # 통신요금 할인 적용
            통신_data = month_data[month_data["카테고리"] == "이동통신"]
            통신_discount = min(통신_data["승인금액"].sum() * 0.03, 1500)

            # 놀이공원 할인 적용
            놀이공원_data = month_data[month_data["카테고리"] == "놀이공원"]
            놀이공원_discount = min(놀이공원_data["승인금액"].sum() * 0.5, 30000)

            # 포인트 적립
            적립_points = total_usage * limits["포인트적립률"]
            적립_금액 = min(적립_points, 100000)  # 월 최대 10만 포인트 적립

            # 혜택과 포인트를 합산
            month_discount = sum([생활편의_discount, 주유_discount, 통신_discount, 놀이공원_discount, 적립_금액])
            total_benefit += month_discount

        # 연회비 면제 적용
        annual_fee = 7000
        if customer_data["승인금액"].sum() >= 3000000:
            annual_fee = 0

        benefits_list.append({"고객ID": customer_id, f"{months}개월_혜택": total_benefit - annual_fee})

    return pd.DataFrame(benefits_list)


# 국민행복카드 신용카드
# 1. 할인 한도 계산 함수
def get_discount_limits_Happy_card(total_usage):
    if total_usage >= 2000000:
        return {"월간통합한도": 30000, "기본적립률": 0.008}
    elif total_usage >= 1000000:
        return {"월간통합한도": 20000, "기본적립률": 0.006}
    elif total_usage >= 500000:
        return {"월간통합한도": 10000, "기본적립률": 0.004}
    elif total_usage >= 300000:
        return {"월간통합한도": 5000, "기본적립률": 0.002}
    else:
        return {"월간통합한도": 0, "기본적립률": 0.002}


# 2. 혜택 계산 함수
def calculate_benefits_Happy_card(df, months):
    df["카테고리"] = df["가맹점업종명"].apply(classify_category)
    df["승인금액"] = df["승인금액"].fillna(0)

    target_month = 202312
    benefits_list = []

    for customer_id in df["고객ID"].unique():
        customer_data = df[df["고객ID"] == customer_id]
        total_benefit = 0

        for i in range(months):
            year = target_month // 100
            month = target_month % 100 - i
            if month <= 0:
                month += 12
                year -= 1
            current_month = year * 100 + month

            month_data = customer_data[customer_data["거래년월"] == current_month]
            total_usage = month_data["승인금액"].sum()
            limits = get_discount_limits_Happy_card(total_usage)

            # 각 혜택 계산
            discount_amount = 0
            for category in ["의료", "온라인쇼핑", "커피", "패밀리레스토랑"]:
                category_data = month_data[month_data["카테고리"] == category]
                discount_amount += category_data["승인금액"].sum() * 0.05

            대중교통_data = month_data[month_data["카테고리"] == "대중교통"]
            대중교통_discount = min(대중교통_data["승인금액"].sum() * 0.05, 1500)

            이동통신_data = month_data[month_data["카테고리"] == "이동통신"]
            이동통신_discount = min(1000, 이동통신_data["승인금액"].sum())

            놀이공원_data = month_data[month_data["카테고리"] == "놀이공원"]
            놀이공원_discount = min(놀이공원_data["승인금액"].sum() * 0.5, 30000)

            # 총 할인 적용
            month_discount = sum([discount_amount, 대중교통_discount, 이동통신_discount, 놀이공원_discount])
            total_benefit += min(month_discount, limits["월간통합한도"])

        benefits_list.append({"고객ID": customer_id, f"{months}개월_혜택": total_benefit})

    return pd.DataFrame(benefits_list)


# 부자되세요 아파트카드
# 1. 할인 한도 계산 함수
def get_discount_limits_Apartment_card(total_usage):
    if total_usage >= 600000:
        return {"통합할인한도": 20000}
    elif total_usage >= 300000:
        return {"통합할인한도": 10000}
    else:
        return {"통합할인한도": 0}

# 2. 혜택 계산 함수
def calculate_benefits_Apartment_card(df, months):
    df["카테고리"] = df["가맹점업종명"].apply(classify_category)
    df["승인금액"] = df["승인금액"].fillna(0)

    target_month = 202312
    benefits_list = []

    for customer_id in df["고객ID"].unique():
        customer_data = df[df["고객ID"] == customer_id]
        total_benefit = 0

        for i in range(months):
            year = target_month // 100
            month = target_month % 100 - i
            if month <= 0:
                month += 12
                year -= 1
            current_month = year * 100 + month

            month_data = customer_data[customer_data["거래년월"] == current_month]
            total_usage = month_data["승인금액"].sum()
            limits = get_discount_limits_Apartment_card(total_usage)

            apt_data = month_data[month_data["카테고리"] == "아파트관리비"]
            apt_discount = 5000 if total_usage >= 300000 else 0
            apt_discount = 10000 if total_usage >= 600000 else apt_discount

            생활할인 = sum([
                min(category_data["승인금액"].sum() * 0.05, 5000)
                for category in ["병원", "약국", "베이커리/디저트"]
                for category_data in [month_data[(month_data["카테고리"] == category) & (month_data["승인금액"] >= 10000)]]
            ])

            total_benefit += min(apt_discount + 생활할인, limits["통합할인한도"])

        benefits_list.append({"고객ID": customer_id, f"{months}개월_혜택": total_benefit})

    return pd.DataFrame(benefits_list)
```

```python
import pandas as pd
import gc  # 가비지 컬렉션 모듈
from IPython.display import display
from tqdm import tqdm  # 진행상황 시각화 모듈

# 사용할 카드 타입을 명확하게 정의
card_type = "VISA"  # 또는 "CUP", 필요에 따라 변경 가능

# 카드 리스트 정의 (card_type을 포함하도록 수정)
cards = [
    ("iM_올레_Super_DC_카드", lambda df, months: calculate_benefits_Super_DC_card(df, months, card_type)),
    ("iM_쇼핑카드", lambda df, months: calculate_benefits_Shopping_card(df, months, card_type)),
    ("단디카드", calculate_benefits_Dandi_card),  # card_type 필요 없음
    ("국민행복카드", calculate_benefits_Happy_card),  # card_type 필요 없음
    ("부자되세요_아파트카드", calculate_benefits_Apartment_card),  # card_type 필요 없음
]

# sample1부터 sample8까지 tqdm을 사용하여 진행률 표시
for i in tqdm(range(1, 9), desc="📂 Processing CSV Files"):
    file_name = f'card_data_2023_sample{i}_utf8.csv'

    try:
        df = pd.read_csv(file_name, encoding='utf-8')
        print(f"\n=== Processing {file_name} ===")

        final_result = None  # 카드별 계산 결과를 저장할 데이터프레임 초기화

        # 카드별 혜택 계산
        for card_name, calc_function in tqdm(cards, desc=f"🃏 Processing Cards for {file_name}", leave=False):
            result_1 = calc_function(df, 1).rename(columns={'1개월_혜택': f'{card_name}_1개월혜택'})
            result_3 = calc_function(df, 3).rename(columns={'3개월_혜택': f'{card_name}_3개월혜택'})
            result_6 = calc_function(df, 6).rename(columns={'6개월_혜택': f'{card_name}_6개월혜택'})
            result_12 = calc_function(df, 12).rename(columns={'12개월_혜택': f'{card_name}_12개월혜택'})

            # 결과 병합
            if final_result is None:
                final_result = result_1.merge(result_3, on="고객ID")\
                                       .merge(result_6, on="고객ID")\
                                       .merge(result_12, on="고객ID")
            else:
                final_result = final_result.merge(result_1, on="고객ID")\
                                           .merge(result_3, on="고객ID")\
                                           .merge(result_6, on="고객ID")\
                                           .merge(result_12, on="고객ID")

        # 결과 저장
        output_file = f'im_i_card_jhoon_sample{i}.csv'
        final_result.to_csv(output_file, index=False, encoding="utf-8")
        print(f"📁 저장 완료: {output_file}")

        # ✅ **메모리 절약을 위한 데이터프레임 삭제**
        del final_result
        del df
        gc.collect()  # 가비지 컬렉션 실행

    except Exception as e:
        print(f"⚠️ 파일 {file_name} 처리 중 오류 발생: {e}")
```
