```python
import pandas as pd
```

```python
df = pd.read_csv("newbie.csv")
```

##### 데이터 로드

```python
df
```

##### 인터넷 신규가입 데이터는 19583개의 행, 11개의 컬럼으로 구성되어 있음

```python
df = df.drop(['who', 'Country', 'Years on Internet'], axis = 1)
```

##### 독립변수 중 'who', 'Country', 'Years on Internet' 3개의 변수는 제거함

```python
target_df = df[['Newbie']]

target_df.info()
```

```python
data_df = df.drop(['Newbie'], axis = 1)

data_df
```

```python
data_df[data_df['Age'].isna()]
```

##### 전체 데이터에서 Age 컬럼에 결측치가 있는 행을 출력해보니 561개가 존재했음

```python
ageNaN_indices = data_df[data_df['Age'].isna()].index

data_df.iloc[ageNaN_indices]
```

##### Age 컬럼 결측치가 평균값으로 잘 채워 졌는지 확인하기 위해 
##### 결측치인 행의 인덱스만 추출하여 iloc 메서드를 통해 부분 데이터프레임 출력

```python
from sklearn.impute import SimpleImputer
import numpy as np

imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
data_df["Age"]= imp_mean.fit_transform(data_df[["Age"]])

data_df.iloc[ageNaN_indices]
```

##### 평균값으로 결측치를 채우기 위해 SimpleImputer 함수를 활용했고, 
##### 인스턴스화하여 fit을 통해 모델을 만들고 transform을 통해 데이터를 모델의 로직에 따라 변환했다.
##### 평균값인 35.2222로 채워진 것을 볼 수 있다.

```python
data_df['Age'].sum() / data_df['Age'].count()
```

```python
data_df.rename({"Household Income":"Household_Income", "Sexual Preference": "Sexual_Preference",
                "Education Attainment":"Education_Attainment", "Major Occupation":"Major_Occupation",
               "Marital Status":"Marital_Status"}, axis=1, inplace=True)

data_df
```

##### 나중에 피쳐 선택을 할 때, formula 식에 적합하게 띄어쓰기가 있는 컬럼명을 수정했다.

```python
incomeNaN_indices = data_df[data_df['Household_Income'].isna()].index

data_df.iloc[incomeNaN_indices]
```

##### 'Household Income' 컬럼이 결측치인 행은 3185개가 나왔다.

```python
imp_mf = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
```

##### Household Income, Sexual Preference, Marital Status 3개의 컬럼은
##### 최빈값으로 결측치를 채우기 위해 SimpleImputer 함수에 strategy값을 'most_frequent'로 설정하여 다시 호출하였다.

```python
data_df["Household_Income"] = imp_mf.fit_transform(data_df[["Household_Income"]])
data_df.iloc[incomeNaN_indices]
```

##### 'Household Income' 컬럼은 최빈값인 $50-74로 채워진 것을 볼 수 있다.

```python
sexualNaN_indices = data_df[data_df['Sexual_Preference'].isna()].index

data_df.iloc[sexualNaN_indices]
```

##### Sexual Preference 컬럼은 1292개의 결측치가 있었다.

```python
data_df["Sexual_Preference"]= imp_mf.fit_transform(data_df[["Sexual_Preference"]])

data_df.iloc[sexualNaN_indices]
```

##### 최빈값인 Heterosexual로 모두 채워넣었다.

```python
marryNaN_indices = data_df[data_df['Marital_Status'].isna()].index

data_df.iloc[marryNaN_indices]
```

##### Marital Status 컬럼의 경우 343개의 결측치가 있었다.

```python
data_df["Marital_Status"]= imp_mf.fit_transform(data_df[["Marital_Status"]])
data_df.iloc[marryNaN_indices]
```

##### Married가 가장 많았기에 Married로 모든 결측치가 채워졌다.

```python
# 범주형 독립변수 : Gender, Household Income, Sexual Preference, Education Attainment, Major Occupation, 
```

```python
onehot_df = pd.get_dummies(data_df, columns = ['Gender', 'Household_Income', 'Sexual_Preference',
                                         'Education_Attainment', 'Major_Occupation', 'Marital_Status'])
onehot_df.columns
```

##### get_dummies라는 pandas 함수를 사용하여 범주형 변수를 원핫인코딩 처리한다.

```python
import statsmodels.formula.api as smf

formula = "Newbie ~ " + "+".join(onehot_df.columns)
formula
```

```python
df = pd.concat([onehot_df, target_df], axis = 1)
```

##### concat함수로 독립변수와 타겟변수 데이터프레임을 하나로 합침.

##### 회귀모델 summary 결과를 보고 pvalue가 유의하지 않은 변수를 제거하려 시도함.

##### 독립변수끼리 상관관계가 0.9 이상인 것도 추출하여 제거해보려 시도함.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(onehot_df, target_df, test_size = 0.2, random_state = 42)
```

##### 테스트 데이터 20% 학습데이터 80%로 독립변수와 타겟 변수를 분할함.

```python
X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver = "liblinear")
model.fit(X_train, y_train)
```

##### LogisticRegression 함수 활용해서 모델을 만든다.

```python
from sklearn.metrics import accuracy_score

pred = model.predict(X_test)
print("정확도 : ")
accuracy_score(y_test, pred)
```

##### 테스트 데이터를 대상으로 값을 예측하고, 정답 데이터와 비교하여 정확도를 측정한다.
