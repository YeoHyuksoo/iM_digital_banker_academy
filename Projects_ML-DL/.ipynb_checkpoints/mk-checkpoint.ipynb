```python
import pandas as pd
import numpy as np

df = pd.read_csv('intensity.csv')
df
```

```python
df = df[df['Weight (kg)'] < 120]
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Gender 컬럼에 대한 레이블 인코딩
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])

# Workout_Type 컬럼에 대한 원핫 인코딩
workout_encoded = pd.get_dummies(df['Workout_Type'], prefix='Workout')

# 원핫 인코딩 결과를 기존 데이터프레임에 추가하고 원본 Workout_Type 컬럼 삭제
df = pd.concat([df.drop('Workout_Type', axis=1), workout_encoded], axis=1)

# 결과 확인
print(df.head())
```

```python
df
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer

# 표준화에서 제외할 컬럼들
exclude_cols = ['Gender', 'Workout_Cardio', 'Workout_HIIT', 'Workout_Strength', 'Workout_Yoga', 'Fat_Percentage']

# 표준화할 수치형 컬럼들
numerical_cols = [col for col in df.columns if col not in exclude_cols]

# ColumnTransformer 설정
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols)
    ],
    remainder='passthrough'  # 나머지 컬럼은 그대로 유지
)

# 전처리 적용
transformed_array = preprocessor.fit_transform(df)

# 새로운 컬럼 순서 설정
new_columns = numerical_cols + exclude_cols

# 결과를 데이터프레임으로 변환
df_processed = pd.DataFrame(transformed_array, columns=new_columns, index=df.index)

print(df_processed.head())
```

```python
# 제외할 컬럼 리스트
exclude_cols = ['64%_BPM', '76%_BPM', 'Resting_BPM', 'Fat_Percentage']

# dfx 생성: 제외할 컬럼들을 제외한 모든 컬럼
dfx = df_processed.drop(columns=exclude_cols)

# dfy 생성: Fat_Percentage 컬럼만 포함
dfy = df_processed['Fat_Percentage']

print("dfx의 컬럼:", dfx.columns.tolist())
print("\ndfx의 shape:", dfx.shape)
print("\ndfy의 shape:", dfy.shape)
```
