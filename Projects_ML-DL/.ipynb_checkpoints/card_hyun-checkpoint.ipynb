```python
categories = {
    "대중교통": {
        "철도": ["철도"],
        "고속버스": ["고속버스"]
    },
    "배달앱": [],
    "OTT (스트리밍)": [],
    "이동통신, 통신요금": {
        "이동통신요금": ["이동통신요금"],
        "kt통신": [
            "이동통신요금",
            "통신기기",
            "통신서비스/소득공제비대상",
            "통신판매업1",
            "일반(통신판매)",
            "위성방송"
        ]
    },
    "영화": {
        "영화관": ["영화관"]
    },
    "커피": {
        "기타음료식품": ["기타음료식품"]
    },
    "생활용품": [
        "화장품",
        "문구용품",
        "침구·수예점",
        "주방용구",
        "주방용식기",
        "연 쇄 점"
    ],
    "편의점": {
        "편 의 점": ["편 의 점"]
    },
    "대한항공 마일리지": {
        "항공사": ["항공사"]
    },
    "마일리지+라운지": [],
    "항공, 면세점(여행)": [
        "항공사",
        "면 세 점"
    ],
    "후불교통, 택시, 고속버스, 기차(KTX포함) (교통)": [
        "택시",
        "고속버스",
        "택시회사"
    ],
    "온라인쇼핑": [
        "인터넷Mall",
        "인터넷종합Mall",
        "PG상품권"
    ],
    "국내 및 해외 가맹점": {
        "제외할 것": ["PG상품권", "전자상거래상품권", "전자상거래상품권전문판매", "상품권전문판매"]
    },
    "홈쇼핑": {
        "CATV홈쇼핑": ["CATV홈쇼핑"]
    },
    "가전, 가구": {
        "냉열기기": ["냉열기기"]
    },
    "외식, 음식점": [
        "일반한식",
        "서양음식",
        "중국식",
        "일식·회집",
        "갈비전문점",
        "한정식",
        "유흥주점",
        "단란주점",
        "주점",
        "칵테일바"
    ],
    "패밀리 레스토랑": [],
    "간편결제": {
        "인터넷P/G": ["인터넷P/G"]
    },
    "관리비, 전기요금, 전기/가스요금": [
        "공공요금대행서비스/소득공제대상",
        "공공요금대행서비스/소득공제비대상"
    ],
    "대형마트 (트레이더스 포함)": [
        "대형할인점",
        "농협하나로클럽",
        "농,축협직영매장",
        "연쇄점"
    ],
    "슈퍼마켓": [],
    "주유": [
        "SK주유소",
        "GS주유소",
        "현대정유(오일뱅크)",
        "쌍용S-OIL",
        "SK가스충전소",
        "GS가스충전소",
        "쌍용S-OIL가스충전소",
        "현대정유가스충전소",
        "LPG취급점",
        "유류판매",
        "윤활유전문판매",
        "주유소",
        "E1가스충전소"
    ],
    "공공시설 입장료": {
        "주차장": ["주차장"]
    },
    "친환경 (에코머니 포함)": {
        "전기차충전소": ["전기차충전소"]
    },
    "백화점": {
        "자사카드발행백화점": ["자사카드발행백화점"]
    },
    "놀이공원": [],
    "도서": [
        "일반서적",
        "전문서적",
        "출판 및 인쇄물",
        "기타서적문구"
    ],
    "전기차": {
        "전기차충전소": ["전기차충전소"]
    },
    "공유모빌리티": ["가타자동차서비스"],
    "반려동물": [
        "애완동물",
        "동물병원"
    ],
    "의료": [
        "병 원(응급실운영)",
        "치과의원",
        "치과병원",
        "한 의 원",
        "한방병원",
        "의료용품",
        "건강진단",
        "의 원",
        "약 국",
        "기타의료기관 및 기타의료기기",
        "한약방"
    ],
     "병원, 한의원(건강)": [
        "병원(응급실운영)",
        "치과의원",
        "치과병원",
        "의 원",
        "한 의 원",
        "약국",
        "종합병원",
        "건강진단"
    ],
    "키즈 (어린이 관련)": {
        "유아원": ["유아원"]
    },
    "학원 (교육 포함)": [
        "보습학원",
        "예·체능계학원",
        "외국어학원",
        "학습지교육",
        "초중고교육기관",
        "독서실",
        "기타교육기관",
        "기능학원",
        "학원"
    ],
    "이/미용실": [
        "미용원",
        "피부미용실"
    ],
    "안경": {
        "안경": ["안경"]
    },
    "제과(던킨, 베라, 뚜레쥬르)": {
        "제과점": ["제과점"]
    },
    "약국": {
        "약국": ["약국"]
    },
    "기타 (분류 불가)": {
        "레저 및 취미 관련": [
            "기타레저업소",
            "골프경기장",
            "골프연습장",
            "당구장",
            "스크린골프",
            "볼 링 장",
            "수 영 장",
            "종합레져타운",
            "노 래 방",
            "스포츠·레져용품",
            "음반영상물",
            "골프용품 전문점",
            "헬스클럽",
            "문화취미기타",
            "레져업소"
        ],
        "자동차 관련": [
            "자동차정비",
            "자동차부품",
            "자동차시트·타이어",
            "가타자동차서비스",
            "카인테리어",
            "렌터카",
            "세차장"
        ],
        "숙박 관련": [
            "특급호텔",
            "1급 호텔",
            "2급 호텔",
            "콘도",
            "기타숙박업"
        ],
        "보험 관련": [
            "생명보험",
            "손해보험",
            "기타보험"
        ],
        "잡화 및 의류 관련": [
            "캐주얼의류",
            "정장",
            "단체복",
            "맞춤복점",
            "내의판매",
            "신발",
            "시 계",
            "옷감·직물",
            "액세서리",
            "기타잡화",
            "완구점",
            "스포츠의류",
            "가 방",
            "신 발",
            "아동의류",
            "안경",
            "제 화"
        ],
        "음식 및 음료 관련": [
            "스넥",
            "정육점",
            "인삼제품",
            "기타건강식품",
            "주류판매점"
        ],
        "농업 및 축산 관련": [
            "농·축·수산품",
            "농축수산가공품",
            "비료·사료·종자기타농업관련",
            "농기계"
        ],
        "건축 및 인테리어 관련": [
            "건축용 요업제품",
            "목재·석재·철물",
            "조명기구",
            "카페트",
            "커튼",
            "천막",
            "지물",
            "기타가구",
            "기계공구",
            "일반가구",
            "기타건축자재",
            "보일러·펌프·샷시",
            "인테리어전문"
        ],
        "그 외 기타 항목들": [
            "가전제품",
            "소프트웨어",
            "부동산중개·임대",
            "혼례서비스업",
            "비영리/대상",
            "비영리/비대상",
            "컴퓨터",
            "업종미등록",
            "위탁급식업",
            "수족관",
            "화물운송",
            "기타수리서비스",
            "미용재료",
            "가정용품수리",
            "민예·공예품",
            "사무서비스",
            "사우나",
            "이륜차판매",
            "관공여행",
            "기념품점",
            "미곡상",
            "귀금속",
            "화방·표구점",
            "DP&E",
            "기타 전문점",
            "기타 사무용품",
            "사무·통신기기수리",
            "종합용역",
            "기타용역서비스",
            "보관및 창고업",
            "여객선",
            "관광여행",
            "비씨카드 정산용(할인)",
            "기타4",
            "기타1",
            "안마/스포츠마사지",
            "전자상거래상품권",
            "전자상거래상품권전문판매",
            "상품권전문판매",
            "제과점",
            "화원",
            "페인트",
            "피부미용실",
            "티켓",
            "기타유통업",
            "홍삼제품",
            "기타의류",
            "기타전기제품",
            "비씨카드 정산용가맹점",
            "구내매점(국가기관등)",
            "상 품 권",
            "조세서비스",
            "사무용 OA기기",
            "가례서비스업",
            "정수기"
        ]
    }
}

def classify_category(business_type):
    """가맹점 업종명을 기반으로 카테고리를 분류하는 함수"""
    for category, details in categories.items():
        if isinstance(details, dict):
            for sub_category, keywords in details.items():
                if business_type in keywords:
                    return category
        elif isinstance(details, list):
            if business_type in details:
                return category
    return "기타 (분류 불가)"  # 어떤 카테고리에도 속하지 않을 경우

def with_category_classification(func):
    """카테고리 분류 기능을 함수에 적용하는 데코레이터"""
    def wrapper(*args, **kwargs):
        # 카테고리 정보 및 분류 함수를 함수의 전역 네임스페이스에 추가
        func.__globals__["categories"] = categories
        func.__globals__["classify_category"] = classify_category
        return func(*args, **kwargs)
    return wrapper
```

# 그린카드V2

```python
from IPython.display import Markdown, display
import pandas as pd

# GreenCardV2 클래스 정의
class GreenCardV2:
    def __init__(self):
        # 기본 정보 초기화
        self.annual_fee = {
            'BC제휴': 10000,
            'Master제휴': 12000
        }

        self.foreign_fee = {
            'Master': '1% + 0.25%',
            'BC제휴': '0.25%'
        }

        # 서비스별 적립 정책
        self.point_policy = {
            '전기차/수소차': {'rates': {30: 0.2, 60: 0.4}, 'cap': 20000},
            '공유모빌리티': {'rate': 0.1, 'cap': 5000},
            '대중교통': {'rate': 0.1, 'cap': 5000},
            '고속버스': {'rate': 0.1, 'cap': 5000},
            '커피전문점': {'rate': 0.1, 'cap': 5000},
            '국내가맹점': {'rate': 0.002, 'cap': {30: 30000, 60: 40000}}
        }

# PointCalculator 클래스 정의
class PointCalculator:
    def __init__(self, card):
        self.card = card
        self.service_list = [
            '전기차/수소차', '공유모빌리티',
            '대중교통', '고속버스', '커피전문점'
        ]

    def determine_tier(self, prev_month):
        if prev_month >= 600000:
            return 60
        elif prev_month >= 300000:
            return 30
        return 0

    def calculate(self, spending):
        tier = self.determine_tier(spending['전월이용금액'])
        if tier == 0:
            return {}

        points = {}

        # 특화 서비스 계산
        total_special = 0
        for service in self.service_list:
            amount = spending.get(service, 0)
            if service == '전기차/수소차':
                rate = self.card.point_policy[service]['rates'][tier]
                points[service] = min(amount * rate, self.card.point_policy[service]['cap'])
            else:
                points[service] = min(amount * self.card.point_policy[service]['rate'],
                                       self.card.point_policy[service]['cap'])
            total_special += points[service]

        # 통합한도 적용
        limit = 40000 if tier == 60 else 30000
        if total_special > limit:
            ratio = limit / total_special
            for service in self.service_list:
                points[service] = round(points[service] * ratio)

        # 국내가맹점 계산
        domestic = spending.get('국내가맹점', 0)
        domestic_cap = self.card.point_policy['국내가맹점']['cap'][tier]
        points['국내가맹점'] = min(domestic * 0.002, domestic_cap)

        return points

# 시뮬레이션 엔진
class GreenCardSimulator:
    def __init__(self, card):
        self.card = card
        self.calculator = PointCalculator(card)

    def calculate_points(self, spending):
        return self.calculator.calculate(spending)

# 카테고리 분류 함수 (예시)
def classify_category(merchant_name):
    # 가맹점 이름에 따라 카테고리를 분류하는 로직을 작성하세요.
    # 예시: 특정 키워드 포함 여부로 분류
    if "전기차" in merchant_name or "수소차" in merchant_name:
        return "전기차/수소차"
    elif "공유모빌리티" in merchant_name:
        return "공유모빌리티"
    elif "대중교통" in merchant_name:
        return "대중교통"
    elif "고속버스" in merchant_name:
        return "고속버스"
    elif "커피" in merchant_name:
        return "커피전문점"
    else:
        return None

# 혜택 계산 함수
def calculate_benefits(df, months, simulator, annual_fee):
    """혜택을 계산하는 함수"""
    # 카테고리 분류 적용
    df['카테고리'] = df['가맹점업종명'].apply(classify_category)

    # 기준월 설정 (202312)
    target_month = 202312

    benefits_list = []

    for customer_id in df['고객ID'].unique():
        customer_data = df[df['고객ID'] == customer_id]
        total_benefit = 0

        # 각 월별로 혜택 계산
        for i in range(months):
            year = target_month // 100
            month = target_month % 100 - i
            if month <= 0:
                month += 12
                year -= 1
            current_month = year * 100 + month

            # 해당 월의 데이터 추출
            month_data = customer_data[customer_data['거래년월'] == current_month]

            # 카테고리별 지출 금액 계산
            spending = {
                '전월이용금액': customer_data[customer_data['거래년월'] == (current_month - 1)]['승인금액'].sum(),
                '전기차/수소차': month_data[month_data['카테고리'] == '전기차/수소차']['승인금액'].sum(),
                '공유모빌리티': month_data[month_data['카테고리'] == '공유모빌리티']['승인금액'].sum(),
                '대중교통': month_data[month_data['카테고리'] == '대중교통']['승인금액'].sum(),
                '고속버스': month_data[month_data['카테고리'] == '고속버스']['승인금액'].sum(),
                '커피전문점': month_data[month_data['카테고리'] == '커피전문점']['승인금액'].sum(),
                '국내가맹점': month_data[month_data['카테고리'].isna()]['승인금액'].sum()
            }

            # 시뮬레이터를 사용하여 포인트 계산
            points = simulator.calculate_points(spending)
            month_benefit = sum(points.values()) if points else 0
            total_benefit += month_benefit

        # 연회비 차감 (개월 수 비례하여 차감)
        net_benefit = total_benefit - (annual_fee * (months / 12))

        # 결과 저장
        benefits_list.append({
            '고객ID': customer_id,
            f'{months}개월_혜택': net_benefit
        })

    return pd.DataFrame(benefits_list)

# 실행 예시
if __name__ == "__main__":
    gc = GreenCardV2()
    simulator = GreenCardSimulator(gc)

    # 데이터 로드 (데이터셋 파일 경로 확인 필요)
    df = pd.read_csv('card_data_2023_sample1_utf8.csv', encoding='utf-8')

    # 각 기간별 혜택 계산
    annual_fee = gc.annual_fee['BC제휴']
    result_1 = calculate_benefits(df, 1, simulator, annual_fee)
    result_3 = calculate_benefits(df, 3, simulator, annual_fee)
    result_6 = calculate_benefits(df, 6, simulator, annual_fee)
    result_12 = calculate_benefits(df, 12, simulator, annual_fee)

    # 결과 병합 및 출력
    final_result = result_1.merge(result_3, on='고객ID', how='outer') \
                           .merge(result_6, on='고객ID', how='outer') \
                           .merge(result_12, on='고객ID', how='outer')

    final_result.columns = ['고객ID', '1개월_혜택', '3개월_혜택', '6개월_혜택', '12개월_혜택']

    print("\n### 각 기간별 혜택 계산 결과")
    print(final_result)

    # 결과 저장 (파일 경로 확인 필요)
    final_result.to_csv('im_GreenCardV2_혜택.csv', index=False, encoding='utf-8')
```

```python
import pandas as pd

class AnywhereGreen:
    def __init__(self):
        # 통합 적립한도 체계
        self.consolidated_limits = {
            'tier1': {'min': 300000, 'max': 599999, 'limit': 30000},
            'tier2': {'min': 600000, 'limit': 40000}
        }

        # 서비스별 적립 정책
        self.point_policy = {
            '전기차/수소차': {'rates': {30: 0.2, 60: 0.4}, 'cap': 20000},
            '공유모빌리티': {'rate': 0.1, 'cap': 5000},
            '대중교통': {'rate': 0.1, 'cap': 5000},
            '고속버스': {'rate': 0.1, 'cap': 5000},
            '커피전문점': {'rate': 0.1, 'cap': 5000},
            '국내가맹점': {'rate': 0.002, 'cap': {30: 30000, 60: 40000}}
        }

        # 연회비 정보
        self.annual_fee = 15000

class PointCalculator:
    def __init__(self, card):  # card 인스턴스를 받아옴
        self.card = card
        self.service_list = [
            '전기차/수소차', '공유모빌리티',
            '대중교통', '고속버스', '커피전문점'
        ]

    def determine_tier(self, prev_month):
        if prev_month >= 600000:
            return 60
        elif prev_month >= 300000:
            return 30
        return 0

    def calculate(self, spending):
        tier = self.determine_tier(spending['전월이용금액'])
        if tier == 0:
            return {}

        points = {}

        # 1. 특화 서비스 계산
        total_special = 0
        for service in self.service_list:
            amount = spending.get(service, 0)
            if service == '전기차/수소차':
                rate = self.card.point_policy[service]['rates'][tier]
                points[service] = min(amount * rate, self.card.point_policy[service]['cap'])
            else:
                points[service] = min(amount * self.card.point_policy[service]['rate'],
                                       self.card.point_policy[service]['cap'])
            total_special += points[service]

        # 2. 통합한도 적용
        limit = 40000 if tier == 60 else 30000
        if total_special > limit:
            ratio = limit / total_special
            for service in self.service_list:
                points[service] = round(points[service] * ratio)

        # 3. 국내가맹점 계산
        domestic = spending.get('국내가맹점', 0)
        domestic_cap = self.card.point_policy['국내가맹점']['cap'][tier]
        points['국내가맹점'] = min(domestic * 0.002, domestic_cap)

        return points

# 시뮬레이션 엔진
class CardSimulator:
    def __init__(self, card):
        self.card = card
        self.calculator = PointCalculator(card)  # AnywhereGreen 인스턴스 전달

    def calculate_points(self, spending):
        """각 사용 시나리오별 포인트 계산"""
        return self.calculator.calculate(spending)

    def calculate_monthly_benefit(self, monthly_spending):
        """월별 혜택 계산"""
        points = self.calculate_points(monthly_spending)
        return sum(points.values()) if points else 0

    def calculate_annual_benefit(self, monthly_spendings):
        """연간 혜택 계산 (연회비 차감)"""
        total_benefit = sum(self.calculate_monthly_benefit(spending) for spending in monthly_spendings)
        return total_benefit - self.card.annual_fee

def calculate_benefits(df, months, simulator):
    """
    각 고객 ID에 대해 지정된 기간 동안의 혜택을 계산합니다.
    """
    target_month = 202312
    benefits_list = []

    for customer_id in df['고객ID'].unique():
        customer_data = df[df['고객ID'] == customer_id]
        total_benefit = 0
        
        # 이전 달의 소비를 저장할 딕셔너리 초기화
        previous_month_spending = {}

        for i in range(months):
            year = target_month // 100
            month = target_month % 100 - i
            if month <= 0:
                month += 12
                year -= 1
            current_month = year * 100 + month
            
            # 이전 달 계산
            prev_year = year
            prev_month = month -1
            if prev_month <=0:
                prev_month +=12
                prev_year -=1
            previous_month = prev_year * 100 + prev_month

            month_data = customer_data[customer_data['거래년월'] == current_month]
            
            prev_month_data = customer_data[customer_data['거래년월'] == previous_month]

            # 월별 소비 내역 집계
            total_spending = month_data['승인금액'].sum()
            
            total_prev_spending = prev_month_data['승인금액'].sum()

            # 서비스별 소비 금액
            spending = {
                '전월이용금액': total_prev_spending,
                '국내가맹점': total_spending,  # 모든 승인금액을 국내가맹점 소비로 간주
                '전기차/수소차': 0,
                '공유모빌리티': 0,
                '대중교통': 0,
                '고속버스': 0,
                '커피전문점': 0
            }

            # 포인트 계산
            points = simulator.calculate_points(spending)
            total_benefit += sum(points.values()) if points else 0

        # 연회비 차감 (개월 수 비례하여 차감) 및 현금으로 변환
        net_benefit = (total_benefit - (card.annual_fee * (months / 12))) * 1  # 1포인트 = 1원
        benefits_list.append({
            '고객ID': customer_id,
            f'{months}개월_혜택': net_benefit
        })

    return pd.DataFrame(benefits_list)

# 실행 예시
if __name__ == "__main__":
    # 카드 및 시뮬레이터 초기화
    card = AnywhereGreen()
    # annual_fee = 15000  # 어디로든 그린카드 연회비 ## removing annual fee from here
    simulator = CardSimulator(card)

    # 데이터 로드
    df = pd.read_csv('card_data_2023_sample1_utf8.csv', encoding='utf-8')

    # 각 기간별 혜택 계산
    result_1 = calculate_benefits(df, 1, simulator)
    result_3 = calculate_benefits(df, 3, simulator)
    result_6 = calculate_benefits(df, 6, simulator)
    result_12 = calculate_benefits(df, 12, simulator)

    # 결과 병합
    final_result = result_1.merge(result_3, on='고객ID', how='outer') \
                           .merge(result_6, on='고객ID', how='outer') \
                           .merge(result_12, on='고객ID', how='outer')

    # 컬럼명 변경 (이미지 결과와 동일하게)
    final_result.columns = ['고객ID', '1개월_혜택', '3개월_혜택', '6개월_혜택', '12개월_혜택']

    # 결과 출력
    print("\n### 각 기간별 혜택 계산 결과")
    print(final_result)

    # CSV 파일로 저장
    final_result.to_csv('im_anywhereGreen.csv', index=False, encoding='utf-8')
```

# 스카이패스 V2 Silver

```python
from IPython.display import Markdown, display
import pandas as pd
from datetime import datetime, timedelta

class SkyPassCard:
    def __init__(self):
        # 기본 정보 초기화 (검색결과 1,3,5)
        self.card_info = {
            '발급대상': '만 19세 이상 개인회원',
            '브랜드': ['BC(국내전용)', 'Mastercard(해외겸용)'],
            '연회비': '19,000원',
            '교통기능': '후불교통 기본 탑재',
            '가족카드규정': '본인 및 가족회원 각각 연회비 적용'
        }
        
        # 마일리지 정책 (검색결과 1,4,5)
        self.mileage_policy = {
            '국내': {'기본': 0.001, '추가': 0, '한도': None},
            '해외': {'기본': 0.001, '추가': 0, '한도': None},
            '전월실적_기준': 300000,
            '제외항목': [
                '단기카드대출', '장기카드대출', '연회비', '제세공과금',
                '아파트관리비', '상품권구매', '무이자할부', '후불교통'
            ],
            '신규회원_특례': '발급월+1개월까지 실적 미충족 시에도 적립'
        }

class MileageSystem:
    def __init__(self, is_new_member=False):
        self.is_new_member = is_new_member
        
    def calculate_miles(self, spending):
        results = {'기본': 0, '추가': 0}
        
        # 전월실적 조건 검사 (검색결과 1,4)
        prev_month = spending.get('전월이용금액',0)
        if prev_month < 300000 and not (self.is_new_member and spending.get('신규카드',False)):
            return results
        
        # 적립 대상 금액 계산
        valid_domestic = spending.get('국내쇼핑', 0)  # 국내쇼핑만 적립 대상으로 변경
        valid_foreign = spending.get('해외이용', 0)
        
        # 국내 마일리지 계산
        results['기본'] += int(valid_domestic * 0.001)
        
        # 해외 마일리지 계산
        results['기본'] += int(valid_foreign * 0.001)
        
        return results

class ForeignTransaction:
    def __init__(self):
        self.fx_rates = {
            'USD': 1300.5,  # 2025-02-20 기준
            'EUR': 1432.1,
            'JPY': 9.85
        }
        
    def convert_amount(self, amount, currency):
        return amount * self.fx_rates.get(currency, 1)
    
    def calculate_fees(self, amount):
        # 수수료 계산 (검색결과 1,5)
        브랜드수수료 = amount * 0.01
        해외서비스수수료 = amount * 0.0025
        return 브랜드수수료 + 해외서비스수수료

class CardSimulator:
    def __init__(self):
        self.history = []
    
    def add_scenario(self, name, spending, is_foreign=False):
        calculator = MileageSystem()
        
        # 해외거래 처리
        if is_foreign:
            fx = ForeignTransaction()
            converted = fx.convert_amount(spending.get('해외이용',0), 'USD')
            fees = fx.calculate_fees(converted)
            spending['해외이용_수수료'] = fees
        
        # 마일리지 계산
        miles = calculator.calculate_miles(spending)
        
        self.history.append({
            'name': name,
            'spending': spending,
            'miles': miles,
        })
    
    def show_results(self):
        for scenario in self.history:
            display(Markdown(f"### {scenario['name']}"))
            print("▷ 소비 내역:")
            print(pd.DataFrame(scenario['spending'].items(), columns=['항목', '금액']))
            print("\n▷ 마일리지 적립 결과:")
            print(pd.DataFrame(scenario['miles'].items(), columns=['유형', '마일리지']))

def calculate_benefits(df, months):
    """Calculate total benefits (mileage value) over a period for each customer."""
    mileage_value_per_point = 20  # 1마일 = 20원
    
    card = SkyPassCard()
    simulator = CardSimulator()
    
    results = []
    
    for customer_id in df['고객ID'].unique():
        customer_data = df[df['고객ID'] == customer_id]
        
        total_mileage = 0  # Initialize total_mileage here
        total_mileage_value = 0  # Initialize total_mileage_value here
        
        # 기준 월 설정
        current_year, current_month = 2023, 12
        
        for i in range(months):
            year = current_year
            month = current_month - i
            
            if month <= 0:
                month += 12
                year -= 1
            
            # 해당 월의 데이터
            month_data = customer_data[
                (customer_data['거래년월'] // 100 == year) &
                (customer_data['거래년월'] % 100 == month)
            ]
            
            # 월별 소비 금액 계산
            total_spending = month_data['승인금액'].sum()
            
            # 전월 소비 금액 계산 (마일리지 적립 기준)
            prev_year, prev_month = year, month - 1
            if prev_month <= 0:
                prev_month = 12
                prev_year -= 1
                
            prev_month_data = customer_data[
                (customer_data['거래년월'] // 100 == prev_year) &
                (customer_data['거래년월'] % 100 == prev_month)
            ]
            prev_month_spending = prev_month_data['승인금액'].sum()

            # 소비 내역
            spending = {
                '전월이용금액': prev_month_spending,
                '국내쇼핑': total_spending,  # 국내 쇼핑 금액만 고려
                '해외이용': 0,             # 해외 이용은 0으로 설정
            }
            
            # 시나리오 추가 및 마일리지 계산
            simulator.add_scenario(f"Customer {customer_id} - {year}-{month}", spending)

            # 각 시나리오의 마일리지를 합산
            for scenario in simulator.history:
                total_mileage += sum(scenario['miles'].values())

        total_mileage_value = total_mileage * mileage_value_per_point
        
        # 연회비 차감
        annual_fee = 19000 * (months/12) # changed 10000 to 19000 to reflect skypass card's annual fee.
        net_benefit = total_mileage_value - annual_fee

        results.append({
            '고객ID': customer_id,
            f'{months}개월_마일리지': total_mileage,
            f'{months}개월_혜택': net_benefit
        })

        # Reset simulator history for the next customer. Important!
        simulator.history = []

    return pd.DataFrame(results)

# 실행 예시
if __name__ == "__main__":
    # 데이터 로드
    df = pd.read_csv('card_data_2023_sample1_utf8.csv', encoding='utf-8')
    
    # 4. 각 기간별 혜택 계산 및 결과 병합
    result_1 = calculate_benefits(df, 1)
    result_3 = calculate_benefits(df, 3)
    result_6 = calculate_benefits(df, 6)
    result_12 = calculate_benefits(df, 12)

    # 결과 병합
    final_result = result_1.merge(result_3, on='고객ID', how='outer') \
                           .merge(result_6, on='고객ID', how='outer') \
                           .merge(result_12, on='고객ID', how='outer')

    # 컬럼명 변경 (이미지 결과와 동일하게)
    final_result.columns = ['고객ID', '1개월_마일리지', '1개월_혜택', '3개월_마일리지', '3개월_혜택', 
                             '6개월_마일리지', '6개월_혜택', '12개월_마일리지', '12개월_혜택']

    # 결과 출력
    print("\n### 각 기간별 혜택 계산 결과")
    print(final_result)

    # CSV 파일로 저장
    final_result.to_csv('im_skysilver.csv', index=False, encoding='utf-8')
```

# 스카이패스 V2 Gold

```python
from IPython.display import Markdown, display
import pandas as pd
from datetime import datetime, timedelta

class SkyPassGoldCard:
    def __init__(self):
        # 기본 정보 초기화
        self.card_info = {
            '발급대상': '개인회원(가족카드 가능)',
            '브랜드': ['BC(국내전용)', 'Mastercard(해외겸용)'],
            '연회비': '35,000원',
            '교통기능': '후불교통 기본 탑재',
            '가족카드규정': '본인 및 가족회원 각각 연회비 적용'
        }
        
        # 마일리지 정책
        self.mileage_policy = {
            '국내': {'기본': 0.001, '추가': 0, '한도': None},
            '해외': {'기본': 0.001, '추가': 0.001, '한도': 2000},
            '전월실적_기준': 300000,
            '제외항목': [
                '단기카드대출', '장기카드대출', '연회비', '제세공과금',
                '아파트관리비', '상품권구매', '무이자할부', '후불교통'
            ]
        }
        
        # 라운지 혜택
        self.lounge_policy = {
            '월_제공': 1,
            '연_제공': 2,
            '전월실적_기준': 300000,
            '사용조건': 'Mastercard 보유 및 정상상태 유지'
        }

class MileageSystem:
    def __init__(self, is_new_member=False):
        self.is_new_member = is_new_member
        self.remaining_lounge = 2  # 연간 잔여 라운지 이용 횟수
        
    def calculate_miles(self, spending):
        results = {'기본': 0, '추가': 0}
        
        # 전월실적 조건 검사
        prev_month = spending.get('전월이용금액', 0)
        if prev_month < 300000 and not (self.is_new_member and spending.get('신규카드', False)):
            return results
        
        # 적립 대상 금액 계산
        valid_domestic = spending.get('국내쇼핑', 0)  # 국내쇼핑만 적립 대상으로 변경
        valid_foreign = spending.get('해외이용', 0)
        
        # 국내 마일리지 계산
        results['기본'] += int(valid_domestic * 0.001)
        
        # 해외 마일리지 계산
        base_foreign = int(valid_foreign * 0.001)
        additional = min(int(valid_foreign * 0.001), 2000)
        results['기본'] += base_foreign
        results['추가'] = additional
        
        return results
    
    def use_lounge(self, prev_month_spending):
        if self.remaining_lounge > 0 and prev_month_spending >= 300000:
            self.remaining_lounge -= 1
            return True
        return False

class ForeignTransaction:
    def __init__(self):
        self.fx_rates = {
            'USD': 1300.5,  # 2025-02-20 기준
            'EUR': 1432.1,
            'JPY': 9.85
        }
        
    def convert_amount(self, amount, currency):
        return amount * self.fx_rates.get(currency, 1)
    
    def calculate_fees(self, amount):
        # 수수료 계산
        브랜드수수료 = amount * 0.01
        해외서비스수수료 = amount * 0.0025
        return 브랜드수수료 + 해외서비스수수료

class CardSimulator:
    def __init__(self):
        self.history = []
    
    def add_scenario(self, name, spending, is_foreign=False):
        calculator = MileageSystem()
        lounge_used = False
        
        # 해외거래 처리
        if is_foreign:
            fx = ForeignTransaction()
            converted = fx.convert_amount(spending.get('해외이용',0), 'USD')
            fees = fx.calculate_fees(converted)
            spending['해외이용_수수료'] = fees
        
        # 마일리지 계산
        miles = calculator.calculate_miles(spending)
        
        # 라운지 이용
        if spending.get('라운지이용', False):
            lounge_used = calculator.use_lounge(spending['전월이용금액'])
        
        self.history.append({
            'name': name,
            'spending': spending,
            'miles': miles,
            'lounge_used': lounge_used,
            'remaining_lounge': calculator.remaining_lounge
        })
    
    def show_results(self):
        for scenario in self.history:
            display(Markdown(f"### {scenario['name']}"))
            print("▷ 소비 내역:")
            print(pd.DataFrame(scenario['spending'].items(), columns=['항목', '금액']))
            print("\n▷ 마일리지 적립 결과:")
            print(pd.DataFrame(scenario['miles'].items(), columns=['유형', '마일리지']))
            print(f"\n▷ 라운지 이용: {'가능' if scenario['lounge_used'] else '불가'}")
            print(f"잔여 라운지: {scenario['remaining_lounge']}회")

def calculate_benefits(df, months):
    """Calculate total benefits (mileage value) over a period for each customer."""
    mileage_value_per_point = 20  # 1마일 = 20원
    
    card = SkyPassGoldCard()
    simulator = CardSimulator()
    
    results = []
    
    for customer_id in df['고객ID'].unique():
        customer_data = df[df['고객ID'] == customer_id]
        
        total_mileage = 0
        
        # 기준 월 설정
        current_year, current_month = 2023, 12
        
        for i in range(months):
            year = current_year
            month = current_month - i
            
            if month <= 0:
                month += 12
                year -= 1
            
            # 해당 월의 데이터
            month_data = customer_data[
                (customer_data['거래년월'] // 100 == year) &
                (customer_data['거래년월'] % 100 == month)
            ]
            
            # 월별 소비 금액 계산
            total_spending = month_data['승인금액'].sum()
            
            # 전월 소비 금액 계산 (마일리지 적립 기준)
            prev_year, prev_month = year, month - 1
            if prev_month <= 0:
                prev_month = 12
                prev_year -= 1
                
            prev_month_data = customer_data[
                (customer_data['거래년월'] // 100 == prev_year) &
                (customer_data['거래년월'] % 100 == prev_month)
            ]
            prev_month_spending = prev_month_data['승인금액'].sum()

            # 소비 내역
            spending = {
                '전월이용금액': prev_month_spending,
                '국내쇼핑': total_spending,  # 국내 쇼핑 금액만 고려
                '해외이용': 0,             # 해외 이용은 0으로 설정
                '라운지이용': False
            }
            
            # 시나리오 추가 및 마일리지 계산
            simulator.add_scenario(f"Customer {customer_id} - {year}-{month}", spending)

        # 총 마일리지 계산 (모든 시나리오를 합산)
        total_mileage = sum(sum(scenario['miles'].values()) for scenario in simulator.history)
        total_mileage_value = total_mileage * mileage_value_per_point

        # 연회비 차감
        annual_fee = 35000 * (months / 12)
        net_benefit = total_mileage_value - annual_fee

        results.append({
            '고객ID': customer_id,
            f'{months}개월_마일리지': total_mileage,
            f'{months}개월_혜택': net_benefit
        })
        
        # Reset simulator history for the next customer
        simulator.history = []

    return pd.DataFrame(results)

# 실행 예시
if __name__ == "__main__":
    # 데이터 로드
    df = pd.read_csv('card_data_2023_sample1_utf8.csv', encoding='utf-8')
    
    # 4. 각 기간별 혜택 계산 및 결과 병합
    result_1 = calculate_benefits(df, 1)
    result_3 = calculate_benefits(df, 3)
    result_6 = calculate_benefits(df, 6)
    result_12 = calculate_benefits(df, 12)

    # 결과 병합
    final_result = result_1.merge(result_3, on='고객ID', how='outer') \
                           .merge(result_6, on='고객ID', how='outer') \
                           .merge(result_12, on='고객ID', how='outer')

    # 컬럼명 변경 (이미지 결과와 동일하게)
    final_result.columns = ['고객ID', '1개월_마일리지', '1개월_혜택', '3개월_마일리지', '3개월_혜택', 
                             '6개월_마일리지', '6개월_혜택', '12개월_마일리지', '12개월_혜택']

    # 결과 출력
    print("\n### 각 기간별 혜택 계산 결과")
    print(final_result)

    # CSV 파일로 저장
    final_result.to_csv('im_skygold.csv', index=False, encoding='utf-8')
```

# 세븐캐쉬백 카드

```python
import pandas as pd
from datetime import datetime, timedelta

class SevenCashbackCard:
    def __init__(self):
        # 카드 기본 정보
        self.card_info = {
            '발급대상': '만 19세 이상 개인회원',
            '브랜드': ['BC(국내전용)', 'Master(해외겸용)'],
            '연회비': {'BC': 10000, 'Master': 12000},
            '교통기능': '후불교통 기본 탑재',
            '가족카드': '발급 가능'
        }
        
        # 캐시백 정책
        self.cashback_policy = {
            'rate': 0.07,
            'min_transaction': 30000,
            'daily_limit': 5000,
            'monthly_tiers': {
                300000: 10000,
                600000: 20000,
                1200000: 40000
            },
            'exclusions': [
                '할부', '후불교통', '제세공과금', 
                '상품권', '대학등록금', '아파트관리비'
            ]
        }

class CashbackCalculator:
    def __init__(self, card, prev_month_spending):
        self.card = card
        self.prev_tier = self._determine_tier(prev_month_spending)
        self.monthly_limit = self._get_monthly_limit()
        self.used_cashback = 0
        self.last_transaction_date = None
        
    def _determine_tier(self, amount):
        tiers = sorted(self.card.cashback_policy['monthly_tiers'].keys(), reverse=True)
        for tier in tiers:
            if amount >= tier:
                return tier
        return 0  # 수정: tiers에 해당되는 값이 없을 경우 0을 반환

    
    def _get_monthly_limit(self):
        return self.card.cashback_policy['monthly_tiers'].get(self.prev_tier, 0)
    
    def calculate(self, date, amount, category):
        # 조건 검증
        if category in self.card.cashback_policy['exclusions']:
            return 0
        if amount < self.card.cashback_policy['min_transaction']:
            return 0
        
        # 일일 제한 확인
        if self.last_transaction_date == date:
            return 0
        
        # 캐시백 계산
        cashback = min(amount * self.card.cashback_policy['rate'], 
                      self.card.cashback_policy['daily_limit'])
        
        # 월 한도 적용
        remaining = self.monthly_limit - self.used_cashback
        final_cashback = min(cashback, remaining)
        
        if final_cashback > 0:
            self.used_cashback += final_cashback
            self.last_transaction_date = date
            
        return final_cashback

# 각 기간별 혜택 계산 함수
def calculate_monthly_benefit(df, year, month, calculator, card):
    """Calculate monthly cashback benefit for a customer."""
    total_cashback = 0
    
    # 해당 월의 데이터만 필터링
    month_data = df[(df['거래년월'] // 100 == year) & (df['거래년월'] % 100 == month)]
    
    # 거래일자별로 정렬
    month_data = month_data.sort_values(by='거래년월')

    # 일별 캐시백 계산
    last_transaction_date = None  # Reset daily transaction date
    for _, row in month_data.iterrows():
        date = datetime(year, month, 1).date()  # Use the first day of the month
        amount = row['승인금액']
        category = row['가맹점업종명']

        # 이미 해당 날짜에 거래가 있는 경우, 건너뜀
        if last_transaction_date == date:
            continue
        
        cashback = calculator.calculate(date, amount, category)
        total_cashback += cashback
        
        if cashback > 0:
            last_transaction_date = date  # Update last transaction date
        
    return total_cashback

# 전체 기간에 대한 혜택 계산 함수
def calculate_total_benefits(df, months, annual_fee, card):
    """Calculate total benefits over a period, considering annual fee."""
    customer_benefits = {}

    for customer_id in df['고객ID'].unique():
        customer_data = df[df['고객ID'] == customer_id]
        total_benefit = 0

        # 기준 월 설정
        current_year, current_month = 2023, 12
        
        for i in range(months):
            year = current_year
            month = current_month - i
            
            if month <= 0:
                month += 12
                year -= 1
            
            # 전월 실적 계산 (이전 달의 총 승인 금액)
            prev_year, prev_month = year, month - 1
            if prev_month <= 0:
                prev_month = 12
                prev_year -= 1
            
            prev_month_data = customer_data[
                (customer_data['거래년월'] // 100 == prev_year) &
                (customer_data['거래년월'] % 100 == prev_month)
            ]
            
            prev_month_spending = prev_month_data['승인금액'].sum()

            # CashbackCalculator 객체 생성
            calculator = CashbackCalculator(card, prev_month_spending)

            # 월별 혜택 계산
            monthly_benefit = calculate_monthly_benefit(
                customer_data, year, month, calculator, card
            )
            total_benefit += monthly_benefit

        # 연회비 차감 (개월 수 비례하여 차감)
        net_benefit = total_benefit - (annual_fee * (months / 12))
        customer_benefits[customer_id] = net_benefit

    # 결과를 DataFrame으로 변환
    customer_ids = list(customer_benefits.keys())
    benefits = list(customer_benefits.values())

    results = pd.DataFrame({'고객ID': customer_ids, f'{months}개월_혜택': benefits})
    return results

# 실행 예시
if __name__ == "__main__":
    # 카드 정보 초기화
    card = SevenCashbackCard()

    # 데이터 로드
    df = pd.read_csv('card_data_2023_sample1_utf8.csv', encoding='utf-8')

    # 연회비 설정
    annual_fee = card.card_info['연회비']['BC']

    # 각 기간별 혜택 계산
    result_1 = calculate_total_benefits(df, 1, annual_fee, card)
    result_3 = calculate_total_benefits(df, 3, annual_fee, card)
    result_6 = calculate_total_benefits(df, 6, annual_fee, card)
    result_12 = calculate_total_benefits(df, 12, annual_fee, card)

    # 결과 병합
    final_result = result_1.merge(result_3, on='고객ID', how='outer') \
                           .merge(result_6, on='고객ID', how='outer') \
                           .merge(result_12, on='고객ID', how='outer')

    # 컬럼명 변경
    final_result.columns = ['고객ID', '1개월_혜택', '3개월_혜택', '6개월_혜택', '12개월_혜택']

    # 결과 출력
    print("\n### 각 기간별 혜택 계산 결과")
    print(final_result)

    # CSV 파일로 저장
    final_result.to_csv('im_7cashback.csv', index=False, encoding='utf-8')
```

# im-one 카드

```python
from IPython.display import Markdown, display
import pandas as pd
from datetime import datetime, timedelta

class IMOneCard:
    def __init__(self):
        # 카드 기본 정보 초기화
        self.card_info = {
            '발급대상': '개인회원 (가족카드 가능)',
            '브랜드': ['BC(국내전용)', 'Master(해외겸용)'],
            '연회비': {'BC': 10000, 'Master': 12000},
            '교통기능': '후불교통 기본 탑재'
        }
        
        # 연회비 면제 조건
        self.annual_fee_condition = 1000000  # 1백만원
        
        # 포인트 적립 정책
        self.point_policy = {
            '스립리스인시애틀': {'rate': 0.1, 'min_spending': 0}, # 최소금액 조건 없음
            '다빈치': {'rate': 0.1, 'min_spending': 0}, # 최소금액 조건 없음
            '주유': {'rate': 0.01, 'min_spending': 0}, # 최소금액 조건 없음
            '쇼핑': {'rate': 0.01, 'min_spending': 0}, # 최소금액 조건 없음
            '여행': {'rate': 0.01, 'min_spending': 0}, # 최소금액 조건 없음
            '건강': {'rate': 0.007, 'min_spending': 0}, # 최소금액 조건 없음
            '교통': {'rate': 0.007, 'min_spending': 0}, # 최소금액 조건 없음
            '생활': {'rate': 0.007, 'min_spending': 0}, # 최소금액 조건 없음
            '기타': {'rate': 0.001, 'min_spending': 0} # 최소금액 조건 없음
        }

        # 캐시백 및 할인 정책
        self.cashback_policy = {
            '외식 할인': {
                '대상': ['아웃백', 'TGIF', 'VIPS'],
                '할인율': 0.2,
                '조건': {'최소금액': 30000, '최대할인': 10000},
                '횟수제한': {'월': 1, '분기': 1}
            },
            '5% 캐쉬백': {
                '대상': ['약국', '편의점', 
                         '이/미용실', 
                         '안경', 
                         '서적', 
                         '제과점'],  # 던킨도너츠, 베스킨라빈스, 뚜레쥬르 -> 제과점으로 통합
                '할인율': 0.05,
                '조건': {'횟수제한 없음'}
            },
            '영화 할인': {
                '대상': ['CGV'],
                '할인금액': 4000,
                '조건': {'최소금액': 10000},
                '횟수제한': {'월': 1, '연': 12}
            }
        }
        
        # 전월 실적 조건 및 할인 한도
        self.discount_limit = {
            300000: 5000,
            600000: 10000,
            1200000: 15000
        }

    def get_annual_fee(self, customer_id, current_year, df, card_type='BC'):
        """전년도 실적에 따른 연회비 계산"""
        # 첫 해에는 무조건 연회비 부과
        if current_year == 2023:  # 첫 해 기준
            return self.card_info['연회비'][card_type]
        else:
            # 전년도 실적 확인 (2차년도부터 면제 조건 적용)
            previous_year = current_year - 1
            previous_year_data = df[(df['고객ID'] == customer_id) & (df['거래년월'] // 100 == previous_year)]
            previous_year_spending = previous_year_data['승인금액'].sum()

            if previous_year_spending >= self.annual_fee_condition:
                return 0  # 연회비 면제
            else:
                return self.card_info['연회비'][card_type]  # 연회비 부과

class CashbackCalculator:
    def __init__(self, card, prev_month_spending):
        self.card = card
        self.prev_month_spending = prev_month_spending
        self.remaining_limit = self._get_discount_limit()
        self.used_restaurant_count = 0  # 월별 레스토랑 할인 사용 횟수
        self.used_movie_count = 0  # 월별 영화 할인 사용 횟수
        self.total_benefits = 0  # Initialize total benefits (cashback + points)
        self.monthly_point = 0 # 월별 포인트 적립액

    def _get_discount_limit(self):
        for tier, limit in sorted(self.card.discount_limit.items(), reverse=True):
            if self.prev_month_spending >= tier:
                return limit
        return 0

    def calculate_cashback_and_points(self, category, amount):
        cashback = 0
        point = 0

        # 외식 할인 계산
        if category in self.card.cashback_policy['외식 할인']['대상']:
            if amount >= self.card.cashback_policy['외식 할인']['조건']['최소금액'] and \
               self.used_restaurant_count < self.card.cashback_policy['외식 할인']['횟수제한']['월']:
                discount = min(amount * self.card.cashback_policy['외식 할인']['할인율'], 
                               self.card.cashback_policy['외식 할인']['조건']['최대할인'])
                available_discount = min(discount, self.remaining_limit)
                cashback = available_discount
                self.remaining_limit -= available_discount
                self.used_restaurant_count += 1  # 레스토랑 할인 횟수 증가

        # 영화 할인 계산
        elif category in self.card.cashback_policy['영화 할인']['대상']:
            if amount >= self.card.cashback_policy['영화 할인']['조건']['최소금액'] and \
               self.used_movie_count < self.card.cashback_policy['영화 할인']['횟수제한']['월']:
                discount = self.card.cashback_policy['영화 할인']['할인금액']
                available_discount = min(discount, self.remaining_limit)
                cashback = available_discount
                self.remaining_limit -= available_discount
                self.used_movie_count += 1

        # 5% 캐시백 계산
        elif category in self.card.cashback_policy['5% 캐쉬백']['대상']:
            discount = amount * self.card.cashback_policy['5% 캐쉬백']['할인율']
            available_discount = min(discount, self.remaining_limit)
            cashback = available_discount
            self.remaining_limit -= available_discount

        # 포인트 적립 계산
        else:
            for policy_category, policy in self.card.point_policy.items():
                if policy_category in category or category in policy_category:
                    if policy['min_spending'] <= amount:
                        point = amount * policy['rate']
                        self.monthly_point += point # 월별 포인트 합산
                        break # 해당하는 정책을 찾았으면 반복문 종료
            else: #적립 정책에 해당하지 않는 경우
                point = amount * self.card.point_policy['기타']['rate']
                self.monthly_point += point # 월별 포인트 합산

        total_benefit = cashback + point # 캐쉬백 + 포인트
        self.total_benefits += total_benefit
        return total_benefit

def classify_category(merchant_name):
    """가맹점명을 기반으로 카테고리를 분류하는 함수"""
    merchant_name = merchant_name.lower()

    if "아웃백" in merchant_name or "tgif" in merchant_name or "vips" in merchant_name:
        return "외식 할인"
    elif "약국" in merchant_name:
        return "약국"
    elif "편의점" in merchant_name:
        return "편의점"
    elif "이미용실" in merchant_name or "미용실" in merchant_name:
        return "이/미용실"
    elif "안경" in merchant_name:
        return "안경"
    elif "서적" in merchant_name or "책" in merchant_name:
        return "서적"
    elif "파리바게트" in merchant_name or "뚜레쥬르" in merchant_name or "크라운베이커리" in merchant_name:
        return "제과점"
    elif "cgv" in merchant_name:
        return "CGV"
    elif "스립리스인시애틀" in merchant_name.lower():
        return "스립리스인시애틀"
    elif "다빈치" in merchant_name.lower():
        return "다빈치"
    elif "sk에너지" in merchant_name.lower() or "주유소" in merchant_name.lower():
        return "주유"
    elif "홈플러스" in merchant_name.lower() or "롯데홈쇼핑" in merchant_name.lower() or "the shilla" in merchant_name.lower():
        return "쇼핑"
    elif "병원" in merchant_name.lower() or "한의원" in merchant_name.lower():
        return "건강"
    elif "택시" in merchant_name.lower() or "고속버스" in merchant_name.lower() or "기차" in merchant_name.lower():
        return "교통"
    elif "음식점" in merchant_name.lower() or "마트" in merchant_name.lower() or "슈퍼마켓" in merchant_name.lower():
        return "생활"

    return "기타"  # 어떤 카테고리에도 해당하지 않는 경우

def calculate_total_benefits(df, months, card):
    """Calculate total benefits over a period for each customer."""
    customer_benefits = {}

    # 필요한 컬럼들만 선택
    df = df[['고객ID', '거래년월', '가맹점업종명', '승인금액']]

    for customer_id in df['고객ID'].unique():
        customer_data = df[df['고객ID'] == customer_id].copy() # SettingWithCopyWarning 방지

        # 카테고리 분류
        customer_data['카테고리'] = customer_data['가맹점업종명'].apply(classify_category)

        # 기준 월 설정
        current_year, current_month = 2023, 12
        
        total_net_benefit = 0  # Initialize net benefit for the customer

        # 카드 종류 설정 (예시로 'BC'로 설정)
        card_type = 'BC'

        # 연회비 계산
        annual_fee = card.get_annual_fee(customer_id, current_year, df, card_type)

        for i in range(months):
            year = current_year
            month = current_month - i
            
            if month <= 0:
                month += 12
                year -= 1

            # 해당 월의 데이터
            month_data = customer_data[((customer_data['거래년월'] // 100) == year) & ((customer_data['거래년월'] % 100) == month)]

            # 전월 실적 계산 (이전 달의 총 승인 금액)
            prev_year, prev_month = year, month - 1
            if prev_month <= 0:
                prev_month = 12
                prev_year -= 1
            prev_month_data = customer_data[((customer_data['거래년월'] // 100) == prev_year) & ((customer_data['거래년월'] % 100) == prev_month)]
            prev_month_spending = prev_month_data['승인금액'].sum()

            # CashbackCalculator 객체 생성
            cashback_calc = CashbackCalculator(card, prev_month_spending)
            cashback_calc.monthly_point = 0 # 월별 포인트 초기화

            monthly_benefit = 0  # Initialize monthly benefit
            
            # 월별 혜택 계산
            for _, row in month_data.iterrows():
                category = row['카테고리']
                amount = row['승인금액']
                total_benefit = cashback_calc.calculate_cashback_and_points(category, amount)
                monthly_benefit += total_benefit

            total_net_benefit += monthly_benefit

        # 총 혜택에서 연회비 차감 (개월 수에 따라 비례하여 차감)
        total_net_benefit -= annual_fee * (months / 12)  # Distribute annual fee over 12 months

        # 고객별 총 혜택 저장
        customer_benefits[customer_id] = total_net_benefit

    # 결과를 DataFrame으로 변환
    customer_ids = list(customer_benefits.keys())
    benefits = list(customer_benefits.values())

    results_df = pd.DataFrame({'고객ID': customer_ids, f'{months}개월_혜택': benefits})
    return results_df

# 실행 예시
if __name__ == "__main__":
    # 카드 정보 초기화
    card = IMOneCard()

    # 데이터 로드
    df = pd.read_csv('card_data_2023_sample1_utf8.csv', encoding='utf-8')

    # 각 기간별 혜택 계산
    result_1 = calculate_total_benefits(df, 1, card)
    result_3 = calculate_total_benefits(df, 3, card)
    result_6 = calculate_total_benefits(df, 6, card)
    result_12 = calculate_total_benefits(df, 12, card)

    # 결과 병합
    final_result = result_1.merge(result_3, on='고객ID', how='outer') \
                           .merge(result_6, on='고객ID', how='outer') \
                           .merge(result_12, on='고객ID', how='outer')

    # 컬럼명 변경
    final_result.columns = ['고객ID', '1개월_혜택', '3개월_혜택', '6개월_혜택', '12개월_혜택']

    # 결과 출력
    print("\n### 각 기간별 혜택 계산 결과")
    print(final_result)

    # CSV 파일로 저장
    final_result.to_csv('im_one.csv', index=False, encoding='utf-8')
```
