```python
from sklearn.model_selection import KFold
import statsmodels.api as sm
import numpy as np

scores = np.zeros(5)
scores
```

```python
cv = KFold(5, shuffle=True, random_state=0)

cv
```

```python
feature_names = df.columns[:-1]
print(feature_names)
```

```python
for i, (idx_train, idx_test) in enumerate(cv.split(df)):
    # print(idx_train, '\n', idx_test)
    df_train = df.iloc[idx_train]
    df_test = df.iloc[idx_test]

    model = sm.OLS.from_formula("Target ~ " + "+".join(feature_names), data = df_train) 
    # Target을 종속변수로, 나머지는 + 뒤에 있으니 독립변수
    result = model.fit()

    pred = result.predict(df_test)
    rss = ((df_test.Target - pred) ** 2).sum() # residual squared sum 잔차제곱합
    tss = ((df_test.Target - df_test.Target.mean()) ** 2).sum() # total squared sum 
    rsquared = 1 - rss / tss

    scores[i] = rsquared
    print("학습 R^2 = {:.8f}, 검증 R^2 = {:.8f}".format(result.rsquared, rsquared))
```

```python
from sklearn.metrics import r2_score

scores = np.zeros(5)
cv = KFold(5, shuffle=True, random_state=0)

for i, (idx_train, idx_test) in enumerate(cv.split(df)):
    df_train = df.iloc[idx_train]
    df_test = df.iloc[idx_test]

    model = sm.OLS.from_formula("Target ~ " + "+".join(feature_names), data = df_train) 
    # Target을 종속변수로, 나머지는 + 뒤에 있으니 독립변수
    result = model.fit()

    pred = result.predict(df_test)
    rsquared = r2_score(df_test.Target, pred)
    scores[i] = rsquared

scores
```

```python
from sklearn.base import BaseEstimator, RegressorMixin
import statsmodels.formula.api as smf
import statsmodels.api as sm

class StatsmodelsOLS(BaseEstimator, RegressorMixin):
    def __init__(self, formula):
        self.formula = formula
        self.model = None
        self.data = None
        self.result = None
    
    def fit(self, dfX, dfy):
        self.data = pd.concat([dfX, dfy], axis = 1)
        self.model = smf.ols(self.formula, data = self.data)
        self.result = self.model.fit()
        
    def predict(self, new_data):
        return self.result.predict(new_data)
        
```

```python
from sklearn.model_selection import cross_val_score

model = StatsmodelsOLS("Target ~ " + "+".join(feature_names))
cv = KFold(5, shuffle=True, random_state = 0)
cross_val_score(model, dfX, dfy, scoring="r2", cv=cv)
```
