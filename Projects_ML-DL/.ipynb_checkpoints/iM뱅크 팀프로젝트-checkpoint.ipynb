```python
import pandas as pd
import numpy as np

# 원본 파일 읽기
df = pd.read_csv('unique_customer_ids.csv')

# 전체 행 수 확인
total_rows = len(df)
# 각 파일당 행 수 계산 (올림)
rows_per_file = int(np.ceil(total_rows / 10))

# 10개의 파일로 나누어 저장
for i in range(10):
    # 시작 인덱스와 끝 인덱스 계산
    start_idx = i * rows_per_file
    end_idx = min((i + 1) * rows_per_file, total_rows)
    
    # 부분 데이터프레임 추출
    subset_df = df.iloc[start_idx:end_idx]
    
    # 파일 저장 (파일명: unique_customer_ids_part1.csv, unique_customer_ids_part2.csv, ...)
    subset_df.to_csv(f'unique_customer_ids_part{i+1}.csv', index=False)
    
    # 저장된 파일의 행 수 출력
    print(f'Part {i+1}: {len(subset_df)} rows saved')
```

```python
import pandas as pd

# 데이터 읽기
df = pd.read_csv('iMBANK_CARD_DATA_2023(K-DigitalTraining).csv', encoding = 'cp949')

# 거래년월 필드에서 최근 3개월 데이터 필터링 (2023년 10월, 11월, 12월)
recent_months = [202310, 202311, 202312]
df_recent = df[df['거래년월'].isin(recent_months)]

# 월별 카드 실적 계산: 고객ID와 거래년월별 승인금액 합계
monthly_performance = df_recent.groupby(['고객ID', '거래년월'])['승인금액'].sum().reset_index()

# 월별 카드 실적이 30만 원 이상인 고객 필터링
valid_customers = monthly_performance[monthly_performance['승인금액'] >= 300000]['고객ID']

# 최근 3개월 동안 모든 월에서 카드 실적이 30만 원 이상인 고객만 선택
valid_customers_count = valid_customers.value_counts()
final_customers = valid_customers_count[valid_customers_count == 3].index

# 최종 고객ID를 데이터프레임으로 변환
final_df = pd.DataFrame({'고객ID': final_customers})

# 결과를 CSV 파일로 저장
final_df.to_csv('recent3month_upper30_ids.csv', index=False)

print("조건을 만족하는 고객ID가 recent3month_upper30_ids.csv에 저장되었습니다.")
```

```python
import pandas as pd

# 데이터 읽기
df = pd.read_csv('iMBANK_CARD_DATA_2023(K-DigitalTraining).csv', encoding = 'cp949')
```

```python
# 거래년월을 datetime 형식으로 변환
df['거래년월'] = pd.to_datetime(df['거래년월'].astype(str), format='%Y%m')
```

```python
# 최근 3개월 데이터 필터링 (2023년 10월, 11월, 12월)
recent_months = pd.to_datetime(['202310', '202311', '202312'], format='%Y%m')
df_recent = df[df['거래년월'].isin(recent_months)]

# 최근 3개월 동안 단 한 번도 카드 실적이 30만 원 이상인 적이 없는 고객 제거
monthly_performance_recent = df_recent.groupby(['고객ID', '거래년월'])['승인금액'].sum().reset_index()
valid_recent_customers = monthly_performance_recent[monthly_performance_recent['승인금액'] >= 300000]['고객ID'].unique()

# 최근 3개월 동안 조건을 만족하지 못한 고객 제외
df_filtered = df[df['고객ID'].isin(valid_recent_customers)]

# 각 고객의 가장 오래된 결제 월과 최신 결제 월 계산
customer_min_max_months = df_filtered.groupby('고객ID')['거래년월'].agg(['min', 'max']).reset_index()

# 각 고객의 전체 기간 동안 월별 카드 실적 계산
monthly_performance = df_filtered.groupby(['고객ID', '거래년월'])['승인금액'].sum().reset_index()

# 카드 실적이 30만 원 이상인 월 계산
monthly_performance['30만_이상'] = monthly_performance['승인금액'] >= 300000

# 각 고객의 전체 기간 중 카드 실적이 30만 원 이상인 월 비율 계산
def calculate_valid_month_ratio(row):
    customer_id = row['고객ID']
    min_month = row['min']
    max_month = row['max']
    
    # 해당 고객의 전체 기간 데이터 필터링
    customer_data = monthly_performance[(monthly_performance['고객ID'] == customer_id) & 
                                        (monthly_performance['거래년월'] >= min_month) & 
                                        (monthly_performance['거래년월'] <= max_month)]
    
    # 비율 계산
    total_months = len(customer_data)
    valid_months = customer_data['30만_이상'].sum()
    return valid_months / total_months if total_months > 0 else 0

customer_min_max_months['valid_ratio'] = customer_min_max_months.apply(calculate_valid_month_ratio, axis=1)

# 유효 비율이 30% 이상인 고객만 필터링
valid_customers_final = customer_min_max_months[customer_min_max_months['valid_ratio'] >= 0.3]['고객ID']

# 최종 결과 데이터프레임 생성
final_df = pd.DataFrame({'고객ID': valid_customers_final})

# 결과를 CSV 파일로 저장
final_df.to_csv('filtered_customer_ids.csv', index=False)

print("조건을 만족하는 고객ID가 filtered_customer_ids.csv에 저장되었습니다.")
```

```python
import pandas as pd

# 데이터 로드
file_path = "card_data_2023_sample1_utf8.csv"
df = pd.read_csv(file_path)

# 날짜 데이터 변환 (거래년월을 기준으로 판단)
df['거래년월'] = pd.to_datetime(df['거래년월'], format='%Y%m')
latest_month = df['거래년월'].max()

# 고객별 혜택 계산을 위한 기간 설정
periods = {
    'A카드_최근1개월혜택': 1,
    'A카드_최근3개월혜택': 3,
    'A카드_최근6개월혜택': 6,
    'A카드_최근12개월혜택': 12
}

# 혜택 조건 설정
discount_categories = {
    '이동통신 5% 할인': {'업종': ['이동통신요금'], '최소금액': 50000, '할인율': 0.05},
    '커피·생활용품·편의점 5% 할인': {
        '업종': ['기타음료식품', '화장품', '문구용품', '침구·수예점', '주방용구', '주방용식기', '연 쇄 점', '편 의 점'],
        '최소금액': 10000, '할인율': 0.05
    },
    '영화 5% 할인': {'업종': ['영화관'], '최소금액': 20000, '할인율': 0.05}
}

# 연회비 설정
annual_fee = 5000

# 결과를 저장할 딕셔너리 초기화
result_dict = {'고객ID': []}
for col_name in periods.keys():
    result_dict[col_name] = []

# 고객별 혜택 계산
for customer_id, customer_df in df.groupby('고객ID'):
    result_dict['고객ID'].append(customer_id)
    
    for col_name, months in periods.items():
        start_date = latest_month - pd.DateOffset(months=months)
        df_period = customer_df[customer_df['거래년월'] > start_date]
        
        monthly_benefit = []
        
        for month, month_group in df_period.groupby(df_period['거래년월']):
            total_benefit = 0
            total_performance = 0
            
            for _, row in month_group.iterrows():
                avg_amount = row['승인금액'] / row['승인건수'] if row['승인건수'] > 0 else 0
                applied = False
                
                for discount_name, condition in discount_categories.items():
                    if row['가맹점업종명'] in condition['업종'] and avg_amount >= condition['최소금액']:
                        total_benefit += row['승인금액'] * condition['할인율']
                        applied = True
                
                if not applied:
                    total_performance += row['승인금액']
            
            # 실적 기준 할인 한도 적용
            if total_performance >= 600000:
                total_benefit = min(total_benefit, 10000)
            elif total_performance >= 300000:
                total_benefit = min(total_benefit, 5000)
            else:
                total_benefit = 0
            
            monthly_benefit.append(total_benefit)
        
        # 기간 내 월별 혜택 합산
        total_benefit = sum(monthly_benefit)
        
        # 연회비 적용
        adjusted_benefit = total_benefit - (annual_fee / 12 * months)
        result_dict[col_name].append(adjusted_benefit)

# 결과 데이터프레임 생성
result = pd.DataFrame(result_dict)

# 결과 저장
output_file = "고객혜택금액_A카드.csv"
result.to_csv(output_file, index=False, encoding='utf-8-sig')
print(f"파일 저장 완료: {output_file}")
```

```python
import pandas as pd

# 데이터 로드
file_path = "card_data_2023_sample1_utf8.csv"
df = pd.read_csv(file_path)

# 날짜 데이터 변환 (거래년월을 기준으로 판단)
df['거래년월'] = pd.to_datetime(df['거래년월'], format='%Y%m')
latest_month = df['거래년월'].max()

# 고객별 혜택 계산을 위한 기간 설정
periods = {
    '대백-iM뱅크_최근1개월혜택': 1,
    '대백-iM뱅크_최근3개월혜택': 3,
    '대백-iM뱅크_최근6개월혜택': 6,
    '대백-iM뱅크_최근12개월혜택': 12
}

# 연회비 설정
annual_fee = 2000
fee_per_month = annual_fee / 12

# 혜택 조건 설정
discount_categories = {
    '주유 4% 할인': {
        '업종': ['SK주유소', 'GS주유소', '현대정유(오일뱅크)', '쌍용S-OIL', 'SK가스충전소', 'GS가스충전소',
               '쌍용S-OIL가스충전소', '현대정유가스충전소', 'LPG취급점', '유류판매', '주유소', 'E1가스충전소'],
        '최대결제금액': 100000, '할인율': 0.04, '월제한': 4
    },
    '놀이공원 20% 할인': {'업종': ['종합레져타운', '문화취미기타', '콘도'], '할인율': 0.2, '월제한': 1},
    '영화 2000원 할인': {'업종': ['영화관'], '고정할인': 2000, '월제한': 2, '연제한': 12},
    '철도 5% 할인': {'업종': ['철도'], '최소금액': 2000, '할인율': 0.05, '최대할인': 5000, '월제한': 1},
    '이동통신 3% 할인': {'업종': ['이동통신요금'], '할인율': 0.03, '최대할인': 5000, '월제한': 1},
    '서적 3% 할인': {'업종': ['일반서적', '전문서적', '출판 및 인쇄물', '기타서적문구'], '할인율': 0.03, '최대할인': 5000, '월제한': 1},
    '백화점 5% 할인': {'업종': ['자사카드발행백화점'], '할인율': 0.05, '월제한': 3}
}

# 결과를 저장할 딕셔너리 초기화
result_dict = {'고객ID': []}
for col_name in periods.keys():
    result_dict[col_name] = []

# 고객별 혜택 계산
for customer_id, customer_df in df.groupby('고객ID'):
    result_dict['고객ID'].append(customer_id)
    
    for col_name, months in periods.items():
        start_date = latest_month - pd.DateOffset(months=months)
        df_period = customer_df[customer_df['거래년월'] > start_date]
        
        monthly_benefit = []
        
        for month, month_group in df_period.groupby(df_period['거래년월']):
            total_benefit = 0
            total_performance = 0
            discount_counts = {key: 0 for key in discount_categories.keys()}
            
            for _, row in month_group.iterrows():
                avg_amount = row['승인금액'] / row['승인건수'] if row['승인건수'] > 0 else 0
                applied = False
                
                for discount_name, condition in discount_categories.items():
                    if row['가맹점업종명'] in condition['업종']:
                        if '최소금액' in condition and avg_amount < condition['최소금액']:
                            continue
                        
                        available_uses = min(row['승인건수'], condition.get('월제한', float('inf')) - discount_counts[discount_name])
                        if available_uses <= 0:
                            continue
                        
                        effective_amount = row['승인금액'] * (available_uses / row['승인건수'])
                        
                        if '최대결제금액' in condition:
                            discount_amount = min(effective_amount, condition['최대결제금액']) * condition['할인율']
                        elif '할인율' in condition:
                            discount_amount = effective_amount * condition['할인율']
                        else:
                            discount_amount = condition['고정할인'] * available_uses
                            
                        if '최대할인' in condition:
                            discount_amount = min(discount_amount, condition['최대할인'])
                        
                        total_benefit += discount_amount
                        discount_counts[discount_name] += available_uses
                        applied = True
                
                if not applied:
                    total_performance += row['승인금액']
            
            # 실적 기준 확인
            if total_performance >= 300000:
                monthly_benefit.append(total_benefit)
            else:
                monthly_benefit.append(0)
        
        # 기간 내 월별 혜택 합산 후 연회비 차감
        total_benefit = sum(monthly_benefit)
        total_benefit -= (fee_per_month * months)
        result_dict[col_name].append(total_benefit)

# 결과 데이터프레임 생성
result = pd.DataFrame(result_dict)

# 결과 저장
output_file = "고객혜택금액_대백-iM뱅크.csv"
result.to_csv(output_file, index=False, encoding='utf-8-sig')
print(f"파일 저장 완료: {output_file}")
```
