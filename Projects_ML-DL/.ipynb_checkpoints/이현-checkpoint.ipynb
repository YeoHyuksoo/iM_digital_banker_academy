### 주제 선정 배경
IM뱅크는 2024년 5월 16일, 지방은행에서 시중은행으로 거듭나 대구경북 지방을 넘어서 전국에 지점 개설을 할 수 있게 되었고 새로운 지점 개설을 통해 고객 확보를 하기 위한 노력을 하고 있다. 비록 2021~2023년 자료에 한정되어 있고, 이 시기 코로나 팬더믹이라는 특수한 사회, 경제 상황과 더불어 구체적인 고객 특성 분석을 하기에는 부족한 칼럼을 가지고 있지만, 대구/경북 지역과 이외 지역의 고객 특성 별 분류를 통해 IM뱅크의 새로운 지점 개설에 도움을 주고자 한다.

---

**프로젝트 목표**

- 대구·경북 지역과 기타 지역 고객의 금융 활동 특성을 비교 분석.
- 신규 지점 개설이 유망한 지역을 데이터 기반으로 선정.
- 금융 상품 이용 패턴 및 고객 성장률을 활용한 최적 입지 추천.

---

**활용 데이터**

- **고객 데이터**: 2021~2023년 IM뱅크 고객의 연령대, 성별, 자택 주소, 금융 상품 이용 내역 등.

# 카드, 고객 데이터 병합

```python
import os
import pandas as pd

# 병합할 파일 리스트 정의
files = [
    'iMBANK_CARD_DATA_2021(K-DigitalTraining).csv',
    'iMBANK_CARD_DATA_2022(K-DigitalTraining).csv',
    'iMBANK_CARD_DATA_2023(K-DigitalTraining).csv'
]

# 데이터를 저장할 리스트
dataframes = []

# 각 파일 읽기
for file in files:
    if os.path.exists(file):  # 파일 존재 여부 확인
        try:
            # euc-kr 인코딩으로 CSV 읽기
            df = pd.read_csv(file, encoding='euc-kr')
            dataframes.append(df)  # DataFrame 리스트에 추가
            print(f"{file} 읽기 성공")
        except Exception as e:
            print(f"{file} 읽기 실패: {e}")
    else:
        print(f"{file} 파일이 존재하지 않습니다.")

# 데이터 병합 (같은 칼럼을 가진 경우)
if dataframes:
    combined_df = pd.concat(dataframes, ignore_index=True)
    print("모든 데이터를 성공적으로 병합했습니다.")
    
    # 병합된 데이터를 CSV 파일로 저장
    combined_df.to_csv('iMBANK_CARD_DATA_TOTAL', index=False, encoding='euc-kr')
    print("병합된 데이터를 'iMBANK_CARD_DATA_TOTAL.csv'로 저장했습니다.")
else:
    print("병합할 데이터가 없습니다.")
```

```python
import pandas as pd

# 데이터 로드
file_path = 'IMBANK_CUSTOMER_DATA_TOTAL.csv'  # 파일 경로를 적절히 수정하세요.
df_customer_total = pd.read_csv(file_path, encoding='euc-kr')

target_regions = ['대구광역시', '경상북도']
daegu_gyeongbuk = df_customer_total[df_customer_total['자택_시도'].isin(target_regions)]
other_regions = df_customer_total[~df_customer_total['자택_시도'].isin(target_regions)]

print(f"대구+경북 고객 수: {len(daegu_gyeongbuk)}")
print(f"기타 지역 고객 수: {len(other_regions)}")

#%% [markdown]
# ## 3. 평균 계산 함수 정의
def calculate_means(df, region_name):
    """지역별 금융상품 평균 계산"""
    return {
        '지역': region_name,
        '수신_요구불예금': df['수신_요구불예금'].mean(),
        '수신_거치식예금': df['수신_거치식예금'].mean(),
        '수신_적립식예금': df['수신_적립식예금'].mean(),
        '수신_펀드': df['수신_펀드'].mean(),
        '수신_외화예금': df['수신_외화예금'].mean(),
        '대출금액': df['대출금액'].mean()
    }

# 대구+경북과 기타 지역 평균 계산
daegu_gyeongbuk_means = calculate_means(daegu_gyeongbuk, "대구+경북")
other_regions_means = calculate_means(other_regions, "기타 지역")

# 결과 저장
results = [daegu_gyeongbuk_means, other_regions_means]

#%% [markdown]
# ## 4. 결과 테이블 출력
result_df = pd.DataFrame(results).set_index('지역')
print("지역별 금융상품 평균 비교:")
display(result_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

# 데이터 로드
file_path = 'IMBANK_CUSTOMER_DATA_TOTAL.csv'  # 파일 경로를 적절히 수정하세요.
df_customer_total = pd.read_csv(file_path, encoding='euc-kr')

# 대구광역시와 경상북도 데이터 필터링
target_regions = ['대구광역시', '경상북도']
daegu_gyeongbuk = df_customer_total[df_customer_total['자택_시도'].isin(target_regions)]
other_regions = df_customer_total[~df_customer_total['자택_시도'].isin(target_regions)]

# 그룹화: 성별, 고객등급, 연령대별 가입자 수 계산
daegu_gyeongbuk_summary = daegu_gyeongbuk.groupby(['성별', '고객등급', '연령대']).size().reset_index(name='가입자수')
other_regions_summary = other_regions.groupby(['성별', '고객등급', '연령대']).size().reset_index(name='가입자수')

# 시각화 함수 정의
def plot_region_summary(summary_df, region_name):
    plt.figure(figsize=(12, 6))
    sns.barplot(
        data=summary_df,
        x='연령대',
        y='가입자수',
        hue='성별',
        ci=None
    )
    plt.title(f'{region_name} 성별/연령대별 가입자 분포')
    plt.xlabel('연령대')
    plt.ylabel('가입자 수')
    plt.legend(title='성별')
    plt.show()

    plt.figure(figsize=(12, 6))
    sns.barplot(
        data=summary_df,
        x='고객등급',
        y='가입자수',
        hue='성별',
        ci=None
    )
    plt.title(f'{region_name} 성별/고객등급별 가입자 분포')
    plt.xlabel('고객등급')
    plt.ylabel('가입자 수')
    plt.legend(title='성별')
    plt.show()

# 대구/경북 지역 시각화
plot_region_summary(daegu_gyeongbuk_summary, '대구/경북')

# 기타 지역 시각화
plot_region_summary(other_regions_summary, '기타 지역')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 설정
plt.rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False

# 기간 필터링
df_filtered = df_customer_total[
    (df_customer_total['기준년월'] >= 202101) & 
    (df_customer_total['기준년월'] <= 202312)
]

# 각 월별 금융상품 가입자 수 계산
# 하나라도 금융상품에 가입한 경우를 계산
df_filtered['금융상품가입여부'] = (
    (df_filtered['수신_요구불예금'] > 0) |
    (df_filtered['수신_거치식예금'] > 0) |
    (df_filtered['수신_적립식예금'] > 0) |
    (df_filtered['수신_펀드'] > 0) |
    (df_filtered['수신_외화예금'] > 0) |
    (df_filtered['대출금액'] > 0)
).astype(int)

# 월별 가입자 수 집계
monthly_subscribers = df_filtered.groupby('기준년월')['금융상품가입여부'].sum()

# 증감률 계산
monthly_change = monthly_subscribers.pct_change() * 100

# 그래프 생성
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))

# 월별 가입자 수 그래프
ax1.plot(range(len(monthly_subscribers)), monthly_subscribers.values, 
         marker='o', linewidth=2)
ax1.set_title('월별 금융상품 가입자 수')
ax1.set_xticks(range(len(monthly_subscribers)))
ax1.set_xticklabels(monthly_subscribers.index, rotation=45)
ax1.grid(True)

# 월별 증감률 그래프 (폭포수 차트)
colors = ['red' if x < 0 else 'green' for x in monthly_change.values]
ax2.bar(range(len(monthly_change)), monthly_change.values, color=colors)
ax2.set_title('월별 가입자 증감률 (%)')
ax2.set_xticks(range(len(monthly_change)))
ax2.set_xticklabels(monthly_change.index, rotation=45)
ax2.grid(True)
ax2.axhline(y=0, color='black', linestyle='-', linewidth=0.5)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 한글 폰트 설정
plt.rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False

# 데이터 필터링: 2021년 1월부터 2023년 12월까지
df_filtered = df_customer_total[
    (df_customer_total['기준년월'] >= 202101) & (df_customer_total['기준년월'] <= 202312)
]

# 기준년월을 문자열로 변환 (X축 정렬 문제 해결)
df_filtered['기준년월'] = df_filtered['기준년월'].astype(str)

# 금융상품 컬럼 리스트
columns = ['수신_요구불예금', '수신_거치식예금', '수신_적립식예금', '수신_펀드', '수신_외화예금', '대출금액']

# 그래프 생성 및 월별 변화율 계산
fig, axes = plt.subplots(len(columns), 1, figsize=(12, len(columns) * 4))

for i, col in enumerate(columns):
    # 월별 총 가입 금액 계산
    monthly_totals = df_filtered.groupby('기준년월')[col].sum()
    
    # 월별 변화율 계산
    monthly_change = monthly_totals.pct_change() * 100

    # 그래프 생성
    axes[i].plot(monthly_totals.index, monthly_totals.values, marker='o', linewidth=2, label='총 가입 금액')
    axes[i].set_title(f'{col} 월별 총 가입 금액 추이', fontsize=16)
    axes[i].set_xlabel('기준년월', fontsize=12)
    axes[i].set_ylabel('총 가입 금액 (원)', fontsize=12)
    axes[i].tick_params(axis='x', rotation=45)
    axes[i].grid(alpha=0.3)
    
plt.tight_layout()
plt.show()
```

   - 대출 상품에서는 기존 차주의 부담을 줄이는 대환대출과 같은 정책적 지원이 중요합니다.

### **대환대출이 대구은행 정책 및 금융상품 가입자 수에 미친 영향 분석**

#### **1. 대환대출의 배경**
- **코로나19 팬데믹**과 **금리 인상**으로 인해 고금리 대출을 이용하던 개인사업자와 소상공인의 금융 부담이 크게 증가했습니다.
- 정부는 이러한 상황을 완화하기 위해 **2023년 5월**부터 대환대출 인프라를 본격적으로 도입하여, 고금리 대출을 저금리 대출로 전환할 수 있는 기회를 제공했습니다[1][3][6].
- 대환대출은 금융회사 간 경쟁을 촉진하고, 소비자의 이자 부담을 줄이는 데 목적이 있었습니다.

---

#### **2. 대구은행의 대환대출 정책**
- 대구은행은 2022년 하반기부터 **"소상공인 저금리 대환대출"** 상품을 출시하여, 7% 이상의 고금리 사업자대출을 연 5.5%의 고정금리로 전환하는 서비스를 제공했습니다[4][5].
- 이를 통해 약 232억 원 규모의 고금리 대출을 저금리로 전환했으며, 고객의 금융비용 부담을 경감시켰습니다.
- 또한, 신용보증기금을 통해 보증료 감면 및 이자 환급 혜택을 제공하여 추가적인 비용 절감 효과를 창출했습니다[4].

---

#### **3. 대환대출이 금융상품 가입자 수에 미친 영향**
1. **2023년 가입자 수 증가**:
   - 상반기(2023년 5월 이후)부터 대환대출 인프라가 본격적으로 시행되면서, 기존 고금리 대출 이용 고객이 저금리 상품으로 이동하기 시작했습니다.
   - 대구은행의 지역 특화 마케팅과 함께, 소상공인 및 개인사업자를 대상으로 한 맞춤형 대환대출 상품이 가입자 수 증가에 기여했을 가능성이 큽니다[4][6].

2. **예금 상품 가입 증가**:
   - 금리가 안정화되면서 예금 상품의 매력이 증가했으며, 이는 대구은행의 기존 고객 유지와 신규 고객 유치에 긍정적 영향을 미쳤습니다.
   - 특히, 대환대출로 이자 부담이 줄어든 고객들이 여유 자금을 예금 상품으로 전환했을 가능성이 있습니다.

3. **대출 상품 가입 감소 방지**:
   - 일반적으로 금리 인상기에는 대출 상품 가입이 감소하지만, 저금리 대환대출은 기존 고금리 차주의 이동을 유도하며 대구은행의 대출 포트폴리오를 유지하는 데 기여했을 가능성이 있습니다[3][6].

#### **4. 국내외 경제/사회 상황과 연계**
1. **코로나19 팬데믹 이후 회복세**:
   - 코로나19로 인해 경제적 불확실성이 커졌던 시기에 비해, 2023년에는 경제 정상화와 소비 심리가 회복되며 금융상품 가입이 활발해졌습니다.
   - 특히, 정부의 민생 금융 지원 정책(저금리 대환 프로그램 등)이 이러한 회복세를 뒷받침했습니다[1][4].

2. **한국은행 기준금리 인상과 안정화**:
   - 한국은행은 2021년 8월부터 기준금리를 인상하기 시작해 2023년 초 3.5%까지 도달했습니다[3].
   - 금리가 안정화된 2023년 하반기에는 예·적금 상품의 매력이 증가했으며, 이는 금융상품 가입자 수 증가로 이어졌습니다.

3. **소상공인 지원 정책 강화**:
   - 정부는 소상공인을 대상으로 한 저금리 대환 프로그램과 같은 지원책을 통해 금융 비용 부담 완화를 추진했습니다.
   - 이러한 정책은 특히 중소기업 및 개인사업자가 많은 지역(예: 대구·경북)에서 긍정적인 영향을 미쳤습니다[4][6].

### 기존의 영업점이 위치해 있지 않은 영역을 바탕으로 코드 작성

기존 IM뱅크 지점의 자택_시도와 자택_시군구 조합 추출
- existing_regions = df_IM_address[['자택_시도', '자택_시군구']].drop_duplicates()

고객 데이터에서 기존 지역에 해당하지 않는 새로운 지역 필터링
- new_regions = df_customer_total_others[['자택_시도', '자택_시군구']].drop_duplicates()

```python
import pandas as pd
import matplotlib.pyplot as plt
# 한글 폰트 설정
plt.rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False
# 데이터 로드
df_IM_address = pd.read_csv('IM_address.csv')  # 기존 IM뱅크 지점 데이터
df_customer_total_others = pd.read_csv('df_customer_total_others.csv', encoding='euc-kr')  # 고객 데이터

# 기존 IM뱅크 지점의 자택_시도와 자택_시군구 조합 추출
existing_regions = df_IM_address[['자택_시도', '자택_시군구']].drop_duplicates()

# 고객 데이터에서 기존 지역에 해당하지 않는 새로운 지역 필터링
new_regions = df_customer_total_others[['자택_시도', '자택_시군구']].drop_duplicates()
new_regions = new_regions.merge(existing_regions, on=['자택_시도', '자택_시군구'], how='left', indicator=True)
new_regions = new_regions[new_regions['_merge'] == 'left_only'].drop(columns=['_merge'])

# 신규 가입자 수 계산 (새로운 지역에서만)
region_new_customers = (
    df_customer_total_others[df_customer_total_others[['자택_시도', '자택_시군구']].apply(tuple, axis=1).isin(new_regions.apply(tuple, axis=1))]
    .groupby(['자택_시도', '자택_시군구'])['고객ID']
    .nunique()
    .reset_index()
)
region_new_customers.rename(columns={'고객ID': '신규가입자수'}, inplace=True)

# 내림차순 정렬
region_new_customers_sorted = region_new_customers.sort_values(by='신규가입자수', ascending=False)

# 상위 10개 지역 선택
top_10_regions = region_new_customers_sorted.head(10)

# 결과 출력
print("상위 10개 새로운 지역 후보:")
print(top_10_regions)

# 시각화: X축과 Y축 교환, 그래프 옆에 숫자 추가
plt.figure(figsize=(10, 6))
bars = plt.barh(top_10_regions['자택_시도'] + " " + top_10_regions['자택_시군구'], top_10_regions['신규가입자수'], color='skyblue')
plt.title('상위 10개 새로운 지역 신규 가입자 수', fontsize=16)
plt.xlabel('신규 가입자 수', fontsize=12)
plt.ylabel('지역', fontsize=12)

# 그래프 옆에 숫자 추가
for bar in bars:
    plt.text(bar.get_width() + 50, bar.get_y() + bar.get_height()/2,
             f'{int(bar.get_width())}', va='center', fontsize=10)
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

# 머신러닝

### **가중치를 사용한 이유**

#### **1. 신규 가입자 수 계산**
- **신규 가입자 수**는 해당 지역에서의 금융 활동 잠재력을 측정하는 중요한 지표입니다.
- 신규 고객이 많을수록 해당 지역은 금융상품에 대한 수요가 높고, 대구은행이 신규 지점을 설립하거나 마케팅을 강화할 경우 성공 가능성이 높습니다.
- 따라서, **고객 성장률(20%)**에 가중치를 부여하여 신규 고객 유입의 중요성을 반영했습니다.

---

#### **2. 각 지역의 금융상품 총합계**
- 금융상품 총합계는 해당 지역의 전반적인 금융 활동 규모를 나타냅니다.
- 예금, 대출, 펀드 등 다양한 금융상품의 총합계는 지역 경제력과 금융시장 규모를 평가하는 데 중요한 역할을 합니다.
- 대출금액은 대구경북 지역에서는 요구불예금 비중이 가장 높았던 것과 달리, 이외의 지역에서는 대출금액이 가장 큰 비율을 보이고 있으므로 따로 가중치를 부여하였습니다.
- 특히, **총금액(35%)**과 **대출금액(25%)**에 높은 가중치를 부여하여, 대구은행이 해당 지역에서 기대할 수 있는 금융 자산 규모를 강조했습니다.

---

#### **3. 평균 금융 활동 증가율**
- 금융 활동 증가율은 해당 지역의 경제 성장 가능성과 금융상품 수요 증가를 예측하는 데 중요한 지표입니다.
- 월별 증감률을 기반으로 계산된 평균 증가율은 지역 경제의 역동성을 평가하며, 성장 가능성이 높은 지역을 식별하는 데 유용합니다.
- 따라서, **총금액 성장률(10%)**과 **대출금액 성장률(10%)**에 가중치를 부여하여, 지속적으로 성장하는 지역의 중요성을 반영했습니다.

---

### **가중치 구성의 의미** (가중치는 다른 요소를 고려하여 수정해서, 신규개설 지점의 목표와 특색에 맞게 수정하여 사용)
| 항목                     | 가중치  | 이유                                                                                     |
|--------------------------|---------|------------------------------------------------------------------------------------------|
| **총금액 (Total)**       | 35%     | 해당 지역의 전체 금융 자산 규모를 가장 중요하게 평가.                                     |
| **대출금액 (Loan)**       | 25%     | 대출 상품은 은행 수익성에 중요한 영향을 미치며, 대출 수요가 높은 지역을 우선적으로 고려.   |
| **고객 성장률 (Customer)**| 20%     | 신규 고객 유입은 은행의 시장 점유율 확대와 직결되므로 중요하게 평가.                      |
| **총금액 성장률**         | 10%     | 지역 경제 및 금융 자산의 지속적인 성장 가능성을 나타냄.                                  |
| **대출금액 성장률**       | 10%     | 대출 상품 수요가 지속적으로 증가하는 지역은 은행의 장기적인 수익성에 긍정적인 영향을 미침.|

---

### **`feature_engineering` 함수를 사용한 이유**

1. **시간 단위 분석 가능**:
   - `기준년월`, `월`, `분기`와 같은 시간 관련 정보를 생성하여 월별/분기별 금융 활동 패턴 분석 및 시계열 기반 예측 가능.

2. **총 금융 활동 규모 계산**:
   - 여러 금융상품의 금액을 합산한 `상품가입금액합계`를 통해 고객 또는 지역의 전체 금융 활동을 평가.

3. **신규 고객 식별**:
   - 신규 고객 여부(`신규고객`)를 통해 은행 성장성과 마케팅 효과를 평가하고, 기존 고객과 차별화된 전략 수립 가능.

4. **데이터 전처리 자동화**:
   - 동일한 전처리 작업을 여러 데이터셋에 일관되게 적용할 수 있어 효율성과 재사용성 향상.

5. **모델 학습 준비**:
   - 머신러닝 모델에서 사용할 주요 피처를 생성하여 학습 및 예측 성능 향상.

### **`Min-Max` 스케일러를 사용한 이유**

1. **데이터의 스케일 차이 조정**:
   - Min-Max 스케일링을 통해 모든 피처를 동일한 범위(0~1)로 변환함으로써, 각 피처가 동일한 중요도로 모델에 반영되도록 합니다.

2. **머신러닝 모델의 성능 향상**:
   - 많은 머신러닝 알고리즘(예: KMeans, XGBoost 등)은 입력 데이터의 스케일에 민감합니다.
   - KMeans: 거리 기반 알고리즘으로, 피처 간 거리 계산에 영향을 미치는 스케일 차이를 제거합니다.
   - XGBoost: 가중치 기반 알고리즘으로, 피처의 스케일 차이가 클 경우 학습이 불안정해질 수 있으므로 이를 방지합니다.

3. **데이터 해석 용이성**:
   - Min-Max 스케일링은 결과값을 0~1 사이로 변환하므로, 각 지표를 상대적인 중요도로 해석하기 쉽습니다.

4. **이상치 영향 감소**:
   - 특정 지역에서 총금액이 매우 크더라도, 이를 최대값으로 정규화하면 다른 지역과 비교 가능한 상대적 값을 얻을 수 있습니다.


### **머신러닝 파이프라인에서 사용된 기술과 이유**

1. **결측치 처리(SimpleImputer)**:
   - 머신러닝 모델은 결측치(NaN)를 처리하지 못하므로, 이를 평균값으로 대체하여 학습 가능한 데이터로 변환합니다.
   - 결측치가 있는 데이터를 처리하여 모델 학습 및 예측 과정에서 오류가 발생하지 않도록 보장.

2. **데이터 정규화(StandardScaler)**:
   - 머신러닝 알고리즘(특히 거리 기반 알고리즘인 KMeans)은 피처 간 스케일 차이에 민감합니다.
   - 모든 피처를 평균 0, 표준편차 1로 변환하여 동일한 스케일로 조정합니다.

3. **크러스터링(Kmeans)**:
   - 비지도 학습 알고리즘인 KMeans를 사용해 지역 데이터를 유사한 특성을 가진 그룹(클러스터)으로 나눕니다.

4. **데이터 분할('train_test_split)**:
   - 데이터를 학습용(train)과 테스트용(test)으로 분리하여 모델의 일반화 성능(새로운 데이터에 대한 예측 정확도)을 평가합니다.
   - 과적합(overfitting)을 방지하고 모델의 성능을 공정하게 평가.

5. **회귀 모델 학습(XGBRegressor)**:
   - XGBoost는 부스팅 알고리즘 기반의 강력한 회귀 모델로, 높은 예측 성능과 효율성을 제공합니다.
   - 주어진 입력 변수(X_train)와 목표 변수(y_train)를 기반으로 회귀 모델을 학습하여 각 지역의 점수를 예측.

4. **모델평가(mean_squared_error,RMSE 계산)**:
   - RMSE(Root Mean Squared Error)는 회귀 모델의 성능을 평가하는 지표로, 예측값과 실제값 간의 평균적인 차이를 나타냅니다.
   - RMSE 값을 통해 모델이 얼마나 정확하게 점수를 예측했는지 평가.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_squared_error
import xgboost as xgb

# 한글 폰트 설정
plt.rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False

# 1. 데이터 로드
df_IM_address = pd.read_csv('IM_address.csv')  # 기존 IM뱅크 지점 데이터
df_customer_total_others = pd.read_csv('df_customer_total_others.csv', encoding='euc-kr')  # 고객 데이터

# 2. 기존 IM뱅크 지점의 자택_시도와 자택_시군구 추출
existing_regions = df_IM_address[['자택_시도', '자택_시군구']].drop_duplicates()

# 3. 고객 데이터에서 기존 지역에 해당하지 않는 새로운 지역 필터링
new_regions = df_customer_total_others[['자택_시도', '자택_시군구']].drop_duplicates()
new_regions = new_regions.merge(existing_regions, on=['자택_시도', '자택_시군구'], how='left', indicator=True)
new_regions = new_regions[new_regions['_merge'] == 'left_only'].drop(columns=['_merge'])

# 4. 신규지역 고객 데이터 선택 및 '지역' 컬럼 생성 (예: "서울 강남구")
filter_condition = df_customer_total_others[['자택_시도', '자택_시군구']].apply(tuple, axis=1).isin(
    new_regions.apply(tuple, axis=1)
)
filtered_data = df_customer_total_others[filter_condition].copy()
filtered_data.loc[:, '지역'] = filtered_data['자택_시도'] + ' ' + filtered_data['자택_시군구']

# 5. 피처 엔지니어링: 기준년월 변환, 월/분기 추출, 예금 항목 합산, 신규 고객 여부 산출
def feature_engineering(df):
    df = df.copy()
    # 기준년월 문자열('202101')을 datetime형('2021-01-01')으로 변환
    df['기준년월'] = pd.to_datetime(df['기준년월'], format='%Y%m')
    df['월'] = df['기준년월'].dt.month
    df['분기'] = df['기준년월'].dt.quarter
    # 예금 관련 컬럼 합산
    deposit_cols = ['수신_요구불예금', '수신_거치식예금', '수신_적립식예금', '수신_펀드', '수신_외화예금']
    df['상품가입금액합계'] = df[deposit_cols].sum(axis=1)
    # 고객별 첫 가입월과 동일하면 신규 고객 (1), 아니면 0
    df['신규고객'] = (df.groupby('고객ID')['기준년월'].transform('min') == df['기준년월']).astype(int)
    return df

processed_data = feature_engineering(filtered_data)

# 6. 2021~2023년 데이터 선택 및 '총금액' 계산 (상품가입금액합계 + 대출금액)
time_filtered = processed_data[(processed_data['기준년월'].dt.year >= 2021) &
                               (processed_data['기준년월'].dt.year <= 2023)].copy()
time_filtered.loc[:, '총금액'] = time_filtered['상품가입금액합계'] + time_filtered['대출금액']

# 7. 지역별 전체 총 금융활동 및 대출금액 합계 산출
total_amt = time_filtered.groupby('지역')['총금액'].sum().reset_index(name='총금액_total')
loan_amt = time_filtered.groupby('지역')['대출금액'].sum().reset_index(name='대출금액_total')

# 8. 월별 고객 수 집계 및 성장률 계산
monthly_customers = time_filtered.groupby(['지역', '기준년월'])['고객ID'].nunique().reset_index(name='고객수')
def calc_growth_rate(series):
    growth = series.pct_change()
    return growth.mean() * 100  # 평균 증감률 (%)
    
customer_growth = monthly_customers.groupby('지역')['고객수'].apply(
    lambda s: calc_growth_rate(s.sort_values())
).reset_index(name='고객_성장률')

# 9. 월별 총금액 집계 및 성장률 계산
monthly_total = time_filtered.groupby(['지역', '기준년월'])['총금액'].sum().reset_index(name='총금액_monthly')
total_growth = monthly_total.groupby('지역')['총금액_monthly'].apply(
    lambda s: calc_growth_rate(s.sort_values())
).reset_index(name='총금액_성장률')

# 10. 월별 대출금액 집계 및 성장률 계산
monthly_loan = time_filtered.groupby(['지역', '기준년월'])['대출금액'].sum().reset_index(name='대출금액_monthly')
loan_growth = monthly_loan.groupby('지역')['대출금액_monthly'].apply(
    lambda s: calc_growth_rate(s.sort_values())
).reset_index(name='대출금액_성장률')

# 11. 지역별 통계 병합
region_stats = total_amt.merge(loan_amt, on='지역', how='left')  \
                        .merge(customer_growth, on='지역', how='left')  \
                        .merge(total_growth, on='지역', how='left')  \
                        .merge(loan_growth, on='지역', how='left')

# 12. 각 지표를 min–max 정규화 (0 ~ 1)
def min_max_normalize(series):
    if series.max() != series.min():
        return (series - series.min()) / (series.max() - series.min())
    else:
        return series
region_stats['총금액_total_norm'] = min_max_normalize(region_stats['총금액_total'])
region_stats['대출금액_total_norm'] = min_max_normalize(region_stats['대출금액_total'])
region_stats['고객_성장률_norm']      = min_max_normalize(region_stats['고객_성장률'])
region_stats['총금액_성장률_norm']      = min_max_normalize(region_stats['총금액_성장률'])
region_stats['대출금액_성장률_norm']    = min_max_normalize(region_stats['대출금액_성장률'])

# 13. 최종 점수 계산 (가중치 적용)
region_stats['점수'] = (
    region_stats['총금액_total_norm'] * 0.35 +
    region_stats['대출금액_total_norm'] * 0.25 +
    region_stats['고객_성장률_norm'] * 0.20 +
    region_stats['총금액_성장률_norm'] * 0.10 +
    region_stats['대출금액_성장률_norm'] * 0.10
) * 100

# ★ 라벨에 결측치가 있으면 XGBoost 오류가 발생하므로, 결측치를 채우거나 제거합니다.
# 여기서는 결측치를 0으로 채웁니다.
region_stats['점수'] = region_stats['점수'].fillna(0)

# 14. 순위 생성 및 정렬
region_stats_sorted = (
    region_stats.sort_values('점수', ascending=False)
    .reset_index(drop=True)
    .assign(순위=lambda x: x.index + 1)
    [['순위', '지역', '총금액_total', '대출금액_total',
      '고객_성장률', '총금액_성장률', '대출금액_성장률', '점수']]
)

# ★ 표로 출력 (Jupyter 환경에서 보기 좋게 출력)
from IPython.display import display

print("최적 입지 추천 지역 (상위 5개):")
display(region_stats_sorted.head(5))

# 15. 머신러닝 파이프라인
features = ['총금액_total_norm', '대출금액_total_norm', 
            '고객_성장률_norm', '총금액_성장률_norm', '대출금액_성장률_norm']
X = region_stats[features].values
y = region_stats['점수'].values

imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)

kmeans = KMeans(n_clusters=3, random_state=42)
region_stats['클러스터'] = kmeans.fit_predict(X_scaled)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
xgb_model = xgb.XGBRegressor(objective='reg:squarederror', random_state=42)
xgb_model.fit(X_train, y_train)
y_pred = xgb_model.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"\nXGBoost RMSE: {rmse:.2f}")

# 16. 상위 5개 지역 점수 비교 시각화
plt.figure(figsize=(12,8))
sns.barplot(
    data=region_stats_sorted.head(5),
    x='지역',
    y='점수',
    palette='viridis'
)
plt.title('상위 5개 지역 점수 비교')
plt.xlabel('')
plt.ylabel('점수 (100점 만점)')
plt.xticks(rotation=45)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

## 은행 밀집도를 통해 신규 지점 개설지점을 설정하는 이유

- **경쟁 분석**  
  특정 지역에 이미 많은 은행이 밀집되어 있다면, 경쟁이 치열할 가능성이 높습니다. 이런 경우 신규 지점을 개설하면 기존 은행들과 고객을 두고 경쟁해야 하므로 수익성이 낮아질 수 있습니다. 반대로 은행 밀집도가 낮은 지역은 경쟁이 적어 고객 확보에 유리할 수 있습니다.

- **고객 접근성 향상**  
  은행 밀집도가 낮은 지역은 금융 서비스에 대한 접근성이 부족할 가능성이 큽니다. 이런 지역에 신규 지점을 개설하면 금융 소외 계층에게 서비스를 제공하며, 지역 사회와의 관계를 강화할 수 있습니다.

- **시장 잠재력 발굴**  
  밀집도가 낮은 지역은 아직 개발되지 않은 시장일 가능성이 있습니다. 이런 곳에 선제적으로 진출하면 초기 시장 점유율을 확보하고 장기적으로 안정적인 고객 기반을 구축할 수 있습니다.

- **운영 비용 최적화**  
  은행 밀집도가 높은 지역에서는 임대료, 인건비 등 운영 비용이 상승할 가능성이 높습니다. 반면, 밀집도가 낮은 지역에서는 이러한 비용이 상대적으로 낮아 운영 효율성을 높일 수 있습니다.



# 신규 은행 지점 후보 위치 분석 및 시각화

이 코드는 **Google Places Nearby Search API**를 사용하여 경기도 광주시 중심 주변의 은행 정보를 조회합니다. 이후, 조회한 은행 정보로 DBSCAN 클러스터링을 수행해 은행들이 밀집해 있는 군집을 파악하고, Folium을 사용해 다음과 같이 지도를 시각화합니다:

- **은행 정보**: 각 은행 위치에 파란색 마커 표시
- **히트맵**: 은행 밀집도를 시각적으로 표시
- **최적 신규 지점 후보**: 가장 많은 은행이 모인 클러스터의 중심을 빨간색 스타 아이콘으로 표시

아래 Markdown 셀에 붙여넣으면 코드와 설명이 함께 보입니다.

### 코드 설명

1. **API 호출 및 데이터 로드**  
   - Google Places Nearby Search API를 이용해 지정한 중심 좌표(광주시)와 반경 내 은행 정보를 조회합니다.
   - 조회 결과에서 각 은행의 이름, 위도, 경도를 추출해 DataFrame으로 구성합니다.

2. **DBSCAN 클러스터링 수행**  
   - `DBSCAN` 알고리즘을 사용하여 은행들 간의 밀집 군집을 찾습니다.
   - `eps=0.0045`로 설정하여 약 500m 내외에 있는 은행을 하나의 군집으로 분류합니다.
   - 노이즈로 분류된 값(-1)을 제외하고, 가장 많은 은행이 모인 클러스터를 선정합니다.
   - 해당 클러스터의 중심위치를 계산하여 최적 신규 지점 후보로 지정합니다.

3. **Folium 지도 생성 및 시각화**  
   - **파란색 마커**: 각 은행의 위치를 표시하여 개별 은행 정보를 시각화합니다.
   - **히트맵**: 은행들의 밀집도를 색상으로 표현해 시각화를 강화합니다.
   - **빨간색 마커**: 클러스터 중심(최적 신규 지점 후보 위치)을 표시합니다.

4. **Jupyter Notebook 내 지도 표시**  
   - `display(m)`를 사용해 Jupyter Notebook 셀에서 Folium 지도를 바로 확인할 수 있습니다.

```python
import requests
import pandas as pd
import folium
from folium.plugins import HeatMap
from sklearn.cluster import DBSCAN
from IPython.display import display

# 기본 설정
API_KEY = "AIzaSyC29v05rhXh6twqVWzg0r-h5sW-Tk26bJc"
center_lat, center_lng = 37.4095, 127.2556  # 경기도 광주시 중심 좌표
radius = 10000  # 검색 반경 (미터)

# Google Places Nearby Search API 호출 URL 구성
url = (
    f"https://maps.googleapis.com/maps/api/place/nearbysearch/json?"
    f"location={center_lat},{center_lng}&radius={radius}&type=bank&key={API_KEY}"
)

# API 호출 및 결과 확인
response = requests.get(url)
data = response.json()

if data.get("status") != "OK":
    print("데이터를 가져오는 중 오류 발생:", data.get("status"))
else:
    banks_list = []
    for bank in data['results']:
        name = bank.get('name')
        lat = bank['geometry']['location']['lat']
        lng = bank['geometry']['location']['lng']
        banks_list.append({'name': name, 'lat': lat, 'lng': lng})
    
    # DataFrame 구성
    df = pd.DataFrame(banks_list)
    
    # DBSCAN 클러스터링: eps 값은 약 0.0045 (약 500m 내외), min_samples=2
    coords = df[['lat', 'lng']].values
    clustering = DBSCAN(eps=0.0045, min_samples=2).fit(coords)
    df['cluster'] = clustering.labels_
    
    # 노이즈(-1) 제외 후, 가장 많은 은행이 모인 클러스터 파악
    valid_clusters = df[df['cluster'] != -1]['cluster']
    if not valid_clusters.empty:
        cluster_counts = valid_clusters.value_counts()
        best_cluster = cluster_counts.idxmax()
        cluster_points = df[df['cluster'] == best_cluster]
        centroid_lat = cluster_points['lat'].mean()
        centroid_lng = cluster_points['lng'].mean()
    else:
        centroid_lat, centroid_lng = None, None
    
    # Folium 지도 생성 (경기도 광주시 중심)
    m = folium.Map(location=[center_lat, center_lng], zoom_start=14)
    
    # 은행 정보에 파란색 마커 추가
    for idx, row in df.iterrows():
        folium.Marker(
            location=[row['lat'], row['lng']],
            popup=row['name'],
            icon=folium.Icon(color='blue', icon='info-sign')
        ).add_to(m)
    
    # 히트맵 추가 (은행 밀집도 시각화)
    heat_data = df[['lat', 'lng']].values.tolist()
    HeatMap(heat_data, radius=15).add_to(m)
    
    # 밀집된 클러스터 중심(최적 신규 지점 후보) 빨간색 아이콘으로 표시
    if centroid_lat and centroid_lng:
        folium.Marker(
            location=[centroid_lat, centroid_lng],
            popup="최적의 신규 지점 후보",
            icon=folium.Icon(color='red', icon='star')
        ).add_to(m)
    
    # Jupyter Notebook에서 지도 표시
    display(m)
```
