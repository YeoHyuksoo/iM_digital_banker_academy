```python
import pandas as pd
import numpy as np

# 원본 파일 읽기
df = pd.read_csv('unique_customer_ids.csv')

# 전체 행 수 확인
total_rows = len(df)
# 각 파일당 행 수 계산 (올림)
rows_per_file = int(np.ceil(total_rows / 10))

# 10개의 파일로 나누어 저장
for i in range(10):
    # 시작 인덱스와 끝 인덱스 계산
    start_idx = i * rows_per_file
    end_idx = min((i + 1) * rows_per_file, total_rows)
    
    # 부분 데이터프레임 추출
    subset_df = df.iloc[start_idx:end_idx]
    
    # 파일 저장 (파일명: unique_customer_ids_part1.csv, unique_customer_ids_part2.csv, ...)
    subset_df.to_csv(f'unique_customer_ids_part{i+1}.csv', index=False)
    
    # 저장된 파일의 행 수 출력
    print(f'Part {i+1}: {len(subset_df)} rows saved')
```

```python
# 최근 3개월 데이터 필터링 (2023년 10월, 11월, 12월)
recent_months = pd.to_datetime(['202310', '202311', '202312'], format='%Y%m')
df_recent = df[df['거래년월'].isin(recent_months)]

# 최근 3개월 동안 단 한 번도 카드 실적이 30만 원 이상인 적이 없는 고객 제거
monthly_performance_recent = df_recent.groupby(['고객ID', '거래년월'])['승인금액'].sum().reset_index()
valid_recent_customers = monthly_performance_recent[monthly_performance_recent['승인금액'] >= 300000]['고객ID'].unique()

# 최근 3개월 동안 조건을 만족하지 못한 고객 제외
df_filtered = df[df['고객ID'].isin(valid_recent_customers)]

# 각 고객의 가장 오래된 결제 월과 최신 결제 월 계산
customer_min_max_months = df_filtered.groupby('고객ID')['거래년월'].agg(['min', 'max']).reset_index()

# 각 고객의 전체 기간 동안 월별 카드 실적 계산
monthly_performance = df_filtered.groupby(['고객ID', '거래년월'])['승인금액'].sum().reset_index()

# 카드 실적이 30만 원 이상인 월 계산
monthly_performance['30만_이상'] = monthly_performance['승인금액'] >= 300000

# 각 고객의 전체 기간 중 카드 실적이 30만 원 이상인 월 비율 계산
def calculate_valid_month_ratio(row):
    customer_id = row['고객ID']
    min_month = row['min']
    max_month = row['max']
    
    # 해당 고객의 전체 기간 데이터 필터링
    customer_data = monthly_performance[(monthly_performance['고객ID'] == customer_id) & 
                                        (monthly_performance['거래년월'] >= min_month) & 
                                        (monthly_performance['거래년월'] <= max_month)]
    
    # 비율 계산
    total_months = len(customer_data)
    valid_months = customer_data['30만_이상'].sum()
    return valid_months / total_months if total_months > 0 else 0

customer_min_max_months['valid_ratio'] = customer_min_max_months.apply(calculate_valid_month_ratio, axis=1)

# 유효 비율이 30% 이상인 고객만 필터링
valid_customers_final = customer_min_max_months[customer_min_max_months['valid_ratio'] >= 0.3]['고객ID']

# 최종 결과 데이터프레임 생성
final_df = pd.DataFrame({'고객ID': valid_customers_final})

# 결과를 CSV 파일로 저장
final_df.to_csv('filtered_customer_ids.csv', index=False)

print("조건을 만족하는 고객ID가 filtered_customer_ids.csv에 저장되었습니다.")
```

```python
import pandas as pd

# 데이터 로드
file_path = "card_data_2023_sample1_utf8.csv"
df = pd.read_csv(file_path)

# 날짜 데이터 변환 (거래년월을 기준으로 판단)
df['거래년월'] = pd.to_datetime(df['거래년월'], format='%Y%m')
latest_month = df['거래년월'].max()

# 고객별 혜택 계산을 위한 기간 설정
periods = {
    'A카드_최근1개월혜택': 1,
    'A카드_최근3개월혜택': 3,
    'A카드_최근6개월혜택': 6,
    'A카드_최근12개월혜택': 12
}

# 혜택 조건 설정
discount_categories = {
    '이동통신 5% 할인': {'업종': ['이동통신요금'], '최소금액': 50000, '할인율': 0.05},
    '커피·생활용품·편의점 5% 할인': {
        '업종': ['기타음료식품', '화장품', '문구용품', '침구·수예점', '주방용구', '주방용식기', '연 쇄 점', '편 의 점'],
        '최소금액': 10000, '할인율': 0.05
    },
    '영화 5% 할인': {'업종': ['영화관'], '최소금액': 20000, '할인율': 0.05}
}

# 연회비 설정
annual_fee = 5000

# 결과를 저장할 딕셔너리 초기화
result_dict = {'고객ID': []}
for col_name in periods.keys():
    result_dict[col_name] = []

# 고객별 혜택 계산
for customer_id, customer_df in df.groupby('고객ID'):
    result_dict['고객ID'].append(customer_id)
    
    for col_name, months in periods.items():
        start_date = latest_month - pd.DateOffset(months=months)
        df_period = customer_df[customer_df['거래년월'] > start_date]
        
        monthly_benefit = []
        
        for month, month_group in df_period.groupby(df_period['거래년월']):
            total_benefit = 0
            total_performance = 0
            
            for _, row in month_group.iterrows():
                avg_amount = row['승인금액'] / row['승인건수'] if row['승인건수'] > 0 else 0
                applied = False
                
                for discount_name, condition in discount_categories.items():
                    if row['가맹점업종명'] in condition['업종'] and avg_amount >= condition['최소금액']:
                        total_benefit += row['승인금액'] * condition['할인율']
                        applied = True
                
                if not applied:
                    total_performance += row['승인금액']
            
            # 실적 기준 할인 한도 적용
            if total_performance >= 600000:
                total_benefit = min(total_benefit, 10000)
            elif total_performance >= 300000:
                total_benefit = min(total_benefit, 5000)
            else:
                total_benefit = 0
            
            monthly_benefit.append(total_benefit)
        
        # 기간 내 월별 혜택 합산
        total_benefit = sum(monthly_benefit)
        
        # 연회비 적용
        adjusted_benefit = total_benefit - (annual_fee / 12 * months)
        result_dict[col_name].append(adjusted_benefit)

# 결과 데이터프레임 생성
result = pd.DataFrame(result_dict)

# 결과 저장
output_file = "고객혜택금액_A카드.csv"
result.to_csv(output_file, index=False, encoding='utf-8-sig')
print(f"파일 저장 완료: {output_file}")
```

```python
import pandas as pd 

df = pd.read_csv('card_data_2023_sample1_utf8.csv', encoding='utf-8')

# 1. 업종 분류 함수 수정 (실적 제외 업종 추가) 
def classify_category(business_type):
    excluded_categories = ["PG상품권", "전자상거래상품권", "전자상거래상품권전문판매", "상품권전문판매", "공공요금대행서비스/소득공제대상", "공공요금대행서비스/소득공제비대상", "생명보험", "손해보험", "기타보험", "상 품 권", "조세서비스"]
    if business_type in excluded_categories:
        return '실적제외'
    elif business_type in ['인터넷Mall', '인터넷종합Mall', 'PG상품권']:
        return '온라인쇼핑'
    elif business_type == '기타음료식품':
        return '커피'
    elif business_type == '편 의 점':
        return '편의점'
    elif business_type == '이동통신요금':
        return '이동통신'
    else:
        return None

# 2. 할인한도 계산 함수 (동일)
def get_discount_limits(total_usage):
    if total_usage >= 600000:
        return {
            '온라인쇼핑': 10000,
            '커피': 8000,
            '편의점': 6000,
            '이동통신': 6000,
            '총한도': None  # 통합 한도 없음
        }
    elif total_usage >= 300000:
        return {
            '온라인쇼핑': 5000,
            '커피': 4000,
            '편의점': 3000,
            '이동통신': 3000,
            '총한도': None  # 통합 한도 없음
        }
    else:
        return None  # 실적이 30만 원 미만이면 혜택 없음

# 3. 혜택 계산 함수 (수정: 실적 제외 업종 반영 및 이동통신요금 1회 제한 적용)
def calculate_benefits(df, months):
    # 카테고리 분류
    df['카테고리'] = df['가맹점업종명'].apply(classify_category)
    
    # 기준월 설정 (202312)
    target_month = 202312
    
    benefits_list = []
    
    for customer_id in df['고객ID'].unique():
        customer_data = df[df['고객ID'] == customer_id]
        total_benefit = 0
        
        # 각 월별로 혜택 계산
        for i in range(months):
            year = target_month // 100
            month = target_month % 100 - i
            if month <= 0:
                month += 12
                year -= 1
            current_month = year * 100 + month
            
            # 해당 월의 데이터
            month_data = customer_data[customer_data['거래년월'] == current_month]
            
            # 실적 제외 업종 제거하고 실적 계산
            total_usage = month_data[~month_data['카테고리'].isin(['실적제외', '온라인쇼핑', '커피', '편의점', '이동통신'])]['승인금액'].sum()
            
            # 할인한도 계산 (30만 원 미만이면 혜택 없음)
            limits = get_discount_limits(total_usage)
            if limits is None:
                continue
            
            # 카테고리별 할인금액 계산
            month_discount = 0
            for category in ['온라인쇼핑', '커피', '편의점']:
                category_usage = month_data[month_data['카테고리'] == category]['승인금액'].sum()
                discount = min(category_usage * 0.1, limits[category])
                month_discount += discount
            
            # 이동통신요금 혜택 (월 1회 제한 적용)
            telecom_data = month_data[month_data['카테고리'] == '이동통신']
            if not telecom_data.empty:
                telecom_usage = telecom_data.sort_values(by='승인금액', ascending=False).iloc[0]['승인금액']
                telecom_discount = min(telecom_usage * 0.1, limits['이동통신'])
                month_discount += telecom_discount
            
            total_benefit += month_discount
        
        # 연회비 차감
        annual_fee = 10000 * (months/12)
        net_benefit = total_benefit - annual_fee
        
        benefits_list.append({
            '고객ID': customer_id,
            f'{months}개월_혜택': net_benefit
        })
    
    return pd.DataFrame(benefits_list)

# 4. 각 기간별 혜택 계산 및 결과 병합
result_1 = calculate_benefits(df, 1)
result_3 = calculate_benefits(df, 3)
result_6 = calculate_benefits(df, 6)
result_12 = calculate_benefits(df, 12)

# 결과 병합
final_result = result_1.merge(result_3, on='고객ID')\
    .merge(result_6, on='고객ID')\
    .merge(result_12, on='고객ID')
```

```python
import pandas as pd

# 데이터 로드
df = pd.read_csv('card_data_2023_sample1_utf8.csv', encoding='utf-8')

# 1. 업종 분류 함수
def classify_category_untact(business_type):
    if business_type == '인터넷P/G':
        return '간편결제'
    else:
        return None

# 2. 실적별 할인한도 계산 함수
def get_discount_limits_untact(total_usage):
    if total_usage >= 600000:
        return {'간편결제': 15000}
    elif total_usage >= 300000:
        return {'간편결제': 10000}
    else:
        return {'간편결제': 0}

# 3. UntacT카드 혜택 계산 함수
def calculate_benefits_untact(df, months):
    target_month = 202312
    benefits_list = []
    
    for customer_id in df['고객ID'].unique():
        customer_data = df[df['고객ID'] == customer_id]
        total_benefit = 0
        
        for i in range(months):
            year = target_month // 100
            month = target_month % 100 - i
            if month <= 0:
                month += 12
                year -= 1
            current_month = year * 100 + month
            
            month_data = customer_data[customer_data['거래년월'] == current_month]
            
            # 실적 계산 (실적 제외 업종 제거)
            excluded_categories = ["PG상품권", "전자상거래상품권", "전자상거래상품권전문판매", "상품권전문판매", 
                                   "공공요금대행서비스/소득공제대상", "공공요금대행서비스/소득공제비대상", "생명보험", 
                                   "손해보험", "기타보험", "상 품 권", "조세서비스"]
            total_usage = month_data[~month_data['가맹점업종명'].isin(excluded_categories)]['승인금액'].sum()
            
            limits = get_discount_limits_untact(total_usage)
            
            month_discount = 0
            
            # 간편결제 할인 계산 (건당 1만 원 이상, 10% 할인)
            payment_data = month_data[month_data['카테고리'] == '간편결제']
            for _, row in payment_data.iterrows():
                if row['승인건수'] > 0:  # 안전한 계산을 위해 체크 추가
                    amount_per_transaction = row['승인금액'] / row['승인건수']
                    if amount_per_transaction >= 10000:
                        discount = row['승인금액'] * 0.1
                        month_discount += discount
            
            # 월 할인한도 적용
            month_discount = min(month_discount, limits['간편결제'])
            total_benefit += month_discount
        
        # 연회비 차감
        annual_fee = 10000 * (months/12)
        net_benefit = total_benefit - annual_fee
        
        benefits_list.append({
            '고객ID': customer_id,
            f'UntacT카드_{months}개월혜택': net_benefit
        })
    
    return pd.DataFrame(benefits_list)

# ✅ 1. 업종 분류를 한 번만 실행
df['카테고리'] = df['가맹점업종명'].apply(classify_category_untact)

# ✅ 2. 각 기간별 혜택 데이터프레임 생성
periods = [1, 3, 6, 12]
results = []

for period in periods:
    benefit_df = calculate_benefits_untact(df.copy(), period)
    results.append(benefit_df)

# ✅ 3. 모든 기간별 혜택을 고객ID 기준으로 병합 (concat 사용)
final_result = pd.concat(results, axis=1)
final_result = final_result.loc[:, ~final_result.columns.duplicated()]

# ✅ 4. 기존 데이터와 병합
existing_result = pd.read_csv('final_card_benefits.csv')
final_result = existing_result.merge(final_result, on='고객ID', how='left')

# ✅ 5. CSV 저장 (utf-8-sig 인코딩 적용)
final_result.to_csv('final_card_benefits_ver2.csv', index=False, encoding='utf-8-sig')

print("파일 저장 완료: final_card_benefits_ver2.csv")
```

```python
import pandas as pd

# 데이터 로드
df = pd.read_csv('card_data_2023_sample1_utf8.csv')

# 1. 업종 분류 함수
def classify_category_living(business_type):
    if business_type in ['공공요금대행서비스/소득공제대상', '공공요금대행서비스/소득공제비대상', '이동통신요금']:
        return '생활요금'
    elif business_type in ['인터넷Mall', '인터넷종합Mall', 'PG상품권', '화장품', '문구용품', 
                           '침구·수예점', '주방용구', '주방용식기']:
        return '생활쇼핑'
    elif business_type in ['대형할인점', '농협하나로클럽', '농,축협직영매장']:
        return '대형마트'
    else:
        return None

# 2. 실적별 할인한도 계산 함수
def get_discount_limits_living(total_usage):
    if total_usage >= 600000:
        return {'생활요금': 10000, '생활쇼핑': 10000, '대형마트': 10000}
    elif total_usage >= 300000:
        return {'생활요금': 5000, '생활쇼핑': 5000, '대형마트': 5000}
    else:
        return {'생활요금': 0, '생활쇼핑': 0, '대형마트': 0}

# 3. LIVING카드 혜택 계산 함수
def calculate_benefits_living(df, months):
    target_month = 202312
    benefits_dict = {}

    for customer_id, customer_data in df.groupby('고객ID'):
        total_benefit = 0

        for i in range(months):
            year = target_month // 100
            month = target_month % 100 - i
            if month <= 0:
                month += 12
                year -= 1
            current_month = year * 100 + month

            month_data = customer_data[customer_data['거래년월'] == current_month]

            # 실적 계산 (혜택 받은 거래 및 실적 제외 업종 제거)
            excluded_categories = ["PG상품권", "전자상거래상품권", "전자상거래상품권전문판매", "상품권전문판매", 
                                   "공공요금대행서비스/소득공제대상", "공공요금대행서비스/소득공제비대상", "생명보험", 
                                   "손해보험", "기타보험", "상 품 권", "조세서비스"]
            total_usage = month_data[~month_data['가맹점업종명'].isin(excluded_categories)]['승인금액'].sum()

            limits = get_discount_limits_living(total_usage)

            month_discount = 0

            # 생활요금 할인 계산 (10%)
            living_data = month_data[month_data['카테고리'] == '생활요금']
            living_discount = min(living_data['승인금액'].sum() * 0.1, limits['생활요금'])
            month_discount += living_discount

            # 생활쇼핑 할인 계산 (5%, 건당 1만원 이상)
            shopping_data = month_data[month_data['카테고리'] == '생활쇼핑']
            shopping_discount = 0
            for _, row in shopping_data.iterrows():
                if row['승인건수'] > 0:
                    amount_per_transaction = row['승인금액'] / row['승인건수']
                    if amount_per_transaction >= 10000:
                        shopping_discount += row['승인금액'] * 0.05
            shopping_discount = min(shopping_discount, limits['생활쇼핑'])
            month_discount += shopping_discount

            # 대형마트 할인 계산 (5%, 건당 3만원 이상)
            mart_data = month_data[month_data['카테고리'] == '대형마트']
            mart_discount = 0
            for _, row in mart_data.iterrows():
                if row['승인건수'] > 0:
                    amount_per_transaction = row['승인금액'] / row['승인건수']
                    if amount_per_transaction >= 30000:
                        mart_discount += row['승인금액'] * 0.05
            mart_discount = min(mart_discount, limits['대형마트'])
            month_discount += mart_discount

            total_benefit += month_discount

        # 연회비 차감
        annual_fee = 10000 * (months/12)
        net_benefit = total_benefit - annual_fee

        if customer_id not in benefits_dict:
            benefits_dict[customer_id] = {}

        benefits_dict[customer_id][f'LIVING카드_{months}개월혜택'] = net_benefit

    return benefits_dict

# ✅ 1. 업종 분류를 한 번만 실행
df['카테고리'] = df['가맹점업종명'].apply(classify_category_living)

# ✅ 2. 각 기간별 혜택 계산 (딕셔너리 활용 → 속도 최적화)
periods = [1, 3, 6, 12]
benefits_dict = {}

for period in periods:
    period_benefits = calculate_benefits_living(df.copy(), period)
    for key, value in period_benefits.items():
        if key not in benefits_dict:
            benefits_dict[key] = value
        else:
            benefits_dict[key].update(value)

# ✅ 3. 데이터 병합 최적화 (join 활용)
benefits_df = pd.DataFrame.from_dict(benefits_dict, orient='index').reset_index().rename(columns={'index': '고객ID'})
existing_result = pd.read_csv('final_card_benefits_ver2.csv').set_index('고객ID')
benefits_df = benefits_df.set_index('고객ID')

final_result = existing_result.join(benefits_df, how='left').reset_index()

# ✅ 4. CSV 저장
final_result.to_csv('final_card_benefits_ver3.csv', index=False, encoding='utf-8-sig')

print("파일 저장 완료: final_card_benefits_ver3.csv")
```

```python
def get_discount_limits_Super_DC_card(total_usage):
    if total_usage >= 700000:
        return {'KT통신': 15000, '커피할인율': 0.1}
    elif total_usage >= 300000:
        return {'KT통신': 7000, '커피할인율': 0.1}
    else:
        return {'KT통신': 0, '커피할인율': 0.0}
```

```python
def calculate_benefits_Super_DC_card(df, months, card_type):
    df['카테고리'] = df['가맹점업종명'].apply(classify_category)

    # ✅ 실적 제외 업종 리스트 추가
    excluded_categories = ["PG상품권", "전자상거래상품권", "전자상거래상품권전문판매", "상품권전문판매",
                           "공공요금대행서비스/소득공제대상", "공공요금대행서비스/소득공제비대상", "상 품 권", "조세서비스"]

    target_month = 202312
    benefits_list = []
    
    for customer_id in df['고객ID'].unique():
        customer_data = df[df['고객ID'] == customer_id]
        total_benefit = 0
        
        for i in range(months):
            year = target_month // 100
            month = target_month % 100 - i
            if month <= 0:
                month += 12
                year -= 1
            current_month = year * 100 + month
            
            month_data = customer_data[customer_data['거래년월'] == current_month]
            
            # ✅ 혜택 적용된 데이터 제외 후 실적 계산
            benefit_applied_data = month_data[month_data['카테고리'].isin(['이동통신'])]  
            eligible_data = month_data[~month_data.index.isin(benefit_applied_data.index)]  
            total_usage = eligible_data[~eligible_data['카테고리'].isin(excluded_categories)]['승인금액'].sum()
            
            limits = get_discount_limits_Super_DC_card(total_usage)
            
            month_discount = 0
            
            # ✅ KT 통신비 할인
            telecom_usage = month_data[month_data['카테고리'] == '이동통신']['승인금액'].sum()
            kt_discount = min(limits['KT통신'], telecom_usage)
            month_discount += kt_discount

            total_benefit += month_discount

        # ✅ 연간 1,500만 원 이상 사용 시, 3만 원 캐시백 추가 (최근 12개월 기준)
        if months == 12 and customer_data['승인금액'].sum() >= 15000000:
            total_benefit += 30000

        # ✅ 연회비 차감
        annual_fee = 8000 * (months / 12) if card_type == 'CUP' else 13000 * (months / 12)
        net_benefit = total_benefit - annual_fee
        
        benefits_list.append({'고객ID': customer_id, f'{months}개월_혜택': net_benefit})
    
    return pd.DataFrame(benefits_list)
```

```python
def get_discount_limits_Shopping_card(total_usage):
    if total_usage >= 1200000:
        return {'할인율': 0.1, '월간최대할인': 20000}
    elif total_usage >= 600000:
        return {'할인율': 0.07, '월간최대할인': 10000}
    elif total_usage >= 300000:
        return {'할인율': 0.05, '월간최대할인': 5000}
    else:
        return {'할인율': 0.0, '월간최대할인': 0}
```

```python
def calculate_benefits_Shopping_card(df, months, card_type):
    df['카테고리'] = df['가맹점업종명'].apply(classify_category)

    # ✅ 실적 제외 업종 리스트 추가
    excluded_categories = ["CATV홈쇼핑", "PG상품권", "전자상거래상품권", "전자상거래상품권전문판매", "상품권전문판매",
                           "공공요금대행서비스/소득공제대상", "공공요금대행서비스/소득공제비대상", "상 품 권", "조세서비스"]

    target_month = 202312
    benefits_list = []

    for customer_id in df['고객ID'].unique():
        customer_data = df[df['고객ID'] == customer_id]
        total_benefit = 0

        for i in range(months):
            year = target_month // 100
            month = target_month % 100 - i
            if month <= 0:
                month += 12
                year -= 1
            current_month = year * 100 + month

            month_data = customer_data[customer_data['거래년월'] == current_month]

            # ✅ 혜택 적용된 데이터 제외 후 실적 계산
            benefit_applied_data = month_data[month_data['카테고리'].isin(
                ['온라인쇼핑', '편의점', '커피', '면세점', '백화점', '대형마트', '슈퍼마켓', '화장품', '의류', '안경']
            )]
            eligible_data = month_data[~month_data.index.isin(benefit_applied_data.index)]  
            total_usage = eligible_data[~eligible_data['카테고리'].isin(excluded_categories)]['승인금액'].sum()

            limits = get_discount_limits_Shopping_card(total_usage)

            month_discount = 0
            category_discount = 0

            # ✅ 온라인쇼핑/편의점/커피/면세점 할인 (5~10%)
            eligible_categories_1 = ["온라인쇼핑", "편의점", "커피", "면세점"]
            for category in eligible_categories_1:
                category_data = month_data[(month_data['카테고리'] == category) & (month_data['승인금액'] >= 10000)]
                category_discount += category_data['승인금액'].sum() * limits['할인율']

            # ✅ 백화점/대형마트/슈퍼마켓/화장품/의류/안경 할인 (5~10%)
            eligible_categories_5 = ["백화점", "대형마트", "슈퍼마켓", "화장품", "의류", "안경"]
            for category in eligible_categories_5:
                category_data = month_data[(month_data['카테고리'] == category) & (month_data['승인금액'] >= 50000)]
                category_discount += category_data['승인금액'].sum() * limits['할인율']

            month_discount = min(category_discount, limits['월간최대할인'])

            # ✅ 영화 할인 (건당 4천 원 정액 할인)
            movie_data = month_data[(month_data['카테고리'] == '영화') & (month_data['승인금액'] >= 16000)]
            if not movie_data.empty:
                month_discount += 4000  # 정액 할인 적용

            total_benefit += month_discount

        # ✅ 연회비 차감
        annual_fee = 5000 if card_type == 'BC Global' else 8000
        if customer_data['승인금액'].sum() >= 5000000:
            annual_fee = 0  # 500만 원 이상 사용 시 연회비 면제

        benefits_list.append({'고객ID': customer_id, f'{months}개월_혜택': total_benefit - annual_fee})

    return pd.DataFrame(benefits_list)
```

```python
import pandas as pd

# ✅ 국민행복카드 업종 분류 함수
def classify_category_happycard(business_type):
    excluded_categories = [
        "병 원(응급실운영)", "의 원", "조산원", "산후조리원", "어린이집", "유치원", "유아원", 
        "제세공과금", "상품권", "치과의원", "치과병원", "한 의 원", "한방병원", "의 원", "종합병원", "유아원"
    ]
    
    if business_type in excluded_categories:
        return "실적제외"
    elif business_type in ["기타음료식품"]:
        return "커피"
    elif business_type in ["서양음식"]:
        return "패밀리레스토랑"
    elif business_type in ["인터넷Mall", "인터넷종합Mall", "CATV홈쇼핑"]:
        return "온라인쇼핑"
    elif business_type in ["종합레져타운"]:
        return "놀이공원"
    elif business_type == "이동통신요금":
        return "이동통신"
    elif business_type in ["병 원(응급실운영)", "의 원", "치과의원", "치과병원"]:
        return "의료"
    elif business_type in ["유아원", "학습지교육"]:
        return "육아/교육"
    else:
        return None

# ✅ 국민행복카드 할인한도 계산 함수
def get_discount_limits_happycard(total_usage):
    if total_usage >= 2000000:
        return {"월간통합한도": 30000}
    elif total_usage >= 1000000:
        return {"월간통합한도": 20000}
    elif total_usage >= 500000:
        return {"월간통합한도": 10000}
    elif total_usage >= 300000:
        return {"월간통합한도": 5000}
    else:
        return {"월간통합한도": 0}


# ✅ 국민행복카드 혜택 계산 함수 (A/B/C 타입 포함)
def calculate_benefits_happycard(df, months):
    df['카테고리'] = df['가맹점업종명'].apply(classify_category_happycard)
    target_month = 202312
    benefits_list = []

    for customer_id in df['고객ID'].unique():
        customer_data = df[df['고객ID'] == customer_id]
        max_benefit = 0

        for i in range(months):
            year = target_month // 100
            month = target_month % 100 - i
            if month <= 0:
                month += 12
                year -= 1
            current_month = year * 100 + month

            month_data = customer_data[customer_data['거래년월'] == current_month]
            total_usage = month_data[~month_data['카테고리'].isin(["실적제외"])]['승인금액'].sum()
            limits = get_discount_limits_happycard(total_usage)

            # ✅ A/B/C 타입별 혜택 계산
            type_benefits = {"A": 0, "B": 0, "C": 0}

            # ✅ A/B 타입 (기존)
            for card_type in ["A", "B"]:
                month_discount = 0

                # 의료(A) / 육아(B) 할인 (5% / 월 2회, 최대 2만원)
                category_name = "의료" if card_type == "A" else "육아/교육"
                category_data = month_data[month_data["카테고리"] == category_name]
                discount = (category_data['승인금액'] / category_data['승인건수']).apply(lambda x: x * 0.05)
                discount = discount.nlargest(2).sum()  # 월 2회 제한
                month_discount += min(discount, 20000)

                # 온라인 쇼핑 할인 (5% / 월 2회, 건당 3만원 이상)
                shopping_data = month_data[(month_data["카테고리"] == "온라인쇼핑") & (month_data['승인금액'] / month_data['승인건수'] >= 30000)]
                discount = (shopping_data['승인금액'] * 0.05).nlargest(2).sum()
                month_discount += discount

                # 패밀리레스토랑 (10% / 월 1회)
                restaurant_data = month_data[month_data["카테고리"] == "패밀리레스토랑"]
                discount = (restaurant_data['승인금액'] * 0.1).nlargest(1).sum()
                month_discount += discount

                # ✅ 월간 통합 한도 적용
                month_discount = min(month_discount, limits["월간통합한도"])
                type_benefits[card_type] += month_discount

            # ✅ C 타입 (포인트 적립)
            type_c_benefit = 0

            # 기본 적립 (국내 가맹점 0.2%~0.8%)
            if total_usage >= 1500000:
                base_rate = 0.008
            elif total_usage >= 1000000:
                base_rate = 0.006
            elif total_usage >= 500000:
                base_rate = 0.004
            elif total_usage >= 200000:
                base_rate = 0.002
            else:
                base_rate = 0.0
            type_c_benefit += total_usage * base_rate  # 기본 적립 반영

            # 의료 & 육아 추가 적립 (1%~4%, 업종별 월 5천점 한도, 통합 월 1만점 한도)
            extra_rate = 0.04 if total_usage >= 1500000 else 0.03 if total_usage >= 1000000 else 0.02 if total_usage >= 500000 else 0.01
            medical_education_data = month_data[month_data["카테고리"].isin(["의료", "육아/교육"])]
            extra_points = (medical_education_data["승인금액"] * extra_rate).sum()
            extra_points = min(extra_points, 5000)  # 업종별 월 5000점 한도 적용
            type_c_benefit += min(extra_points, 10000)  # 통합 월 1만점 한도 적용

            # KTX, 고속버스 5% 적립 (월 2회 한도)
            ktx_data = month_data[month_data["카테고리"] == "대중교통"]
            discount = (ktx_data['승인금액'] * 0.05).nlargest(2).sum()  # 월 2회 제한
            type_c_benefit += discount

            # ✅ 가장 높은 혜택 선택 (A/B/C 타입 중 최댓값)
            max_benefit += max([type_benefits["A"], type_benefits["B"], type_c_benefit])

        # ✅ 연회비 차감
        annual_fee = 5000 * (months/12)
        net_benefit = max_benefit - annual_fee

        benefits_list.append({
            '고객ID': customer_id,
            f'국민행복카드_{months}개월혜택': net_benefit
        })

    return pd.DataFrame(benefits_list)

# ✅ 실행
df = pd.read_csv("card_data_2023_sample1_utf8.csv")
result_happy_1 = calculate_benefits_happycard(df, 1)
result_happy_3 = calculate_benefits_happycard(df, 3)
result_happy_6 = calculate_benefits_happycard(df, 6)
result_happy_12 = calculate_benefits_happycard(df, 12)

# ✅ 최종 병합 & 저장
final_result = result_happy_1.merge(result_happy_3, on="고객ID").merge(result_happy_6, on="고객ID").merge(result_happy_12, on="고객ID")
final_result.to_csv("국민행복카드_최종.csv", index=False, encoding="utf-8-sig")

print("국민행복카드 혜택 계산 완료!")

```

```python
def calculate_benefits_Apartment_card(df, months):
    df["카테고리"] = df["가맹점업종명"].apply(classify_category)

    # ✅ 실적 제외 업종 리스트 추가
    excluded_categories = ["PG상품권", "전자상거래상품권", "전자상거래상품권전문판매", "상품권전문판매",
                           "공공요금대행서비스/소득공제대상", "공공요금대행서비스/소득공제비대상", "상 품 권", "조세서비스"]

    target_month = 202312
    benefits_list = []

    for customer_id in df["고객ID"].unique():
        customer_data = df[df["고객ID"] == customer_id]
        total_benefit = 0

        for i in range(months):
            year = target_month // 100
            month = target_month % 100 - i
            if month <= 0:
                month += 12
                year -= 1
            current_month = year * 100 + month

            month_data = customer_data[customer_data["거래년월"] == current_month]

            # ✅ 혜택 적용된 데이터 제외 후 실적 계산
            benefit_applied_data = month_data[month_data["카테고리"].isin(
                ["아파트관리비", "병원", "약국", "베이커리/디저트"]
            )]
            eligible_data = month_data[~month_data.index.isin(benefit_applied_data.index)]
            total_usage = eligible_data[~eligible_data["카테고리"].isin(excluded_categories)]["승인금액"].sum()

            # ✅ 통합 할인 한도 설정
            if total_usage >= 600000:
                total_limit = 20000
                apt_discount = 10000  # 아파트관리비 할인 최대 1만 원
            elif total_usage >= 300000:
                total_limit = 10000
                apt_discount = 5000  # 아파트관리비 할인 최대 5천 원
            else:
                total_limit = 0
                apt_discount = 0

            month_discount = 0

            # ✅ 아파트관리비 할인 적용
            apt_data = month_data[month_data["카테고리"] == "아파트관리비"]
            apt_discount = min(apt_discount, apt_data["승인금액"].sum())

            # ✅ 생활편의 할인 (5%) 적용 (건당 1만 원 이상)
            convenience_discount = 0
            for category in ["병원", "약국", "베이커리/디저트"]:
                category_data = month_data[(month_data["카테고리"] == category) & (month_data["승인금액"] >= 10000)]
                convenience_discount += category_data["승인금액"].sum() * 0.05

            # ✅ 생활편의 할인 한도 적용 (월 최대 5천 원)
            convenience_discount = min(convenience_discount, 5000)

            # ✅ 통합 할인 한도 적용
            month_discount = min(apt_discount + convenience_discount, total_limit)
            total_benefit += month_discount

        benefits_list.append({"고객ID": customer_id, f"부자되세요_{months}개월혜택": total_benefit})

    return pd.DataFrame(benefits_list)
```

```python
import pandas as pd
from tqdm import tqdm  # 진행상황 시각화 모듈

# ✅ 사용할 카드 목록 (Super DC, iM 쇼핑은 card_type이 필요하므로 람다 함수 사용)
card_type = "VISA"
cards = [
    ("iM_올레_Super_DC_카드", lambda df, months: calculate_benefits_Super_DC_card(df, months, card_type)),
    ("iM_쇼핑카드", lambda df, months: calculate_benefits_Shopping_card(df, months, card_type)), 
    ("부자되세요_아파트카드", calculate_benefits_Apartment_card)
]

# ✅ sample1 파일만 로드
file_name = 'card_data_2023_sample1_utf8.csv'
df = pd.read_csv(file_name)

```
