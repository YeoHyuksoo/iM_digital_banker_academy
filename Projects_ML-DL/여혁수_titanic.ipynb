(문제2) 타이타닉 데이터를 사용하여 생존여부(Survived)를 분류하는 딥러닝 모델을 만드시오.
- 처리 조건 -
(1) titanic 폴더에 있는 훈련데이터(train.csv)와 테스트 데이터(test.csv)를 사용한다.
(2) 성능을 최대한 높일수 있도록 모델을 설계한다.
(3) 학습한 모델을 titanic.keras로 저장한다. 그리고 불러와서 테스트 데이터에 대하여 평가(evaluate()사용)한다.
(4) 저장 문서 : 홍길동_titanic.ipynb

```python
import pandas as pd

train_data = pd.read_csv("./titanic/train.csv")

train_data
```

```python
test_data = pd.read_csv("./titanic/test.csv")

test_data
```

```python
survived = pd.read_csv("./titanic/gender_submission.csv")
```

```python
test_data = pd.merge(left = test_data, right = survived)
```

```python
train_data['Age'] = train_data['Age'].fillna(train_data['Age'].median())
test_data['Age'] = test_data['Age'].fillna(test_data['Age'].median())

train_data['Fare'] = train_data['Fare'].fillna(train_data['Fare'].median())
test_data['Fare'] = test_data['Fare'].fillna(test_data['Fare'].median())

train_data = train_data.drop(['Cabin', 'Name', 'Ticket', 'PassengerId'], axis = 1)
test_data = test_data.drop(['Cabin', 'Name', 'Ticket', 'PassengerId'], axis = 1)

train_data['Embarked'] = train_data['Embarked'].fillna(train_data['Embarked'].mode()[0])
test_data['Embarked'] = test_data['Embarked'].fillna(test_data['Embarked'].mode()[0])
```

```python
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
train_data['Sex'] = le.fit_transform(train_data['Sex'])
test_data['Sex'] = le.fit_transform(test_data['Sex'])
```

```python
train_data = pd.get_dummies(train_data, columns=['Embarked'], drop_first=True)
test_data = pd.get_dummies(test_data, columns=['Embarked'], drop_first=True)
```

```python
train_data
```

```python
X_train = train_data.drop('Survived', axis = 1)
y_train = train_data['Survived']

X_test = test_data.drop('Survived', axis = 1)
y_test = test_data['Survived']
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
import np_utils
from keras.utils import to_categorical
import matplotlib.pyplot as plt
```

```python
model = Sequential()

model.add(Dense(256, input_shape = (8, )))
model.add(Activation('relu'))
model.add(Dropout(0.2))

# model.add(Dense(128))
# model.add(Activation('relu'))
# model.add(Dropout(0.2))

model.add(Dense(128))
model.add(Activation('relu'))
model.add(Dropout(0.2))

model.add(Dense(32))
model.add(Activation('relu'))
model.add(Dropout(0.1))

model.add(Dense(1))
model.add(Activation('sigmoid')) # 이진분류

model.summary()
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = 'accuracy')

hist = model.fit(X_train, y_train, epochs = 100)
```

```python
# results = model.evaluate(X_test, y_test, verbose = 2)
# print("loss = ", results[0])
# print("accuracy = ", results[1])
```

```python
filename = "titanic2.keras"
model.save(filename)
```

```python
import tensorflow as tf

load_model = tf.keras.models.load_model(filename)

test_loss, test_acc = load_model.evaluate(X_test, y_test)

# 평가 결과 출력
print(f"Test Loss: {test_loss}")
print(f"Test ACC: {test_acc}")
```
