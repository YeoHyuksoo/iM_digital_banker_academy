### RNN 이용해서 텍스트 생성

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
from tensorflow.keras.utils import to_categorical
```

```python
text = """경마장에 있는 말이 뛰고 있다\n그의 말이 법이다\n가는 말이 고와야 오는 말이 곱다"""
```

```python
t = Tokenizer()
t.fit_on_texts([text])
vocab_size = len(t.word_index) + 1
print("단어 집합의 크기 : %d" % vocab_size)
```

```python
t.word_index
```

```python
sequences = list()
for line in text.split('\n'):
    encoded = t.texts_to_sequences([line])[0]
    for i in range(1, len(encoded)):
        sequence = encoded[:i+1]
        sequences.append(sequence)

print("학습이 사용할 샘플의 개수 : %d" % len(sequences))
```

```python
t.texts_to_sequences([text.split('\n')[0]])
```

```python
max_len = max(len(l) for l in sequences)
```

```python
sequences = pad_sequences(sequences, maxlen = max_len, padding = 'pre')
sequences
```

```python
# sequences = np.array(sequences)

X = sequences[:, :-1]
y = sequences[:, -1]
```

```python
X
```

```python
y
```

```python
y = to_categorical(y, num_classes = vocab_size)
y
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Dense, SimpleRNN
```

```python
model = Sequential()
model.add(Embedding(vocab_size, 6, input_length = max_len-1))
model.add(SimpleRNN(32))
model.add(Dense(vocab_size, activation='softmax'))
model.summary()
```

```python
model = Sequential()
model.add(Embedding(vocab_size, 6))
model.add(SimpleRNN(32))
model.add(Dense(vocab_size, activation='softmax'))
model.build(input_shape = (None, max_len-1))
model.summary() # tensorflow 2.16 이상 버전에서는 이 형식만 가능
```

```python
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
model.fit(X, y, epochs = 200, verbose = 2)
```

```python
def sentence_generation(model, t, current_word, n):
    # init_word = current_word
    sentence=  ''
    for _ in range(n):
        encoded = t.texts_to_sequences([current_word])[0]
        encoded = pad_sequences([encoded], maxlen = max_len-1, padding = 'pre')
        temp = model.predict(encoded)
        result = np.argmax(temp, axis = 1)
        for word, index in t.word_index.items():
            if index == result:
                break

        current_word = current_word + ' ' + word
        sentence = sentence + ' ' + word

    # sentence = init_word + sentence
    sentence = current_word
    return sentence
    
```

```python
print(sentence_generation(model, t, '말이', 10))
```

#### LSTM을 이용하여 텍스트 생성

```python
import pandas as pd
from string import punctuation
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
from tensorflow.keras.utils import to_categorical
```

```python
df = pd.read_csv("ArticlesApril2018.csv")
df.head()
```

```python
df.info()
```

```python
print("열의 개수: ", len(df.columns))
df.columns
```

```python
df['headline'].isnull().values.any()
```

```python
headline = []
# 헤드라인의 값들을 리스트로 저장
headline.extend(list(df.headline.values))
headline[:5]
```

```python
print('총 샘플의 개수 : {}'.format(len(headline)))
```

```python
headline = [word for word in headline if word != "Unknown"]
print('노이즈값 제거 후 샘플의 개수 : {}'.format(len(headline)))
```

```python
headline[:5]
```

```python
punctuation
```

```python
def repreprocessing(s):
    s = s.encode("utf8").decode("ascii", "ignore")
    return ''.join(c for c in s if c not in punctuation).lower()
```

```python
text = [repreprocessing(x) for x in headline]
text[:5]
```

```python
t = Tokenizer()
t.fit_on_texts(text)
vocab_size = len(t.word_index) + 1
print("단어 집합의 크기 : %d" % vocab_size)
```

```python
t.word_index
```

```python
sequences = list()
for line in text:
    #print(t.texts_to_sequences([line]))
    encoded = t.texts_to_sequences([line])[0]
    # print(encoded)
    for i in range(1, len(encoded)):
        sequence = encoded[:i+1]
        # print(sequence)
        sequences.append(sequence)
        
print("학습에 사용할 샘플의 개수: %d" % len(sequences))
```

```python
sequence
```

```python
max_len = max(len(l) for l in sequences)
print("샘플의 최대 길이 : {}".format(max_len))
```

```python
sequences = pad_sequences(sequences, maxlen = max_len, padding = 'pre')
```

```python
sequences
```

```python
sequences = np.array(sequences)
X = sequences[:, :-1]
y = sequences[:, -1]
```

```python
y = to_categorical(y, num_classes = vocab_size)
y[1]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Dense, LSTM, Bidirectional
```

```python
model = Sequential()
model.add(Embedding(vocab_size, 100))
model.add(Bidirectional(LSTM(128)))
model.add(Dense(vocab_size, activation = 'softmax'))
model.build(input_shape = (None, max_len-1))
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, y, epochs=30, verbose=2)
```

```python
def sentence_generation(model, t, current_word, n):
    # init_word = current_word
    sentence=  ''
    for _ in range(n):
        encoded = t.texts_to_sequences([current_word])[0]
        encoded = pad_sequences([encoded], maxlen = max_len-1, padding = 'pre')
        temp = model.predict(encoded)
        result = np.argmax(temp, axis = 1)
        for word, index in t.word_index.items():
            if index == result:
                break

        current_word = current_word + ' ' + word
        sentence = sentence + ' ' + word

    # sentence = init_word + sentence
    sentence = current_word
    return sentence
    
```

```python
print(sentence_generation(model, t, 'i', 10))
```
