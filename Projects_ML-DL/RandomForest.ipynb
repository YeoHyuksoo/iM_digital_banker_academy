##### 랜덤 포레스트

```python
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
```

```python
iris = datasets.load_iris()
print(iris.target_names)
print("데이터 수 : ", len(iris.data))
print("데이터 열 이름 : ", iris.feature_names)
```

```python
data = pd.DataFrame(
    {
        'sepal_length': iris.data[:, 0],
        'sepal_width': iris.data[:, 1],
        'petal_length': iris.data[:, 2],
        'petal_width': iris.data[:, 3],
        'species': iris.target
    })
data.head()
```

```python
x = data[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y = data['species']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
print(x_train.shape, y_train.shape)
```

```python
forest = RandomForestClassifier(n_estimators = 100)
forest.fit(x_train, y_train)
```

```python
y_pred = forest.predict(x_test)
y_pred
```

```python
print("정확도 : ", metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

feature_importances_values = forest.feature_importances_
feature_importances = pd.Series(feature_importances_values, index = x_train.columns)
feature_top4 = feature_importances.sort_values(ascending = False)[:4]
print(feature_top4)

plt.figure(figsize = (8, 6))
plt.title("Top 4 feature Importances")
sns.barplot(x = feature_top4, y = feature_top4.index)
plt.show()
```

```python
df = pd.read_csv("boston_house.csv")
df.head(2)
```

```python
X = df.drop('Target', axis = 1)
```

```python
y = df['Target']
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
reg = RandomForestRegressor(
    n_estimators = 50,
    criterion = 'squared_error',
    max_depth = 4,
    max_features = 1/3,
    max_samples = 1.0,
    bootstrap = True,
    random_state = 100
).fit(X, y)
```

```python
print(reg.predict(X)[:3])
```

```python
print('정확도 : ', reg.score(X, y)) # r2score - 결정계수
```

```python
print(reg.feature_importances_)
```
