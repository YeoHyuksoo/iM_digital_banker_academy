1.6 열의 이름을 사용하여 여러 개 열을 추출하고 DataFrame 객체 생성

```python
df = data[['INDUS', 'AGE']]
df.info()
```

1.7 열의 번호를 사용하여 여러 개 열을 추출하고 DataFrame 객체 생성

```python
df = data.iloc[:, 2:5]
df.info()
```

1.8 df.iloc[행, 열]을 사용하여 원소에 접근하고 수정

```python
df = data
df.iloc[5, 5]
```

```python
df.iloc[5, 5] = 27.3
df.head(7)
```

1.9 수정된 DataFrame을 새로운 csv파일로 저장

```python
df.to_csv("boston_train_update.csv")
```

## 3. Boston Housing Data Set 조건 검색 출력
### boston_train.csv 파일을 사용하여 boolean 조건 검색 결과를 DataFrame으로 출력하는 코드를 구현하세요

3.1 CRIM 열의 평균값보다 큰 값들만 출력

```python
df[df['CRIM'] > df['CRIM'].mean()]
```

3.2 'AGE' 열의 평균값보다 작은 값들만 출력

```python
df[df['AGE'] < df['AGE'].mean()]
```

3.3  'MEDV' 열의 중앙값 보다 작은 값들만 출력

```python
df3 = df[df['MEDV'] < df['MEDV'].median()]
df3
```

## 4. Boston Housing Data Set 데이터 연쇄하기
- boston_train.csv 파일과 boston_test.csv 파일을 pandas로 모두 읽어서 <br>
- boston_train.csv  에서는 처음 10개 행과  boston_test.csv 의 처음
10개 행을 차례대로 하나로 합쳐 <br>
- 결과를 출력하고 <br>
- boston_batch.csv 파일로 인덱스 값을 제외하고
저장하는 코드를 구현하세요


```python
train_df = pd.read_csv("boston_train.csv")
test_df = pd.read_csv("boston_test.csv")
```

```python
augment_df = pd.concat([train_df.head(10), test_df.head(10)], axis = 0)
```

```python
augment_df
```

```python
augment_df = augment_df.reset_index(drop = True)
augment_df
```

```python
augment_df.to_csv("boston_batch.csv", header=False, index=False, mode='w')
data = pd.read_csv("boston_batch.csv")

data.head()
```
