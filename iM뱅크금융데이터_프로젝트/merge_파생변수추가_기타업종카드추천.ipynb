```python
import pandas as pd

def preprocess_card_data(file_path):
    # 데이터 로드
    df = pd.read_csv(file_path, encoding='utf-8')

    # 거래년월을 datetime 형식으로 변환
    df['거래년월'] = pd.to_datetime(df['거래년월'], format='%Y%m')

    # 최신 거래년월 찾기 (2023-12)
    latest_month = df['거래년월'].max()

    # 최근 n개월 설정
    periods = {1: 1, 3: 3, 6: 6, 12: 12}
    result_dfs = []

    for n, months in periods.items():
        start_date = latest_month - pd.DateOffset(months=months-1)
        df_n = df[(df['거래년월'] >= start_date) & (df['거래년월'] <= latest_month)]

        # 고객ID별 업종 소비 금액 합계 계산
        grouped = df_n.groupby(['고객ID', '가맹점업종명'])['승인금액'].sum().reset_index()

        # 소비 비율 계산
        total_spending = grouped.groupby('고객ID')['승인금액'].transform('sum')
        grouped['소비비율'] = grouped['승인금액'] / total_spending

        # TOP5 업종 선택
        top5 = grouped.sort_values(['고객ID', '승인금액'], ascending=[True, False])
        top5 = top5.groupby('고객ID').head(5)

        # Pivot하여 TOP5 업종 및 비율 컬럼화
        top5['순위'] = top5.groupby('고객ID').cumcount() + 1
        pivoted = top5.pivot(index='고객ID', columns='순위', values=['가맹점업종명', '소비비율'])
        pivoted.columns = [f'n{n}_TOP{col[1]}_{col[0]}' for col in pivoted.columns]
        pivoted.reset_index(inplace=True)

        result_dfs.append(pivoted)

    # 📌 고객별 **월별 총 승인금액(2023년 1~12월)** 계산
    df['년월'] = df['거래년월'].dt.strftime('%Y-%m')
    monthly_spending = df.groupby(['고객ID', '년월'])['승인금액'].sum().unstack().reset_index()

    # 컬럼명 변경 (월별 총 승인금액)
    monthly_spending.columns = ['고객ID'] + [f'월별총승인금액_{col.replace("-", "_")}' for col in monthly_spending.columns[1:]]
    
    # 모든 데이터 병합 (업종별 소비 패턴 + 월별 총 승인금액)
    final_df = result_dfs[0]
    for df_n in result_dfs[1:]:
        final_df = pd.merge(final_df, df_n, on='고객ID', how='outer')

    final_df = pd.merge(final_df, monthly_spending, on='고객ID', how='outer')

    return final_df

# 사용 예시
file_path = "card_data_2023_sample1_utf8.csv"
final_result = preprocess_card_data(file_path)
print(final_result.head())
```

```python
import pandas as pd

def add_financial_data(card_data_path, financial_data_path, output_path):
    # 1️⃣ 카드 소비 분석 데이터 로드
    final_result = pd.read_csv(card_data_path)

    # 2️⃣ 금융상품 데이터 로드
    financial_data = pd.read_csv(financial_data_path, encoding='cp949')

    # 3️⃣ 기준년월을 datetime으로 변환 후 'YYYY-MM' 형식으로 변환
    financial_data['기준년월'] = pd.to_datetime(financial_data['기준년월'], format='%Y%m')
    financial_data['년월'] = financial_data['기준년월'].dt.strftime('%Y-%m')

    # 4️⃣ 고객 ID별 월별 요구불예금 금액을 정리
    demand_deposit = financial_data.pivot_table(
        index='고객ID',
        columns='년월',
        values='수신_요구불예금',
        aggfunc='sum'
    ).reset_index()

    # 5️⃣ 컬럼명 변경 (요구불예금_2023_01, 요구불예금_2023_02, ...)
    demand_deposit.columns = ['고객ID'] + [f'요구불예금_{col.replace("-", "_")}' for col in demand_deposit.columns[1:]]

    # 6️⃣ 기존 final_result와 병합
    final_result = pd.merge(final_result, demand_deposit, on='고객ID', how='left')

    # 7️⃣ 최종 데이터 저장
    final_result.to_csv(output_path, index=False, encoding='utf-8-sig')

    return final_result

# 사용 예시
card_data_path = "고객별 소비TOP5업종_월별총승인금액_sample1.csv"  # 기존 카드 소비 분석 데이터
financial_data_path = "iMBANK_CUSTOMER_DATA_2023(K-DigitalTraining).csv"  # 고객 금융상품 데이터
output_path = "final_result_with_earning.csv"  # 최종 결과 저장

final_result_with_financial = add_financial_data(card_data_path, financial_data_path, output_path)
print(final_result_with_financial.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 정육점 & 농축수산품 업종 필터링
target_categories = ["정육점", "농·축·수산품"]

# 승인건수를 숫자로 변환
card_2023["승인건수"] = pd.to_numeric(card_2023["승인건수"], errors="coerce")
target_data = card_2023[card_2023["가맹점업종명"].isin(target_categories)].copy()

# 건당 소비 금액 계산 (승인금액 / 승인건수)
target_data["건당 소비 금액"] = target_data["승인금액"] / target_data["승인건수"]

# 연령대별 건당 평균 소비 금액 계산
age_group_avg = target_data.groupby("연령대")["건당 소비 금액"].mean().reset_index()

# 시각화 - 연령대별 건당 평균 소비 금액
plt.figure(figsize=(8,5))
sns.barplot(data=age_group_avg, x="연령대", y="건당 소비 금액", palette="Blues_r")
plt.title("연령대별 정육점 & 농축수산품 건당 소비 금액 평균")
plt.ylabel("건당 평균 소비 금액 (원)")
plt.xlabel("연령대")
plt.show()

# 시각화 - 건당 소비 금액 분포 (히스토그램)
plt.figure(figsize=(8,5))
sns.histplot(target_data["건당 소비 금액"], bins=30, kde=True)
plt.title("정육점 & 농축수산품 건당 소비 금액 분포")
plt.xlabel("건당 소비 금액 (원)")
plt.ylabel("건수")
plt.show()

# 특정 구간(1만 원, 3만 원, 5만 원 이상)별 비율 계산
bins = [0, 10000, 30000, 50000, target_data["건당 소비 금액"].max()]
labels = ["1만 원 이하", "1~3만 원", "3~5만 원", "5만 원 이상"]
target_data["소비 구간"] = pd.cut(target_data["건당 소비 금액"], bins=bins, labels=labels, right=False)

# 각 구간별 비율 계산
category_ratio = target_data["소비 구간"].value_counts(normalize=True) * 100
print(category_ratio)
```
