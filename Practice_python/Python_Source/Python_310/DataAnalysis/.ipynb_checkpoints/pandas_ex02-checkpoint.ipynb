#### 타이타닉 생존자 컬럼 정보
- PassengerId: 승객의 아이디.
- Survived: 생존여부 
- Pclass : 승객 등급
- Name: 승객 이름
- Sex: 성별
- Age: 나이
- SibSp: 동승한 배우자와 형제(자매)수
- Parch: 동승한 부모님 수와 자식수
- Ticket: 티켓 번호
- Fare: 운임 요금
- Cabin: 객실 구역
- Embarked: 어디서 탑승했는지 등에 대한 정보
