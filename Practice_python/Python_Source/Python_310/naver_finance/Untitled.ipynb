```python
import plotly.graph_objects as go

fig = go.Figure()


# 거래량에 대한 막대형 차트를 추가합니다.

fig.add_trace(go.Bar(x=stock_ligNex1.index, y=stock_ligNex1['Volume'], marker_color='purple'))


# 더 나은 시각적 모양을 위해 레이아웃을 업데이트하세요.

fig.update_layout(
    title='Trading Volume for last 1 year',
    xaxis_title='Date',
    yaxis_title='Volume',
    template='plotly_white',
    xaxis_rangeslider_visible=True  # Enable the range slider for zooming
)
```
